### Bedingter Ausdruck: kurze Schreibweise für `if-else`

Für eine einfache Entscheidung kann Python eine kurze Schreibweise verwenden.

Der Fachbegriff lautet:

> **bedingter Ausdruck**

Umgangssprachlich wird oft auch vom **ternären Operator** gesprochen.

### Normale Schreibweise mit `if-else`

````python
alter = 17

if alter >= 18:
    status = "Erwachsen"
else:
    status = "Minderjährig"

print(status)
````

### Kurze Schreibweise

````python
alter = 17

status = "Erwachsen" if alter >= 18 else "Minderjährig"

print(status)
````

## Bedingter Ausdruck

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```

Man liest die Zeile so:

> Verwende `"Erwachsen"`, wenn `alter >= 18` wahr ist, sonst `"Minderjährig"`.

### Vergleich mit Java

Java verwendet dafür die klassische ternäre Schreibweise:

```java
String status = alter >= 18
        ? "Erwachsen"
        : "Minderjährig";
```

Der Aufbau in Java lautet:

```text
bedingung ? wert_wenn_wahr : wert_wenn_falsch
```

Python verwendet dagegen:

```text
wert_wenn_wahr if bedingung else wert_wenn_falsch
```

### Wann ist der bedingte Ausdruck sinnvoll?

Gut geeignet:

- eine kurze Zuweisung
- genau zwei einfache Ergebnisse
- eine Bedingung, die sofort verständlich bleibt

````python
zahl = 8

art = "gerade" if zahl % 2 == 0 else "ungerade"

print(art)
````

Weniger geeignet ist er bei langen oder verschachtelten Entscheidungen.

Das funktioniert zwar:

````python
alter = 15

status = (
    "Erwachsen"
    if alter >= 18
    else "Jugendlich"
    if alter >= 14
    else "Kind"
)

print(status)
````

Diese verschachtelte Kurzform ist für Anfänger und spätere Wartung schwerer zu lesen.

Hier ist eine normale Kaskade besser:

````python
alter = 15

if alter >= 18:
    status = "Erwachsen"
elif alter >= 14:
    status = "Jugendlich"
else:
    status = "Kind"

print(status)
````

★ **Merksatz**

> Der bedingte Ausdruck ist gut für eine kurze Entscheidung mit genau zwei übersichtlichen Ergebnissen.

### `match-case`: Python-Alternative zu `switch-case`

In Java gibt es `switch-case`.

Python verwendet seit Version 3.10:

```python
match
case
```

`match-case` eignet sich besonders, wenn ein Wert gegen mehrere feste Fälle geprüft wird.

### Beispiel mit Schulnoten

````python
note = 2

match note:
    case 1:
        print("Sehr gut")
    case 2:
        print("Gut")
    case 3:
        print("Befriedigend")
    case _:
        print("Andere Note")
````

Das `_` steht für:

> alle übrigen Fälle

Es entspricht ungefähr `default` bei Java.

### Dasselbe mit `if-elif-else`

````python

if note == 1:
    print("Sehr gut")
elif note == 2:
    print("Gut")
elif note == 3:
    print("Befriedigend")
else:
    print("Andere Note")
````

Beide Varianten sind richtig.

`match-case` ist oft übersichtlicher, wenn immer derselbe Wert mit festen Möglichkeiten verglichen wird.

### Beispiel mit Befehlen

````python
befehl = "start"

match befehl:
    case "start":
        print("Programm startet")
    case "stop":
        print("Programm stoppt")
    case "pause":
        print("Programm pausiert")
    case _:
        print("Unbekannter Befehl")
````

### Mehrere Werte in einem Fall

Mit `|` können mehrere feste Werte zusammengefasst werden:

````python
tag = "Samstag"

match tag:
    case "Samstag" | "Sonntag":
        print("Wochenende")
    case _:
        print("Werktag")
````

Hier bedeutet `|`:

> oder

### `match-case` mit Guard

Ein **Guard** ist eine zusätzliche Bedingung hinter `if`.

````python
alter = 20

match alter:
    case wert if wert >= 18:
        print("Volljährig")
    case _:
        print("Minderjährig")
````

Diese Schreibweise ist möglich.

Für einen einfachen Zahlenbereich ist normales `if` aber meist leichter zu lesen:

````python
alter = 20

if alter >= 18:
    print("Volljährig")
else:
    print("Minderjährig")
````

### Wann `match-case` sinnvoll ist

Gut geeignet:

- feste Befehle
- Statuswerte
- Menüauswahl
- einzelne Zahlen oder Texte
- bestimmte Muster in Daten

Weniger geeignet:

- Zahlenbereiche
- viele Vergleichsoperatoren
- komplexe Bedingungen mit `and` und `or`

★ **Merksatz**

> `match-case` prüft feste Fälle und Muster.  
> `if-elif-else` prüft Bedingungen und Bereiche.

### 1 Truthy und Falsy

Python bewertet nicht nur die beiden Werte `True` und `False`.

Manche Werte gelten automatisch als **falsch**. Man nennt sie **falsy**.

Typische falsy-Werte sind:

```python
False
None
0
0.0
""
[]
{}
set()
```

Fast alle anderen Werte gelten als **wahr** beziehungsweise **truthy**.

````python
name = ""

if name:
    print("Name vorhanden")
else:
    print("Name fehlt")
````

## Truthy, Falsy, Bereiche und Sammelprüfungen

```python
if len(name) > 0:
    print("Name vorhanden")
```

★ **Merksatz**

> Leere Werte sind in Python häufig `False`, gefüllte Werte häufig `True`.

### 2 Verkettete Vergleiche

Python kann Zahlenbereiche besonders lesbar prüfen.

Statt:

````python
alter = 16

if alter >= 14 and alter < 18:
    print("Jugendlich")
````

kann man schreiben:

````python
alter = 16

if 14 <= alter < 18:
    print("Jugendlich")
````

Die verkettete Schreibweise liest sich fast wie Mathematik:

> 14 ist kleiner oder gleich `alter` und `alter` ist kleiner als 18.

<a id="any-all"></a>

### 3 `any()` und `all()`

`any()` und `all()` prüfen mehrere Wahrheitswerte in einer Sammlung.

- `any()` ist wahr, wenn mindestens ein Wert wahr ist.
- `all()` ist wahr, wenn alle Werte wahr sind.

````python
rechte = [True, True, False]

print("Mindestens ein Recht vorhanden:", any(rechte))
print("Alle Rechte vorhanden:", all(rechte))
````

Verbindung zu den logischen Operatoren:

- `any()` ähnelt einer langen `or`-Verknüpfung
- `all()` ähnelt einer langen `and`-Verknüpfung

### Abgrenzung: `break` und `try-except`

`break` und `try-except` tauchen häufig zusammen mit Bedingungen auf.  
Fachlich gehören sie aber zu anderen Themenbereichen:

- `break` gehört zu **Schleifen**
- `try-except` gehört zur **Fehlerbehandlung**

## break und try-except

`break` beendet eine laufende `for`- oder `while`-Schleife sofort.

````python
for zahl in range(1, 10):
    if zahl == 5:
        break

    print(zahl)
````

Ausgabe:

```text
1
2
3
4
```

Sobald `zahl == 5` wahr ist, beendet `break` die Schleife.

★ **Merksatz**

> `break` beendet die nächstgelegene Schleife.

### Kein `break` bei Python-`match-case`

Bei Python wird nach einem passenden `case` automatisch das gesamte `match` verlassen.

````python
note = 2

match note:
    case 1:
        print("Sehr gut")
    case 2:
        print("Gut")
    case _:
        print("Andere Note")
````

Anders als beim klassischen Java-`switch` ist in Python kein `break` nötig.

Java benötigt bei der klassischen Schreibweise häufig `break`, damit nicht in den nächsten Fall weitergelaufen wird.

### `try-except` fängt Fehler kontrolliert ab

`try-except` verhindert einen Fehler nicht.

Es sorgt dafür, dass ein auftretender Fehler kontrolliert behandelt wird, statt das Programm sofort zu beenden.

````python
eingabe = "abc"

try:
    zahl = int(eingabe)
    print("Zahl:", zahl)
except ValueError:
    print("Die Eingabe ist keine gültige ganze Zahl.")
````

Ablauf:

1. Python versucht den Code im `try`-Block.
2. `int("abc")` erzeugt einen `ValueError`.
3. Python springt in den passenden `except`-Block.
4. Das Programm kann kontrolliert weiterlaufen.

### Bedingungen und Fehlerbehandlung sind nicht dasselbe

Eine Bedingung prüft einen erwartbaren Zustand:

```python
if alter >= 18:
    print("volljährig")
```

`try-except` behandelt dagegen einen Fehler, der während der Ausführung auftreten kann:

```python
try:
    zahl = int(eingabe)
except ValueError:
    print("Ungültige Eingabe")
```

★ **Abgrenzung**

- `if` prüft Bedingungen
- `break` beendet Schleifen
- `match-case` prüft feste Fälle
- `try-except` fängt Fehler kontrolliert ab

Eine ausführliche Behandlung von Exceptions gehört später in ein eigenes Kapitel zur Fehlerbehandlung.

### Mini-Übung

Was passiert in diesem Beispiel?

````python
werte = ["10", "20", "abc", "30"]

for wert in werte:
    try:
        zahl = int(wert)
        print("Zahl:", zahl)
    except ValueError:
        print("Ungültiger Wert:", wert)
        break
````

#### Erklärung

- `"10"` und `"20"` werden erfolgreich umgewandelt.
- Bei `"abc"` entsteht ein `ValueError`.
- `except` fängt den Fehler ab.
- `break` beendet anschließend die Schleife.
- `"30"` wird deshalb nicht mehr verarbeitet.
