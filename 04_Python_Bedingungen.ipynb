# Python 04: Bedingungen und bedingte Ausdrücke

Bedingungen sorgen dafür, dass ein Programm Entscheidungen treffen kann.

Zum Beispiel:

- Ist jemand volljährig?
- Ist ein Passwort lang genug?
- Ist eine Zahl gerade?
- Darf eine Division durchgeführt werden?

In diesem Notebook schauen wir uns an:

1. Wahrheitswerte
2. Vergleichsoperatoren
3. `if`
4. `if-else`
5. `if-elif-else`
6. `and`, `or` und `not`
7. Einrückungen
8. bedingte Ausdrücke als Kurzform
9. typische Fehler
10. Aufgaben


## 1. Wahrheitswerte

Eine Bedingung liefert immer:

- `True`
- oder `False`


````python
print(5 > 3)
print(5 < 3)
print(5 == 5)
print(5 != 5)
````

## 2. Vergleichsoperatoren

- `==` gleich
- `!=` ungleich
- `<` kleiner
- `>` größer
- `<=` kleiner oder gleich
- `>=` größer oder gleich

Wichtig:

> Ein `=` speichert einen Wert. Zwei `==` vergleichen zwei Werte.


````python
alter = 18

print(alter == 18)
print(alter != 18)
print(alter >= 18)
````

## 3. Die `if`-Bedingung

Mit `if` wird Code nur dann ausgeführt, wenn die Bedingung wahr ist.


````python
alter = 20

if alter >= 18:
    print("Du bist volljährig.")
````

## 4. Einrückung

Alles, was eingerückt unter dem `if` steht, gehört zu dieser Bedingung.

Python nimmt Einrückungen ernst. Das ist nicht nur für die Optik.


````python
alter = 20

if alter >= 18:
    print("Diese Zeile gehört zur Bedingung.")
    print("Diese hier auch.")

print("Diese Zeile kommt danach.")
````

## 5. `if-else`

Mit `else` wird festgelegt, was passieren soll, wenn die Bedingung falsch ist.


````python
alter = 17

if alter >= 18:
    print("Volljährig")
else:
    print("Minderjährig")
````

## 6. `if-elif-else`

Mit `elif` können mehrere Fälle geprüft werden.


````python
alter = 16

if alter < 13:
    status = "Kind"
elif alter < 18:
    status = "Jugendlich"
else:
    status = "Erwachsen"

print(status)
````

Die Bedingungen werden von oben nach unten geprüft.

Sobald eine Bedingung passt, wird der passende Block ausgeführt.


## 7. `and`

Bei `and` müssen beide Bedingungen wahr sein.


````python
alter = 20
hat_ticket = True

if alter >= 18 and hat_ticket:
    print("Einlass erlaubt")
````

## 8. `or`

Bei `or` muss mindestens eine Bedingung wahr sein.


````python
ist_admin = False
ist_moderator = True

if ist_admin or ist_moderator:
    print("Zugriff erlaubt")
````

## 9. `not`

`not` dreht einen Wahrheitswert um.


````python
ist_gesperrt = False

if not ist_gesperrt:
    print("Konto ist aktiv")
````

## 10. Bedingter Ausdruck

Für kurze Entscheidungen gibt es eine kompakte Schreibweise.

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```


````python
alter = 17

status = "Erwachsen" if alter >= 18 else "Minderjährig"

print(status)
````

Gelesen wird das so:

> Nimm den ersten Wert, wenn die Bedingung wahr ist, sonst den zweiten.


### Bedingter Ausdruck in einem f-String


````python
alter = 20

print(f"Status: {'Volljährig' if alter >= 18 else 'Minderjährig'}")
````

## 11. Verschachtelte Kurzformen

Das geht technisch:


````python
alter = 16

status = "Kind" if alter < 13 else "Jugendlich" if alter < 18 else "Erwachsen"

print(status)
````

Ehrlich gesagt wird das schnell unübersichtlich.

Bei mehreren Fällen ist `if-elif-else` meistens die bessere Wahl.


## 12. Typische Fehler

### `=` statt `==`

Falsch:

```python
if alter = 18:
```

Richtig:

```python
if alter == 18:
```

### Doppelpunkt vergessen

Nach `if`, `elif` und `else` gehört ein Doppelpunkt.

### Einrückung vergessen

Der zugehörige Code muss eingerückt sein.


## Wahrheitstabellen für `and` und `or`

### `and`

Nur wenn beide Seiten `True` sind, ist das Ergebnis ebenfalls `True`.


````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

### `or`

Bei `or` reicht ein einziges `True`.


````python
print(True or False)
print(False or True)
print(False or False)
````

## Bereiche prüfen

Python kann Vergleiche miteinander verbinden:


````python
alter = 20

if 18 <= alter <= 67:
    print("Das Alter liegt im gewünschten Bereich.")
````

## Strings vergleichen


````python
antwort = "Ja"

if antwort.lower() == "ja":
    print("Bestätigt")
else:
    print("Nicht bestätigt")
````

## Leere Strings prüfen

Ein leerer String wird als `False` behandelt.


````python
name = ""

if name:
    print("Name vorhanden")
else:
    print("Kein Name eingegeben")
````

## Aufgabe 1: Gerade oder ungerade

Lass eine Zahl eingeben und prüfe, ob sie gerade oder ungerade ist.

### Hinweis

Eine Zahl ist gerade, wenn bei der Division durch 2 kein Rest bleibt.

### Dein Versuch


### Musterlösung


````python
zahl = int(input("Bitte eine ganze Zahl eingeben: "))

if zahl % 2 == 0:
    print("Die Zahl ist gerade.")
else:
    print("Die Zahl ist ungerade.")
````

## Aufgabe 2: Notenbewertung

Gib zu einer Note die passende Bewertung aus:

- 1 → Sehr gut
- 2 → Gut
- 3 → Befriedigend
- 4 → Ausreichend
- alles andere → Ungültig

### Dein Versuch


### Musterlösung


````python
note = int(input("Bitte eine Note eingeben: "))

if note == 1:
    print("Sehr gut")
elif note == 2:
    print("Gut")
elif note == 3:
    print("Befriedigend")
elif note == 4:
    print("Ausreichend")
else:
    print("Ungültige Note")
````

## Aufgabe 5: Rabatt nur ab 100 Euro

Ein Kunde erhält nur dann `10 %` Rabatt, wenn der Warenwert mindestens `100 €` beträgt.

Gib den Rabatt und den Endpreis aus.


### Dein Versuch


### Musterlösung


````python
warenwert = 125.00

if warenwert >= 100:
    rabatt = warenwert * 0.10
else:
    rabatt = 0

endpreis = warenwert - rabatt

print(f"Warenwert: {warenwert:.2f} €")
print(f"Rabatt: {rabatt:.2f} €")
print(f"Endpreis: {endpreis:.2f} €")
````

### Erklärung

Die Berechnung des Rabatts steckt innerhalb der Bedingung. Unter `100 €` bleibt der Rabatt einfach `0`.
