## 1. Was ist ein String?

Ein String ist eine Zeichenkette.

Strings werden in Python mit einfachen oder doppelten Anführungszeichen geschrieben.

Beispiele:

```python
name = "Max"
stadt = 'Würzburg'
```

Ein String kann Buchstaben, Zahlen, Leerzeichen und Sonderzeichen enthalten.


````python
name = "Max"
stadt = 'Würzburg'
nummer_als_text = "123"

print(name, type(name))
print(stadt, type(stadt))
print(nummer_als_text, type(nummer_als_text))
````

## 2. String-Verkettung

Strings können mit `+` verbunden werden.

Das nennt man **Verkettung** oder **Konkatenation**.


````python
mein_text_1 = "Hallo"
mein_text_2 = "Welt"

ergebnis = mein_text_1 + " " + mein_text_2
print(ergebnis)
````

Eine Zahl kann nicht direkt mit einem String verbunden werden.

Das hier wäre falsch:

```python
"Alter: " + 49
```

Die Zahl muss vorher mit `str()` umgewandelt werden.


````python
alter = 49

print("Alter: " + str(alter))
````

## 3. Anführungszeichen und Escaping

Problem:

```text
Hans sagt: "Hallo"
```

Da der String selbst doppelte Anführungszeichen enthält, gibt es mehrere Möglichkeiten:

- außen einfache Anführungszeichen
- innen die Anführungszeichen mit `\` maskieren
- einfache und doppelte Anführungszeichen sinnvoll kombinieren


````python
satz_1 = 'Hans sagt: "Hallo"'
satz_2 = "Hans sagt: \"Hallo\""
satz_3 = "Hans sagt: 'Hallo'"

print(satz_1)
print(satz_2)
print(satz_3)
````

Auch HTML lässt sich leichter schreiben, wenn außen einfache Anführungszeichen verwendet werden.


````python
html = '<a href="http://www.beispiel.de">Link</a>'

print(html)
````

## 4. Steuerzeichen

Wichtige Steuerzeichen:

- `\n` – Zeilenumbruch
- `\r` – Wagenrücklauf
- `\t` – Tabulator
- `\\` – Backslash
- `\"` – doppeltes Anführungszeichen

Unter Windows wird ein Zeilenumbruch häufig als `\r\n` dargestellt.


````python
mehrzeilen = "Hallo\r\nWelt"
tabulator = "Max sagt:\tTest"

print(mehrzeilen)
print(tabulator)
````

## 5. Windows-Pfade und Raw Strings

Bei Windows-Pfaden können Zeichen wie `\t` versehentlich als Steuerzeichen gelesen werden.

Problematisch:

```python
pfad = "c:\windows\tmp"
```

Besser ist ein Raw String:

```python
pfad = r"c:\windows\tmp"
```


````python
pfad_mit_escape = "c:\\windows\\tmp"
pfad_raw = r"c:\windows\tmp"

print(pfad_mit_escape)
print(pfad_raw)
````

Ein Raw String ist besonders für Windows-Pfade und reguläre Ausdrücke praktisch.


## 6. Mehrzeilige Strings

Mit dreifachen Anführungszeichen kann ein String über mehrere Zeilen geschrieben werden.


````python
gedicht = '''Rosen sind rot,
Veilchen sind blau,
Python macht Spaß,
das weißt du genau!'''

print(gedicht)
````

Wichtig:

Ein dreifach geschriebener String ist **kein echter Kommentar**.

Er bleibt ein String. Wird er keiner Variablen zugewiesen, wird er zwar oft wie ein Kommentar verwendet, fachlich ist das aber nicht dasselbe.


## 7. String-Formatierung mit `format()`

Mit `format()` können Werte in eine Textvorlage eingesetzt werden.


````python
vorname = "Max"
stadt = "Würzburg"
alter = 49

satzmaske = "{name} ist {alter} Jahre alt und wohnt in {city}"

ergebnis = satzmaske.format(
    name=vorname,
    alter=alter,
    city=stadt
)

print(ergebnis)
````

Die Zuordnung kann auch über Positionen erfolgen.


````python
satzmaske = "{0} wohnt in {1} und ist {2} Jahre alt"

ergebnis = satzmaske.format(vorname, stadt, alter)

print(ergebnis)
````

## 8. String-Formatierung mit f-Strings

f-Strings sind heute meistens die übersichtlichste Lösung.

Vor den String wird ein `f` geschrieben.

Variablen stehen in geschweiften Klammern.


````python
name = "Max"
city = "Würzburg"
alter = 49

satz = f"{name} wohnt in {city} und ist {alter} Jahre alt"

print(satz)
````

### Rechnen innerhalb eines f-Strings

Auch Berechnungen können direkt in einem f-String stehen.


````python
preis = 100
steuer = 0.19

print(f"Bruttopreis: {preis + preis * steuer}")
````

### Methoden innerhalb eines f-Strings

Auch Methoden können direkt aufgerufen werden.


````python
text = "hello this is my world"

print(f"{text.title()}")
````

### Dictionaries in f-Strings


````python
person = {
    "name": "John",
    "age": 19
}

print(f"{person['name']} ist {person['age']} Jahre alt.")
````

## 9. Zahlen formatieren

Mit Formatangaben können Zahlen passend dargestellt werden.

Beispiele:

- `.2f` – zwei Nachkommastellen
- `>8` – rechtsbündig
- `<8` – linksbündig
- `^8` – zentriert


````python
preis = 9.99875

print(f"{preis:.2f}")
print(f"{preis:>8.2f}")
print(f"{preis:<8.2f}")
print(f"{preis:^8.2f}")
````

Rechtsbündige Darstellung ist besonders bei Tabellen praktisch.


````python
zahlen = [100, 99, 4]

for zahl in zahlen:
    print(f"{zahl:>8.3f}")
````

Dasselbe funktioniert auch mit `format()`.


````python
zahlen = [100, 99, 4]

for zahl in zahlen:
    print("{zahl:>8.3f}".format(zahl=zahl))
````

### Wichtig: Formatierte Zahlen sind Strings

Nach der Formatierung ist das Ergebnis kein Zahlenwert mehr, sondern ein String.

Damit sollte nicht direkt weitergerechnet werden.


````python
zahl = 100
formatierte_zahl = f"{zahl:>8.2f}"

print(zahl, type(zahl))
print(formatierte_zahl, type(formatierte_zahl))
````

## 10. Bedingter Ausdruck im f-String

Kurze Entscheidungen können direkt in einem f-String stehen.


````python
alter = 20

status = f"Du bist: {'Erwachsen' if alter >= 18 else 'Minderjährig'}"

print(status)
````

## 11. String-Methoden

Strings besitzen viele eingebaute Methoden.

Wichtige Methoden:

- `upper()` – alles groß
- `lower()` – alles klein
- `capitalize()` – erstes Zeichen groß
- `title()` – jedes Wort beginnt groß
- `strip()` – Zeichen an beiden Enden entfernen
- `lstrip()` – links entfernen
- `rstrip()` – rechts entfernen
- `replace()` – ersetzen
- `center()` – zentrieren
- `split()` – zerlegen
- `join()` – zusammensetzen
- `find()` – Position suchen


````python
mein_string = "    hello mein, World!-------"

print("Länge:", len(mein_string))
print("Groß:", mein_string.upper())
print("Klein:", mein_string.lower())
print("Erster Buchstabe groß:", mein_string.capitalize())
print("Jedes Wort groß:", mein_string.title())

print("Leerzeichen entfernen:", mein_string.strip())
print("Bindestriche entfernen:", mein_string.strip("-"))
print("Beides entfernen:", mein_string.strip().strip("-"))

print("Ersetzen:", mein_string.replace("World", "Mars"))
print("Zentrieren:", mein_string.center(60, "#"))
````

### String zerlegen mit `split()`


````python
zeichenkette = "Apfel,Birne,Banane"

obst = zeichenkette.split(",")

print(obst)
print(type(obst))
````

### Mit `join()` wieder zusammensetzen


````python
obst = ["Apfel", "Birne", "Banane"]

ergebnis = " | ".join(obst)

print(ergebnis)
````

## 12. Prüffunktionen für Strings

Wichtige Prüffunktionen:

- `isspace()`
- `startswith()`
- `endswith()`
- `isalnum()`
- `isdecimal()`
- `isdigit()`
- `isnumeric()`
- `isidentifier()`
- `isalpha()`

Die Rückgabe ist immer `True` oder `False`.


````python
leerzeichen = " "
datei = "bild.jpg"
test = "id_123"

print("isspace:", leerzeichen.isspace())
print("startswith:", datei.startswith("bild"))
print("endswith:", datei.endswith(".jpg"))

print("isalnum:", test.isalnum())
print("isdecimal:", test.isdecimal())
print("isdigit:", test.isdigit())
print("isnumeric:", test.isnumeric())
print("isidentifier:", test.isidentifier())
print("isalpha:", test.isalpha())
````

## 13. `find()`

`find()` sucht nach einem Zeichen oder Text.

Gefunden wird die Position des ersten Vorkommens.

Wird nichts gefunden, liefert `find()` den Wert `-1`.


````python
text = "Hallo Welt"

print(text.find("l"))
print(text.find("Welt"))
print(text.find("Python"))
````

## 14. String-Slicing

Strings können über ihren Index angesprochen werden.

Der erste Index ist `0`.

Mit negativen Indizes wird von hinten gezählt.


````python
string = "Hallo, Welt!"

print(string[0])
print(string[7])
print(string[-2])
````

### Teilbereiche auswählen


````python
string = "Hallo, Welt!"

print(string[0:6])
print(string[:6])
print(string[5:])
print(string[-5:])
print(string[-5:-2])
````

### Schrittweite

Der dritte Wert beim Slicing ist die Schrittweite.

Aufbau:

```python
string[start:ende:schrittweite]
```


````python
string = "Hallo, Welt!"

print(string[::3])
````

### String umdrehen

Mit `[::-1]` wird die Reihenfolge umgedreht.


````python
bitfolge = "11110000"

print(bitfolge[::-1])
````

### Negative Schrittweite


````python
string = "ABCDEFGHIJ"

print(string[8:1:-2])
````

## 15. HTML-Entitäten

`&amp;` ist die HTML-Schreibweise für `&`.

Python wandelt diese Schreibweise nicht automatisch um.

## 16. Einführung in reguläre Ausdrücke

Reguläre Ausdrücke werden mit dem Modul `re` verwendet.

Wichtige Funktionen:

- `re.findall()` – alle Treffer als Liste
- `re.search()` – einen Treffer suchen
- `re.fullmatch()` – ganze Zeichenkette prüfen
- `re.sub()` – Treffer ersetzen


````python
import re

text = "Artikel A123 kostet 49 Euro und Artikel B456 kostet 99 €."
````

### Alle Zahlen finden


````python
print(re.findall(r"[0-9]+", text))
````

### Artikelnummern finden

Das Muster bedeutet:

- ein Großbuchstabe von A bis Z
- danach mindestens eine Ziffer


````python
print(re.findall(r"[A-Z][0-9]+", text))
````

### Preise finden

Das Muster sucht:

- eine oder mehrere Ziffern
- danach Leerraum
- danach eine erlaubte Währungsangabe


````python
treffer = re.findall(r"(\d+)\s+(Euro|euro|€|EUR)", text)

preise = [eintrag[0] for eintrag in treffer]

print(preise)
````

## Aufgabe 1: Passwort ohne Regex analysieren

Prüfe, ob ein Passwort:

- mindestens 8 Zeichen lang ist
- einen Großbuchstaben enthält
- einen Kleinbuchstaben enthält
- eine Zahl enthält

Diese Aufgabe passt zu den String-Prüfmethoden.


````python
passwort = "MeinGeheimes&§123"

mindestens_acht_zeichen = len(passwort) >= 8
enthaelt_grossbuchstaben = any(zeichen.isupper() for zeichen in passwort)
enthaelt_kleinbuchstaben = any(zeichen.islower() for zeichen in passwort)
enthaelt_zahl = any(zeichen.isdigit() for zeichen in passwort)

print("Mindestens 8 Zeichen:", mindestens_acht_zeichen)
print("Großbuchstabe:", enthaelt_grossbuchstaben)
print("Kleinbuchstabe:", enthaelt_kleinbuchstaben)
print("Zahl:", enthaelt_zahl)

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
):
    print("Das Passwort erfüllt alle Bedingungen.")
else:
    print("Das Passwort erfüllt nicht alle Bedingungen.")
````

## Aufgabe 2: Wörter eines Satzes umkehren

Die Reihenfolge der Wörter wird umgedreht.

Vorgehen:

1. mit `split()` zerlegen
2. mit `[::-1]` umdrehen
3. mit `join()` wieder verbinden


````python
text = "Hallo du schöne Sonne"

woerter = text.split()
umgekehrte_woerter = woerter[::-1]
ergebnis = " ".join(umgekehrte_woerter)

print(ergebnis)
````
