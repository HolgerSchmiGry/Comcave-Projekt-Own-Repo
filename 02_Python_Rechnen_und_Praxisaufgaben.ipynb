## Aufgabe 1: Sozialversicherung berechnen

Gegeben sind:

- Bruttolohn
- Arbeitslosenversicherung in Prozent
- Rentenversicherung in Prozent
- Krankenversicherung in Prozent
- Pflegeversicherung in Prozent

Ausgegeben werden sollen:

- Bruttolohn
- jeder einzelne Versicherungsbeitrag
- Summe der Beiträge
- verbleibender Betrag nach diesen Abzügen

### Vereinfachung

Beitragsbemessungsgrenzen, Steuern, Kirchensteuer, Kinderlosenzuschlag und weitere echte Lohnabrechnungsregeln werden hier noch nicht berücksichtigt.

Der Dozent nennt als Ergebnis das „zu versteuernde Einkommen“. Fachlich genauer ist es in dieser vereinfachten Aufgabe aber nur der **verbleibende Betrag nach den Sozialversicherungsbeiträgen**.


### Planung

1. Bruttolohn und Beitragssätze speichern.
2. Prozentsätze durch `100` teilen.
3. Jeden Beitrag berechnen.
4. Beiträge addieren.
5. Gesamtabzug vom Bruttolohn abziehen.


### Dein Versuch


### Musterlösung


````python
bruttolohn = 3500.00

arbeitslosenversicherung_prozent = 1.3
rentenversicherung_prozent = 9.3
krankenversicherung_prozent = 7.3
pflegeversicherung_prozent = 1.7

arbeitslosenversicherung = (
    bruttolohn * arbeitslosenversicherung_prozent / 100
)
rentenversicherung = bruttolohn * rentenversicherung_prozent / 100
krankenversicherung = bruttolohn * krankenversicherung_prozent / 100
pflegeversicherung = bruttolohn * pflegeversicherung_prozent / 100

sozialversicherungsbeitraege = (
    arbeitslosenversicherung
    + rentenversicherung
    + krankenversicherung
    + pflegeversicherung
)

verbleibender_betrag = bruttolohn - sozialversicherungsbeitraege

print(f"Bruttolohn: {bruttolohn:.2f} €")
print(f"Arbeitslosenversicherung: {arbeitslosenversicherung:.2f} €")
print(f"Rentenversicherung: {rentenversicherung:.2f} €")
print(f"Krankenversicherung: {krankenversicherung:.2f} €")
print(f"Pflegeversicherung: {pflegeversicherung:.2f} €")
print(f"Beiträge gesamt: {sozialversicherungsbeitraege:.2f} €")
print(f"Verbleibender Betrag: {verbleibender_betrag:.2f} €")
````

### Erklärung

Bei jedem Beitrag wird der Bruttolohn mit dem Prozentsatz multipliziert und anschließend durch `100` geteilt. Erst danach werden alle Beiträge zusammengerechnet.


## Aufgabe 2: Warenwert mit Rabatten und Zusatzkosten

Gegeben ist ein Warenwert.

Berechnet werden:

- Mengenrabatt: `10 %`
- Sonderrabatt: `5 %`
- Versandkosten: `4,90 €`
- Verpackungskosten: `2 %`
- Mehrwertsteuer: `19 %`

### Rechenreihenfolge

Für diese Aufgabe nehmen wir an:

1. Beide Rabatte werden vom ursprünglichen Warenwert berechnet und abgezogen.
2. Die Verpackungskosten werden ebenfalls vom ursprünglichen Warenwert berechnet.
3. Versand und Verpackung werden anschließend hinzugerechnet.
4. Auf diesen Nettobetrag kommt zum Schluss die Mehrwertsteuer.

Diese Annahme ist wichtig, weil die Aufgabenstellung nicht genauer sagt, von welchem Zwischenbetrag die `2 %` Verpackungskosten berechnet werden.


### Dein Versuch


### Musterlösung


````python
warenwert = 200.00

mengenrabatt_prozent = 10
sonderrabatt_prozent = 5
versandkosten = 4.90
verpackungskosten_prozent = 2
mehrwertsteuer_prozent = 19

mengenrabatt = warenwert * mengenrabatt_prozent / 100
sonderrabatt = warenwert * sonderrabatt_prozent / 100
verpackungskosten = warenwert * verpackungskosten_prozent / 100

nettobetrag = (
    warenwert
    - mengenrabatt
    - sonderrabatt
    + versandkosten
    + verpackungskosten
)

mehrwertsteuer = nettobetrag * mehrwertsteuer_prozent / 100
endpreis = nettobetrag + mehrwertsteuer

print(f"Warenwert: {warenwert:.2f} €")
print(f"Mengenrabatt: -{mengenrabatt:.2f} €")
print(f"Sonderrabatt: -{sonderrabatt:.2f} €")
print(f"Versandkosten: +{versandkosten:.2f} €")
print(f"Verpackungskosten: +{verpackungskosten:.2f} €")
print(f"Nettobetrag: {nettobetrag:.2f} €")
print(f"Mehrwertsteuer: +{mehrwertsteuer:.2f} €")
print(f"Endpreis: {endpreis:.2f} €")
````

### Erklärung

Rabatte werden abgezogen, Zuschläge werden addiert. Die Mehrwertsteuer wird erst auf den neuen Nettobetrag berechnet.


## Aufgabe 3: Zinseszins berechnen

Speichere:

- Startkapital
- Zinssatz
- Anzahl der Jahre

Berechne anschließend das Endkapital mit Zinseszins.

Formel:

```text
Endkapital = Startkapital × (1 + Zinssatz) ** Jahre
```


### Dein Versuch


### Musterlösung


````python
startkapital = 1000.00
zinssatz_prozent = 3
jahre = 5

zinssatz = zinssatz_prozent / 100
endkapital = startkapital * (1 + zinssatz) ** jahre
gewinn = endkapital - startkapital

print(f"Startkapital: {startkapital:.2f} €")
print(f"Zinssatz: {zinssatz_prozent:.2f} %")
print(f"Laufzeit: {jahre} Jahre")
print(f"Endkapital: {endkapital:.2f} €")
print(f"Gewinn: {gewinn:.2f} €")
````

### Erklärung

Der Ausdruck `(1 + zinssatz) ** jahre` sorgt dafür, dass die Zinsen jedes Jahr wieder mitverzinst werden.
