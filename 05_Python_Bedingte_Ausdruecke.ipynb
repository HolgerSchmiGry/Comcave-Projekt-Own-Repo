## 1. Normale `if-else`-Bedingung

Mit `if` und `else` wird eine Entscheidung getroffen.

Beispiel:

- Ist eine Person mindestens 18 Jahre alt, lautet der Status `Erwachsen`.
- Sonst lautet der Status `Minderjährig`.


````python
alter = 17

if alter >= 18:
    status = "Erwachsen"
else:
    status = "Minderjährig"

print(status)
````

## 2. Bedingter Ausdruck

Für kurze Entscheidungen kann Python eine kompakte Schreibweise verwenden.

Aufbau:

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```

Beispiel:

```python
status = "Erwachsen" if alter >= 18 else "Minderjährig"
```


````python
alter = 17

status = "Erwachsen" if alter >= 18 else "Minderjährig"

print(status)
````

## 3. Leserichtung

Ein bedingter Ausdruck wird so gelesen:

```text
Nimm den ersten Wert,
wenn die Bedingung wahr ist,
sonst nimm den zweiten Wert.
```

Beispiel:

```python
"Erwachsen" if alter >= 18 else "Minderjährig"
```

Bedeutung:

- Wenn `alter >= 18` wahr ist → `"Erwachsen"`
- sonst → `"Minderjährig"`


## 4. Verwendung mit Zahlen

Ein bedingter Ausdruck kann nicht nur Texte, sondern auch Zahlen zurückgeben.


````python
zahl = 10

ergebnis = 1 if zahl > 0 else 0

print(ergebnis)
````

## 5. Verwendung in f-Strings

Ein bedingter Ausdruck kann direkt in einem f-String verwendet werden.


````python
alter = 20

print(f"Status: {'Volljährig' if alter >= 18 else 'Minderjährig'}")
````

## 6. Verschachtelte bedingte Ausdrücke

Mehrere Bedingungen können theoretisch miteinander verschachtelt werden.

Beispiel:

```python
status = "Kind" if alter < 13 else "Jugendlich" if alter < 18 else "Erwachsen"
```

Das funktioniert, wird aber schnell unübersichtlich.


````python
alter = 16

status = "Kind" if alter < 13 else "Jugendlich" if alter < 18 else "Erwachsen"

print(status)
````

Für längere oder kompliziertere Entscheidungen ist ein normaler `if-elif-else`-Block besser lesbar.


````python
alter = 16

if alter < 13:
    status = "Kind"
elif alter < 18:
    status = "Jugendlich"
else:
    status = "Erwachsen"

print(status)
````

## 7. Unterschied zu anderen Programmiersprachen

In manchen Sprachen gibt es den sogenannten ternären Operator:

```text
bedingung ? wert_wenn_wahr : wert_wenn_falsch
```

Python verwendet diese Schreibweise **nicht**.

In Python lautet die Reihenfolge:

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```


## 8. Typische Fehler

### Fehler 1: Reihenfolge vertauschen

Falsch:

```python
if alter >= 18 "Erwachsen" else "Minderjährig"
```

Richtig:

```python
"Erwachsen" if alter >= 18 else "Minderjährig"
```

### Fehler 2: Zu viele Bedingungen in eine Zeile packen

Technisch möglich heißt nicht automatisch gut lesbar.

Bei komplizierten Entscheidungen besser `if-elif-else` verwenden.


## Aufgabe 1: Gerade oder ungerade

Prüfe, ob eine Zahl gerade oder ungerade ist.

Hinweis:

Eine Zahl ist gerade, wenn bei der Division durch 2 kein Rest bleibt.


````python
zahl = 7

ergebnis = "Gerade" if zahl % 2 == 0 else "Ungerade"

print(ergebnis)
````

## Aufgabe 2: Positiv oder nicht positiv

Prüfe, ob eine Zahl größer als 0 ist.


````python
zahl = -5

status = "Positiv" if zahl > 0 else "Nicht positiv"

print(status)
````

## Aufgabe 3: Volljährigkeit

Speichere ein Alter und gib mit einem bedingten Ausdruck aus, ob die Person volljährig oder minderjährig ist.


````python
alter = 18

status = "Volljährig" if alter >= 18 else "Minderjährig"

print(status)
````
