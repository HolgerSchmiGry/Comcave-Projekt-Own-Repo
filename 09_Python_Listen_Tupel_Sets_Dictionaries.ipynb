# 1. Überblick über die Datenstrukturen

| Datenstruktur | Schreibweise | Veränderbar? | Duplikate? | Typischer Einsatz |
|---|---|---:|---:|---|
| Liste | `[]` | Ja | Ja | veränderbare Reihenfolge |
| Tupel | `()` | Nein | Ja | feste Werte oder Koordinaten |
| Set | `{wert1, wert2}` | Ja | Nein | eindeutige Werte und Vergleiche |
| Dictionary | `{"schlüssel": wert}` | Ja | Schlüssel eindeutig | Eigenschaften eines Objekts |

★ **Merksatz**

> Listen speichern Reihenfolgen, Tupel schützen feste Werte, Sets vergleichen eindeutige Werte und Dictionaries beschreiben Objekte.

# 2. Listen

Eine Liste speichert mehrere Werte in einer festen Reihenfolge.

Eigenschaften:

- Zugriff über einen Index
- veränderbar
- Duplikate sind erlaubt
- verschiedene Datentypen sind möglich

````python
obstkorb = ["Apfel", "Banane", "Zitrone", "Birne", "Kiwi", "Banane"]

print(obstkorb)
print(type(obstkorb))
print("Anzahl:", len(obstkorb))
````

## Der Index beginnt bei `0`

| Index | Wert |
|---:|---|
| `0` | Apfel |
| `1` | Banane |
| `2` | Zitrone |

Der erste Eintrag besitzt also nicht den Index `1`, sondern den Index `0`.

````python
print("Erstes Element:", obstkorb[0])
print("Zweites Element:", obstkorb[1])
print("Vorletztes Element:", obstkorb[-2])
````

# 3. Zugriff und Slicing

Mit **Slicing** wird ein Teilbereich aus einer Liste geholt.

Grundform:

```python
liste[start:ende:schrittweite]
```

Der Wert bei `ende` ist nicht mehr enthalten.

````python
obstkorb = ["Apfel", "Banane", "Zitrone", "Birne", "Kiwi", "Banane"]

print("Ab Index 2:", obstkorb[2:])
print("Bis Index 5:", obstkorb[:5])
print("Index 3 bis 4:", obstkorb[3:5])
print("Letzte drei:", obstkorb[-3:])
print("Mit Schrittweite 2:", obstkorb[0:5:2])
print("Rückwärts:", obstkorb[::-1])
````

## Negative Indizes

Negative Indizes zählen von hinten:

- `-1` → letztes Element
- `-2` → vorletztes Element
- `-3` → drittletztes Element

# 4. Listen verändern

Listen können nachträglich verändert werden.

## Wert über den Index ersetzen

````python
planeten = ["Merkur", "Venus", "Errdee", "Mars", "Jupiter", "Saturn", "Uranus", "Neptun"]

planeten[2] = "Erde"

print(planeten)
````

## `append()` – am Ende ergänzen

````python
obstkorb = ["Apfel", "Banane"]
obstkorb.append("Melone")

print(obstkorb)
````

## `insert()` – an einer bestimmten Position einfügen

````python
obstkorb.insert(1, "Kirsche")

print(obstkorb)
````

## `remove()` – ersten passenden Wert entfernen

`remove()` sucht nach einem Wert. Ist der Wert nicht vorhanden, entsteht ein Fehler.

````python
obstkorb.remove("Banane")

print(obstkorb)
````

## `pop()` – über den Index entfernen

Ohne Index entfernt `pop()` das letzte Element. Der entfernte Wert wird zurückgegeben.

````python
entfernt = obstkorb.pop()

print("Entfernt:", entfernt)
print("Übrig:", obstkorb)
````

## Listen richtig kopieren

Diese Zuweisung erstellt **keine echte Kopie**:

```python
neue_liste = alte_liste
```

Beide Variablen zeigen dann auf dieselbe Liste.

Eine echte, unabhängige Kopie entsteht mit:

```python
neue_liste = alte_liste.copy()
```

````python
original = ["Erde", "Mars", "Jupiter"]

nur_zweiter_name = original
echte_kopie = original.copy()

nur_zweiter_name.append("Saturn")

print("Original:", original)
print("Zweiter Name:", nur_zweiter_name)
print("Echte Kopie:", echte_kopie)
````

# 5. Listen verbinden und sortieren

## `extend()` – bestehende Liste erweitern

`extend()` verändert die erste Liste.

````python
autos = ["VW", "Audi", "BMW"]
obst = ["Apfel", "Banane"]

autos.extend(obst)

print(autos)
````

## Listen mit `+` verbinden

Dabei entsteht eine neue Liste.

````python
autos = ["VW", "Audi", "BMW"]
obst = ["Apfel", "Banane"]

verbunden = autos + obst

print(verbunden)
````

## Listen mit dem Unpacking-Operator `*` verbinden

````python
verbunden = [*autos, *obst]

print(verbunden)
````

## Alphabetisch sortieren

`sort()` verändert die vorhandene Liste.

Groß- und Kleinschreibung können mit `key=str.lower` ignoriert werden.

````python
obstkorb = ["Banane", "Zitrone", "Birne", "Kiwi", "Apfel", "zwetschge", "affe"]

obstkorb.sort(key=str.lower)

print(obstkorb)
````

## Nach der Länge sortieren

````python
obstkorb.sort(key=len)

print(obstkorb)
````

## Eigene Sortierfunktion als `key`

````python
def laenge_des_wortes(wort):
    return len(wort)

obstkorb.sort(key=laenge_des_wortes)

print(obstkorb)
````

Das Gleiche wäre später auch mit einer Lambda-Funktion möglich:

```python
obstkorb.sort(key=lambda wort: len(wort))
```

# 7. Verschachtelte Listen

Eine verschachtelte Liste enthält weitere Listen.

````python
tabelle = [
    ["Susi", "Sauer", "Susistraße 1"],
    ["Max", "Mustermann", "Musterstraße 1"],
    [
        ["Hallo"]
    ]
]

print(tabelle[1][2])
print(tabelle[1][1][3])
print(tabelle[2][0][0])
````

Der Zugriff wird von links nach rechts gelesen:

```python
tabelle[1][1][3]
```

1. `tabelle[1]` → zweite Zeile  
2. `[1]` → zweiter Wert dieser Zeile: `"Mustermann"`  
3. `[3]` → viertes Zeichen: `"t"`

# 8. Tupel

Ein Tupel ist eine geordnete, aber unveränderbare Sammlung.

Tupel eignen sich für Werte, die nach der Erstellung nicht direkt verändert werden sollen.

````python
obst_tupel = ("Apfel", "Banane", "Zitrone", "Birne", "Kiwi")

print(obst_tupel)
print(type(obst_tupel))
print("Anzahl:", len(obst_tupel))
````

## Tupel mit genau einem Element

Das Komma ist entscheidend:

````python
ein_wert = ("Apfel",)

print(ein_wert)
print(type(ein_wert))
````

Ohne Komma wäre es nur ein String in Klammern:

```python
kein_tupel = ("Apfel")
```

## Zugriff und Slicing funktionieren wie bei Listen

````python
obst_tupel = ("Apfel", "Banane", "Zitrone", "Birne", "Kiwi")

print(obst_tupel[0])
print(obst_tupel[-2])
print(obst_tupel[1:4])
````

## Ein Tupel indirekt erweitern

Ein Tupel kann nicht direkt verändert werden. Es kann aber vorübergehend in eine Liste umgewandelt werden.

````python
obst_tupel = ("Apfel", "Banane")

temporaere_liste = list(obst_tupel)
temporaere_liste.append("Melone")
obst_tupel = tuple(temporaere_liste)

print(obst_tupel)
````

Eine weitere Möglichkeit ist, ein neues Tupel zu erzeugen:

````python
obst_tupel = obst_tupel + ("Aprikose",)

print(obst_tupel)
````

## Unpacking

Beim Unpacking werden die Werte eines Tupels direkt auf Variablen verteilt.

````python
position = (120, 350)

x, y = position

print("X:", x)
print("Y:", y)
````

## Tupel als Rückgabewert einer Funktion

````python
def erstelle_abmessung():
    breite = 800
    hoehe = 600

    return breite, hoehe


breite, hoehe = erstelle_abmessung()

print("Breite:", breite)
print("Höhe:", hoehe)
````

# 9. Sets

Ein Set speichert eindeutige Werte.

Eigenschaften:

- keine Duplikate
- kein Zugriff über einen Index
- Reihenfolge nicht als fachliche Sortierung verwenden
- Werte können hinzugefügt und entfernt werden
- sehr gut für Mengenvergleiche geeignet

> Ein Set ist veränderbar. Die einzelnen enthaltenen Werte müssen jedoch hashbar sein. Listen können deshalb beispielsweise nicht direkt in einem Set liegen.

````python
mein_set = {"Apfel", "Zitrone", "Banane"}

print(mein_set)
print(type(mein_set))
````

## Duplikate verschwinden

Achtung: In Python gelten `True` und `1` bei Sets als gleichwertig.

````python
werte = {"Apfel", "Zitrone", "Zitrone", True, 1, 2}

print(werte)
````

## `add()` – einen Wert hinzufügen

````python
mein_set.add("Melone")

print(mein_set)
````

## `update()` – mehrere Werte ergänzen

````python
mein_set.update({"Ananas", "Kiwi"})

print(mein_set)
````

## `remove()` und `discard()`

- `remove()` erzeugt einen Fehler, wenn der Wert fehlt.
- `discard()` entfernt nur dann, wenn der Wert vorhanden ist.

````python
mein_set.discard("Pfirsich")
print(mein_set)
````

Ein bewusstes Fehlerbeispiel sollte nicht automatisch ausgeführt werden:

```python
mein_set.remove("Pfirsich")
```

Ist `"Pfirsich"` nicht vorhanden, entsteht ein `KeyError`.

## `union()` – Vereinigungsmenge

````python
menge_a = {"Apfel", "Zitrone", "Banane", "Kiwi"}
menge_b = {"Ananas", "Kiwi"}

print(menge_a.union(menge_b))
print(menge_a | menge_b)
````

## `intersection()` – Schnittmenge

````python
print(menge_a.intersection(menge_b))
print(menge_a & menge_b)
````

## `difference()` – nur in der ersten Menge

````python
print(menge_a.difference(menge_b))
print(menge_a - menge_b)
````

## `symmetric_difference()` – Unterschiede auf beiden Seiten

````python
print(menge_a.symmetric_difference(menge_b))
print(menge_a ^ menge_b)
````

## Praxisbeispiel: Alte und neue Herstellerliste

Aus den Mengen lassen sich drei Gruppen bilden:

- löschen oder deaktivieren
- aktualisieren
- neu einfügen

````python
original = {"PE-123456", "PE-123457", "PE-123458", "PE-123459"}
neu = {"PE-123456", "PE-123457", "PE-123460"}

produkte_deaktivieren = original - neu
produkte_aktualisieren = original & neu
produkte_einfuegen = neu - original

print("Deaktivieren:", produkte_deaktivieren)
print("Aktualisieren:", produkte_aktualisieren)
print("Einfügen:", produkte_einfuegen)
````

> In einer echten Shop-Datenbank sollten alte Produkte meist nicht sofort gelöscht werden. Häufig werden sie auf `lieferbar = False` oder `aktiv = False` gesetzt, damit alte Bestellungen und Rechnungen erhalten bleiben.

## Werte finden, die in beiden Sets fehlen

Dafür wird eine Grundmenge benötigt.

````python
menge_a = {1, 2, 4, 6, 7, 8, 9}
menge_b = {1, 2, 3, 6, 7, 8, 9}
grundmenge = set(range(1, 10))

fehlend = grundmenge - (menge_a | menge_b)

print(fehlend)
````

Für die Zahlen von `1` bis `1000`:

```python
grundmenge = set(range(1, 1001))
```

`1001` wird angegeben, weil der Endwert von `range()` nicht enthalten ist.

## Leeres Set

```python
leeres_set = set()
```

`{}` wäre kein leeres Set, sondern ein leeres Dictionary.

# 10. Dictionaries

Ein Dictionary speichert Daten als Schlüssel-Wert-Paare.

Beispiel:

```python
"firstName": "Susi"
```

- `"firstName"` ist der Schlüssel.
- `"Susi"` ist der zugehörige Wert.

````python
person = {
    "firstName": "Susi",
    "lastName": "Sauer",
    "gender": "w",
    "height": 163,
    "weight": 60.6,
    "hobbies": ["singen", "tanzen"],
    "hair": "blond"
}

print(person)
````

## Werte auslesen

````python
print(person["firstName"])
print(person["weight"])
print(person["hobbies"][1])
print(person["firstName"] + " " + person["lastName"])
````

## Wert ändern

````python
person["lastName"] = "Müller"

print(person)
````

## Schlüssel und Wert entfernen

````python
person.pop("hair")

print(person)
````

## Zwei Dictionaries zusammenführen

Moderne Schreibweise:

```python
gesamt = person | adresse
```

Dabei entsteht ein neues Dictionary.

````python
adresse = {
    "street": "Musterstraße 1",
    "zip": "12345",
    "city": "Entenhausen"
}

gesamt = person | adresse

print(gesamt)
````

Die im Dozentenmaterial verwendete Schreibweise mit Dictionary-Unpacking ist ebenfalls korrekt:

````python
gesamt_mit_unpacking = {**person, **adresse}

print(gesamt_mit_unpacking)
````

## Gleicher Schlüssel in beiden Dictionaries

Der Wert aus dem rechten Dictionary gewinnt.

````python
alte_daten = {"name": "Susi", "alter": 30}
neue_daten = {"alter": 31, "stadt": "Köln"}

zusammengefuehrt = alte_daten | neue_daten

print(zusammengefuehrt)
````

## `update()` verändert das vorhandene Dictionary

````python
alte_daten.update(neue_daten)

print(alte_daten)
````

# 11. Verschachtelte Dictionaries

Dictionaries können weitere Dictionaries und Listen enthalten. Solche Strukturen begegnen dir später häufig bei JSON, APIs und Shop-Systemen.

````python
product = {
    "product": {
        "id": "e9d10e5c6a1e4b3e944fb8fcde5bb883",
        "parentId": None,
        "name": "Premium Leather Jacket",
        "description": "A classic, genuine leather jacket with premium stitching.",
        "productNumber": "SW10001",
        "stock": 45,
        "availableStock": 45,
        "active": True,
        "shippingFree": False,
        "calculatedPrice": {
            "unitPrice": 199.99,
            "quantity": 1,
            "totalPrice": 199.99,
            "calculatedTaxes": [
                {
                    "tax": 31.94,
                    "taxRate": 19.0,
                    "price": 199.99
                }
            ],
            "regulationPrice": None
        },
        "cover": {
            "id": "7a8b9c0d1e2f3g4h5i6j7k8l9m0n1o2p",
            "media": {
                "id": "media_id_12345",
                "url": "https://yourshopdomain.com",
                "alt": "Premium Leather Jacket front view",
                "width": 1200,
                "height": 1200
            }
        },
        "properties": [
            {
                "id": "prop_color_id",
                "name": "Color",
                "group": {
                    "id": "group_color_id",
                    "name": "Colors"
                },
                "translated": {
                    "name": "Black"
                }
            },
            {
                "id": "prop_size_id",
                "name": "Size",
                "group": {
                    "id": "group_size_id",
                    "name": "Sizes"
                },
                "translated": {
                    "name": "L"
                }
            }
        ],
        "manufacturer": {
            "id": "manufacturer_id_100",
            "name": "FashionBrand",
            "link": "https://yourshopdomain.com"
        }
    }
}
````

## Werte schrittweise auslesen

Es ist lesbarer, zuerst die innere Produktebene in einer Variablen zu speichern.

````python
produktdaten = product["product"]

produktname = produktdaten["name"]
produkt_id = produktdaten["id"]
preis = produktdaten["calculatedPrice"]["calculatedTaxes"][0]["price"]
steuersatz = produktdaten["calculatedPrice"]["calculatedTaxes"][0]["taxRate"]
hersteller = produktdaten["manufacturer"]["name"]

print("Produktname:", produktname)
print("Produkt-ID:", produkt_id)
print("Preis:", preis, "Euro")
print("Steuersatz:", steuersatz, "%")
print("Hersteller:", hersteller)
````

## Eigenschaften mit einer Schleife ausgeben

````python
for eigenschaft in produktdaten["properties"]:
    name = eigenschaft["name"]
    wert = eigenschaft["translated"]["name"]

    print(f"{name}: {wert}")
````
