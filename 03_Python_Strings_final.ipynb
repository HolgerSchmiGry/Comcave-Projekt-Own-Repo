## 1. Was ist ein String?

Ein String ist einfach Text.

Strings stehen in einfachen oder doppelten Anführungszeichen:


````python
name = "Max"
stadt = 'Würzburg'
nummer_als_text = "123"

print(name, type(name))
print(stadt, type(stadt))
print(nummer_als_text, type(nummer_als_text))
````

`"123"` sieht zwar wie eine Zahl aus, ist durch die Anführungszeichen aber Text.


## 2. Strings verbinden

Strings lassen sich mit `+` verbinden.

Das nennt man **Verkettung** oder **Konkatenation**.


````python
mein_text_1 = "Hallo"
mein_text_2 = "Welt"

print(mein_text_1 + " " + mein_text_2)
````

Eine Zahl kann nicht direkt an einen String gehängt werden.

Falsch:

```python
"Alter: " + 49
```

Richtig:


````python
alter = 49

print("Alter: " + str(alter))
````

## 3. Anführungszeichen und Escaping

Sobald im Text selbst Anführungszeichen vorkommen, muss man ein bisschen aufpassen.


````python
satz_1 = 'Hans sagt: "Hallo"'
satz_2 = "Hans sagt: \"Hallo\""
satz_3 = "Hans sagt: 'Hallo'"

print(satz_1)
print(satz_2)
print(satz_3)
````

Der Backslash `\` maskiert das nächste Zeichen.

Das bedeutet: Python soll das Zeichen als normalen Teil des Textes behandeln.


### HTML als String

Bei HTML ist es oft praktisch, außen einfache Anführungszeichen zu verwenden:


````python
html = '<a href="http://www.beispiel.de">Link</a>'

print(html)
````

## 4. Steuerzeichen

Wichtige Steuerzeichen:

- `\n` – neue Zeile
- `\r` – Wagenrücklauf
- `\t` – Tabulator
- `\\` – Backslash


````python
print("Hallo\nWelt")
print("Max sagt:\tTest")
````

## 5. Windows-Pfade und Raw Strings

Bei Windows-Pfaden kann `\t` versehentlich als Tabulator gelesen werden.

Deshalb ist ein Raw String oft die angenehmere Lösung:


````python
pfad = r"c:\windows\tmp"

print(pfad)
````

Das `r` vor dem String bedeutet sinngemäß:

> Behandle die Backslashes möglichst als normale Zeichen.


## 6. Mehrzeilige Strings

Mit dreifachen Anführungszeichen kann ein String über mehrere Zeilen gehen.


````python
gedicht = '''Rosen sind rot,
Veilchen sind blau,
Python macht Spaß,
das weißt du genau!'''

print(gedicht)
````

Wichtig:

Ein solcher Block ist **kein echter Kommentar**, sondern weiterhin ein String.


## 7. String-Formatierung mit `format()`

Mit `format()` lassen sich Werte in Platzhalter einsetzen.


````python
vorname = "Max"
stadt = "Würzburg"
alter = 49

satzmaske = "{name} ist {alter} Jahre alt und wohnt in {city}"

ergebnis = satzmaske.format(
    name=vorname,
    alter=alter,
    city=stadt
)

print(ergebnis)
````

Die Platzhalter können auch über ihre Position angesprochen werden:


````python
satzmaske = "{0} wohnt in {1} und ist {2} Jahre alt"

print(satzmaske.format(vorname, stadt, alter))
````

## 8. f-Strings

f-Strings sind heute meistens die angenehmste Lösung.

Vor den String kommt ein `f`, die Variablen stehen in geschweiften Klammern.


````python
name = "Max"
stadt = "Würzburg"
alter = 49

print(f"{name} wohnt in {stadt} und ist {alter} Jahre alt.")
````

### Rechnen im f-String


````python
preis = 100
steuer = 0.19

print(f"Bruttopreis: {preis + preis * steuer:.2f} Euro")
````

### Methoden im f-String


````python
text = "hello this is my world"

print(f"{text.title()}")
````

### Dictionary im f-String


````python
person = {"name": "John", "age": 19}

print(f"{person['name']} ist {person['age']} Jahre alt.")
````

## 9. Zahlen schön ausgeben

Wichtige Formatangaben:

- `.2f` – zwei Nachkommastellen
- `>8` – rechtsbündig
- `<8` – linksbündig
- `^8` – zentriert


````python
preis = 9.99875

print(f"{preis:.2f}")
print(f"{preis:>8.2f}")
print(f"{preis:<8.2f}")
print(f"{preis:^8.2f}")
````

Rechtsbündige Zahlen sind besonders bei Tabellen praktisch:


````python
print(f"{100:>8.3f}")
print(f"{99:>8.3f}")
print(f"{4:>8.3f}")
````

### Wichtig

Eine formatierte Zahl ist danach ein String.

Damit sollte man nicht weiterrechnen.


````python
zahl = 100
formatierte_zahl = f"{zahl:>8.2f}"

print(zahl, type(zahl))
print(formatierte_zahl, type(formatierte_zahl))
````

## 10. String-Methoden

Strings bringen viele praktische Methoden direkt mit.


````python
mein_string = "    hello mein, World!-------"

print("Länge:", len(mein_string))
print("Groß:", mein_string.upper())
print("Klein:", mein_string.lower())
print("Erster Buchstabe groß:", mein_string.capitalize())
print("Jedes Wort groß:", mein_string.title())
print("Leerzeichen weg:", mein_string.strip())
print("Leerzeichen und Bindestriche weg:", mein_string.strip().strip("-"))
print("Ersetzt:", mein_string.replace("World", "Mars"))
print("Zentriert:", mein_string.center(60, "#"))
````

`strip()` entfernt nur am Anfang und Ende.

Es räumt nicht mitten im Text auf.


## 11. Strings zerlegen und zusammensetzen

Mit `split()` wird ein String geteilt.


````python
zeichenkette = "Apfel,Birne,Banane"

obst = zeichenkette.split(",")

print(obst)
print(type(obst))
````

Mit `join()` werden die einzelnen Teile wieder verbunden.


````python
obst = ["Apfel", "Birne", "Banane"]

print(" | ".join(obst))
````

## 12. Prüffunktionen

Viele String-Methoden liefern `True` oder `False`.


````python
datei = "bild.jpg"
test = "id_123"
leerzeichen = " "

print("isspace:", leerzeichen.isspace())
print("startswith:", datei.startswith("bild"))
print("endswith:", datei.endswith(".jpg"))
print("isalnum:", test.isalnum())
print("isdecimal:", test.isdecimal())
print("isdigit:", test.isdigit())
print("isnumeric:", test.isnumeric())
print("isidentifier:", test.isidentifier())
print("isalpha:", test.isalpha())
````

## 13. `find()`

`find()` sucht die Position des ersten Treffers.

Wird nichts gefunden, kommt `-1` zurück.


````python
text = "Hallo Welt"

print(text.find("l"))
print(text.find("Welt"))
print(text.find("Python"))
````

## 14. Slicing

Mit Slicing können wir Teile eines Strings auswählen.

Der erste Index ist `0`.


````python
text = "Hallo, Welt!"

print(text[0])
print(text[7])
print(text[-2])
print(text[0:6])
print(text[:6])
print(text[5:])
print(text[-5:])
print(text[-5:-2])
````

### Schrittweite

Der dritte Wert ist die Schrittweite:


````python
text = "ABCDEFGHIJ"

print(text[::3])
print(text[8:1:-2])
````

### String umdrehen


````python
bitfolge = "11110000"

print(bitfolge[::-1])
````

## 15. HTML-Entitäten

`&amp;` ist die HTML-Schreibweise für `&`.

Python wandelt das nicht automatisch um.


````python
import html

text = "Vance &amp; Hince"

print(text)
print(html.unescape(text))
````

### Noch ein Hinweis zu `isalnum()`

`isalnum()` bedeutet: Der String besteht nur aus Buchstaben und Zahlen.

Ein Unterstrich zählt dabei nicht mit:


````python
print("id123".isalnum())     # True
print("id_123".isalnum())    # False
````

### Kleiner Hinweis zu Raw Strings

Raw Strings sind praktisch für Windows-Pfade und Regex.

Ein Raw String darf aber nicht mit einem einzelnen Backslash enden:

```python
r"C:\Ordner\"
```

Das wäre ungültig.

## Aufgabe 1: Wörter eines Satzes umkehren

Gegeben ist:

```python
text = "Hallo du schöne Sonne"
```

Das Ergebnis soll lauten:

```text
Sonne schöne du Hallo
```

### Hinweis

Nutze `split()`, Slicing mit `[::-1]` und `join()`.


### Dein Versuch


````python
text = "Hallo du schöne Sonne"

# Schreibe deine Lösung hier darunter.
````

### Musterlösung


````python
text = "Hallo du schöne Sonne"

woerter = text.split()
umgekehrte_woerter = woerter[::-1]
ergebnis = " ".join(umgekehrte_woerter)

print(ergebnis)
````

### Erklärung

`split()` macht aus dem Satz eine Liste mit einzelnen Wörtern. `[::-1]` dreht die Reihenfolge um. `join()` setzt alles wieder zu einem Satz zusammen.
