## 1. Warum braucht man Schleifen?

Ohne Schleife könnte man so etwas schreiben:


````python
print("Hallo")
print("Hallo")
print("Hallo")
print("Hallo")
print("Hallo")
````

Das funktioniert. Schön ist es aber nicht.

Mit einer Schleife geht es kürzer:


````python
for i in range(5):
    print("Hallo")
````

Beide Beispiele geben fünfmal `Hallo` aus.

Der Unterschied ist nur: Bei der Schleife müssen wir den gleichen Befehl nicht ständig neu schreiben.


## 2. Die `for`-Schleife

Der grundsätzliche Aufbau sieht so aus:

```python
for variable in bereich:
    anweisung
```

Für den Anfang nehmen wir meistens `range()` als Bereich.


````python
for i in range(5):
    print(i)
````

Die Ausgabe lautet:

```text
0
1
2
3
4
```

Das wirkt am Anfang etwas schräg:

`range(5)` läuft fünfmal, startet aber bei `0` und endet deshalb bei `4`.

> **Merksatz:** Der Endwert bei `range()` ist nicht mehr dabei.


## 3. Was steckt in der Schleifenvariable?

In diesem Beispiel heißt die Schleifenvariable `i`:


````python
for i in range(5):
    print("Aktueller Wert von i:", i)
````

Bei jedem Durchlauf bekommt `i` den nächsten Wert.

Also nacheinander:

```text
0, 1, 2, 3, 4
```

Die Schleifenvariable kann auch anders heißen:


````python
for zahl in range(5):
    print("Aktuelle Zahl:", zahl)
````

`zahl` ist hier meistens verständlicher als irgendein kryptischer Buchstabe.

Bei sehr kleinen Schleifen sieht man trotzdem oft `i`. Das ist eine übliche Abkürzung für einen Zähler.


## 4. Was bedeutet `_`?

Manchmal brauchen wir nur die Wiederholung, aber nicht die aktuelle Zahl.

Beispiel:


````python
for _ in range(5):
    print("Hallo")
````

Das `_` ist technisch gesehen ein ganz normaler Variablenname.

Es bedeutet sinngemäß:

> Die Schleifenvariable ist zwar da, aber ihr Wert interessiert mich hier nicht.

Normalerweise könnte man auch schreiben:


````python
for i in range(5):
    print("Hallo")
````

Da `i` im Code aber gar nicht benutzt wird, ist `_` die ehrlichere Schreibweise.

Wichtig:

`_` ist kein besonderes Python-Schlüsselwort. Es ist nur eine übliche Vereinbarung.


## 5. Einrückung

Alles, was eingerückt unter der Schleife steht, gehört zur Schleife.


````python
for i in range(3):
    print("Durchlauf:", i)
    print("Diese Zeile gehört ebenfalls zur Schleife.")

print("Diese Zeile kommt erst nach der Schleife.")
````

Die Einrückung ist in Python nicht nur Optik. Sie entscheidet darüber, welcher Code zusammengehört.

Ohne richtige Einrückung meckert Python – und diesmal zu Recht.

## 7. Schleifen über Strings

Eine Schleife kann auch Zeichen eines Strings nacheinander durchlaufen.


````python
wort = "Hallo"

for zeichen in wort:
    print(zeichen)
````

Bei jedem Durchlauf steckt in `zeichen` genau ein Buchstabe.


## 8. Werte zusammenrechnen

Schleifen können auch Werte sammeln.

Dafür brauchen wir eine Variable, die vor der Schleife startet.


````python
summe = 0

for zahl in range(1, 6):
    summe += zahl

print("Summe:", summe)
````

Die Schleife rechnet:

```text
0 + 1 + 2 + 3 + 4 + 5
```

Das Ergebnis ist `15`.


## 9. Unser Kommazahlen-Beispiel als Schleife

Im Grundlagen-Notebook hatten wir `0.1` mehrmals ausgeschrieben.

Mit einer Schleife geht es kürzer:


````python
wert = 0

for _ in range(10):
    wert += 0.1

print(wert)
print(round(wert, 2))
````

Jetzt passt auch die Erklärung zu `_`:

Wir wollen einfach zehnmal `0.1` addieren. Die aktuelle Nummer des Durchlaufs brauchen wir dafür nicht.


## 10. Die `while`-Schleife

Eine `while`-Schleife läuft so lange, wie eine Bedingung wahr ist.


### Wann nehme ich `for` und wann `while`?

Ganz grob:

- `for` nehme ich, wenn ich weiß, wie oft etwas laufen soll oder wenn ich eine Liste bzw. einen String durchlaufe.
- `while` nehme ich, wenn etwas so lange laufen soll, bis sich eine Bedingung ändert.


````python
zahl = 1

while zahl <= 5:
    print(zahl)
    zahl += 1
````

Die Schleife läuft, solange `zahl <= 5` wahr ist.

Wichtig ist:

```python
zahl += 1
```

Ohne diese Änderung würde die Bedingung immer wahr bleiben. Dann hätten wir eine Endlosschleife – und Python würde fröhlich bis zum Sankt-Nimmerleins-Tag weiterlaufen.


````python
# Achtung: Dieser Code würde endlos laufen.
# Deshalb bleibt er auskommentiert.

# zahl = 1
# while zahl <= 5:
#     print(zahl)
````

## 11. `break`

Mit `break` kann eine Schleife sofort beendet werden.


````python
for zahl in range(10):
    if zahl == 5:
        break

    print(zahl)
````

Sobald `zahl` den Wert `5` hat, wird die Schleife beendet.


## 12. `continue`

Mit `continue` wird nur der aktuelle Durchlauf übersprungen.


````python
for zahl in range(6):
    if zahl == 3:
        continue

    print(zahl)
````

Die `3` wird nicht ausgegeben. Danach läuft die Schleife normal weiter.

## Aufgabe 9: Passwort analysieren

Die Aufgabe stammt aus dem Dozentenmaterial.

Prüfe, ob ein Passwort:

- mindestens 8 Zeichen lang ist
- mindestens einen Großbuchstaben enthält
- mindestens einen Kleinbuchstaben enthält
- mindestens eine Zahl enthält

Hier lösen wir es bewusst mit einer normalen Schleife. So ist gut sichtbar, was bei jedem Zeichen geprüft wird.


### Dein Versuch


````python
passwort = "MeinGeheimes&§123"

# Schreibe deine Lösung hier darunter.
````

### Musterlösung


````python
passwort = "MeinGeheimes&§123"

hat_grossbuchstaben = False
hat_kleinbuchstaben = False
hat_zahl = False

for zeichen in passwort:
    if zeichen.isupper():
        hat_grossbuchstaben = True
    elif zeichen.islower():
        hat_kleinbuchstaben = True
    elif zeichen.isdigit():
        hat_zahl = True

lang_genug = len(passwort) >= 8

print("Mindestens 8 Zeichen:", lang_genug)
print("Großbuchstabe vorhanden:", hat_grossbuchstaben)
print("Kleinbuchstabe vorhanden:", hat_kleinbuchstaben)
print("Zahl vorhanden:", hat_zahl)

if lang_genug and hat_grossbuchstaben and hat_kleinbuchstaben and hat_zahl:
    print("Das Passwort erfüllt alle Regeln.")
else:
    print("Das Passwort erfüllt nicht alle Regeln.")
````

### Erklärung

Die drei Prüfvariablen starten mit `False`. Sobald die Schleife ein passendes Zeichen findet, wird die jeweilige Variable auf `True` gesetzt.
