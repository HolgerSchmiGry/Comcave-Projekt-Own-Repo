<a id="warum-braucht-man-schleifen"></a>
## 1. Warum braucht man Schleifen?

Ohne Schleife könnte man so etwas schreiben:


````python
print("Hallo")
print("Hallo")
print("Hallo")
print("Hallo")
print("Hallo")
````

Das funktioniert. Schön ist es aber nicht.

Mit einer Schleife geht es kürzer:


````python
for i in range(5):
    print("Hallo")
````

Beide Beispiele geben fünfmal `Hallo` aus.

Der Unterschied ist nur: Bei der Schleife müssen wir den gleichen Befehl nicht ständig neu schreiben.


<a id="die-for-schleife"></a>
## 2. Die `for`-Schleife

Der grundsätzliche Aufbau sieht so aus:

```python
for variable in bereich:
    anweisung
```

Für den Anfang nehmen wir meistens `range()` als Bereich.


````python
for i in range(5):
    print(i)
````

Die Ausgabe lautet:

```text
0
1
2
3
4
```

Das wirkt am Anfang etwas schräg:

`range(5)` läuft fünfmal, startet aber bei `0` und endet deshalb bei `4`.

> **Merksatz:** Der Endwert bei `range()` ist nicht mehr dabei.


<a id="was-steckt-in-der-schleifenvariable"></a>
## 3. Was steckt in der Schleifenvariable?

In diesem Beispiel heißt die Schleifenvariable `i`:


````python
for i in range(5):
    print("Aktueller Wert von i:", i)
````

Bei jedem Durchlauf bekommt `i` den nächsten Wert.

Also nacheinander:

```text
0, 1, 2, 3, 4
```

Die Schleifenvariable kann auch anders heißen:


````python
for zahl in range(5):
    print("Aktuelle Zahl:", zahl)
````

`zahl` ist hier meistens verständlicher als irgendein kryptischer Buchstabe.

Bei sehr kleinen Schleifen sieht man trotzdem oft `i`. Das ist eine übliche Abkürzung für einen Zähler.


<a id="was-bedeutet-_"></a>
## 4. Was bedeutet `_`?

Manchmal brauchen wir nur die Wiederholung, aber nicht die aktuelle Zahl.

Beispiel:


````python
for _ in range(5):
    print("Hallo")
````

Das `_` ist technisch gesehen ein ganz normaler Variablenname.

Es bedeutet sinngemäß:

> Die Schleifenvariable ist zwar da, aber ihr Wert interessiert mich hier nicht.

Normalerweise könnte man auch schreiben:


````python
for i in range(5):
    print("Hallo")
````

Da `i` im Code aber gar nicht benutzt wird, ist `_` die ehrlichere Schreibweise.

Wichtig:

`_` ist kein besonderes Python-Schlüsselwort. Es ist nur eine übliche Vereinbarung.


<a id="einrückung"></a>
## 5. Einrückung

Alles, was eingerückt unter der Schleife steht, gehört zur Schleife.


````python
for i in range(3):
    print("Durchlauf:", i)
    print("Diese Zeile gehört ebenfalls zur Schleife.")

print("Diese Zeile kommt erst nach der Schleife.")
````

Die Einrückung ist in Python nicht nur Optik. Sie entscheidet darüber, welcher Code zusammengehört.

Ohne richtige Einrückung meckert Python – und diesmal zu Recht.

Schleifen können auch Werte sammeln.

Dafür brauchen wir eine Variable, die vor der Schleife startet.


````python
summe = 0

for zahl in range(1, 6):
    summe += zahl

print("Summe:", summe)
````

<a id="unser-kommazahlen-beispiel-als-schleife"></a>
## 9. Unser Kommazahlen-Beispiel als Schleife

Im Grundlagen-Notebook hatten wir `0.1` mehrmals ausgeschrieben.

Mit einer Schleife geht es kürzer:


````python
wert = 0

for _ in range(10):
    wert += 0.1

print(wert)
print(round(wert, 2))
````

Jetzt passt auch die Erklärung zu `_`:

Wir wollen einfach zehnmal `0.1` addieren. Die aktuelle Nummer des Durchlaufs brauchen wir dafür nicht.


<a id="die-while-schleife"></a>
## 10. Die `while`-Schleife

Eine `while`-Schleife läuft so lange, wie eine Bedingung wahr ist.


### Wann nehme ich `for` und wann `while`?

Ganz grob:

- `for` nehme ich, wenn ich weiß, wie oft etwas laufen soll oder wenn ich eine Liste bzw. einen String durchlaufe.
- `while` nehme ich, wenn etwas so lange laufen soll, bis sich eine Bedingung ändert.


````python
zahl = 1

while zahl <= 5:
    print(zahl)
    zahl += 1
````

Die Schleife läuft, solange `zahl <= 5` wahr ist.

Wichtig ist:

```python
zahl += 1
```

Ohne diese Änderung würde die Bedingung immer wahr bleiben. Dann hätten wir eine Endlosschleife – und Python würde fröhlich bis zum Sankt-Nimmerleins-Tag weiterlaufen.


````python
# Achtung: Dieser Code würde endlos laufen.
# Deshalb bleibt er auskommentiert.

# zahl = 1
# while zahl <= 5:
#     print(zahl)
````

<a id="break"></a>
## 11. `break`

Mit `break` kann eine Schleife sofort beendet werden.


````python
for zahl in range(10):
    if zahl == 5:
        break

    print(zahl)
````

Sobald `zahl` den Wert `5` hat, wird die Schleife beendet.


<a id="continue"></a>
## 12. `continue`

Mit `continue` wird nur der aktuelle Durchlauf übersprungen.


````python
for zahl in range(6):
    if zahl == 3:
        continue

    print(zahl)
````

Die `3` wird nicht ausgegeben. Danach läuft die Schleife normal weiter.

## Aufgabe 9: Passwort analysieren

Die Aufgabe stammt aus dem Dozentenmaterial.

Prüfe, ob ein Passwort:

- mindestens 8 Zeichen lang ist
- mindestens einen Großbuchstaben enthält
- mindestens einen Kleinbuchstaben enthält
- mindestens eine Zahl enthält

Hier lösen wir es bewusst mit einer normalen Schleife. So ist gut sichtbar, was bei jedem Zeichen geprüft wird.


### Dein Versuch


````python
passwort = "MeinGeheimes&§123"

# Schreibe deine Lösung hier darunter.
````

### Musterlösung


````python
passwort = "MeinGeheimes&§123"

hat_grossbuchstaben = False
hat_kleinbuchstaben = False
hat_zahl = False

for zeichen in passwort:
    if zeichen.isupper():
        hat_grossbuchstaben = True
    elif zeichen.islower():
        hat_kleinbuchstaben = True
    elif zeichen.isdigit():
        hat_zahl = True

lang_genug = len(passwort) >= 8

print("Mindestens 8 Zeichen:", lang_genug)
print("Großbuchstabe vorhanden:", hat_grossbuchstaben)
print("Kleinbuchstabe vorhanden:", hat_kleinbuchstaben)
print("Zahl vorhanden:", hat_zahl)

if lang_genug and hat_grossbuchstaben and hat_kleinbuchstaben and hat_zahl:
    print("Das Passwort erfüllt alle Regeln.")
else:
    print("Das Passwort erfüllt nicht alle Regeln.")
````

### Erklärung

Die drei Prüfvariablen starten mit `False`. Sobald die Schleife ein passendes Zeichen findet, wird die jeweilige Variable auf `True` gesetzt.


<a id="for-und-while-schleifen-im-direkten-vergleich"></a>
## For- und While-Schleifen im direkten Vergleich

Schleifen wiederholen Anweisungen.

Die beiden wichtigsten Schleifenarten in Python sind:

- `for`
- `while`

### Wann verwendet man `for`?

Eine `for`-Schleife ist sinnvoll, wenn:

- eine Sammlung durchlaufen wird
- eine feste Anzahl von Wiederholungen bekannt ist
- ein Bereich mit `range()` durchlaufen wird

````python
for zahl in range(1, 6):
    print(zahl)
````

Ausgabe:

```text
1
2
3
4
5
```

★ **Merksatz**

> `for` eignet sich für bekannte Wiederholungen oder vorhandene Sammlungen.

### Wann verwendet man `while`?

Eine `while`-Schleife ist sinnvoll, wenn vorher nicht sicher feststeht, wie oft etwas wiederholt werden muss.

Sie läuft so lange, wie ihre Bedingung wahr ist.

````python
zahl = 1

while zahl <= 5:
    print(zahl)
    zahl += 1
````

Wichtig ist hier:

```python
zahl += 1
```

Ohne diese Veränderung würde `zahl <= 5` dauerhaft wahr bleiben.

### Direkter Vergleich

| Frage | `for` | `while` |
|---|---|---|
| Anzahl der Durchläufe bekannt? | meist ja | oft nein |
| Sammlung durchlaufen? | sehr gut geeignet | möglich, aber untypisch |
| Bedingung entscheidet über Ende? | indirekt | direkt |
| Gefahr einer Endlosschleife? | gering | höher |

<a id="kopfgesteuerte-schleife"></a>
## Kopfgesteuerte Schleife

Bei einer **kopfgesteuerten Schleife** wird die Bedingung geprüft, bevor der Schleifenblock ausgeführt wird.

Python-`while` ist kopfgesteuert.

````python
zahl = 10

while zahl < 5:
    print(zahl)
````

Hier erfolgt keine Ausgabe.

Warum?

Die Bedingung:

```python
zahl < 5
```

ist bereits beim ersten Prüfen falsch.

★ **Merksatz**

> Eine kopfgesteuerte Schleife kann auch nullmal ausgeführt werden.

### Ablauf einer kopfgesteuerten Schleife

```text
Start
  ↓
Bedingung prüfen
  ├── wahr  → Schleifenblock → zurück zur Bedingung
  └── falsch → Ende
```

<a id="fußgesteuerte-schleife"></a>
## Fußgesteuerte Schleife

Bei einer **fußgesteuerten Schleife** wird der Schleifenblock zuerst ausgeführt.

Die Bedingung wird erst danach geprüft.

Deshalb läuft eine fußgesteuerte Schleife mindestens einmal.

### Java: `do-while`

Java besitzt dafür eine eigene Syntax:

```java
do {
    System.out.println("Wird mindestens einmal ausgeführt");
} while (bedingung);
```

### Python besitzt kein eigenes `do-while`

In Python wird dieses Verhalten meist mit `while True` und `break` nachgebildet.

````python
while True:
    eingabe = "ende"

    print("Eingabe:", eingabe)

    if eingabe == "ende":
        break
````

Der Schleifenblock wird zuerst ausgeführt.

Erst danach entscheidet die `if`-Bedingung, ob mit `break` beendet wird.

### Praxisbeispiel: Positive Zahl verlangen

````python
while True:
    eingabe = "5"
    zahl = int(eingabe)

    if zahl > 0:
        break

print("Gültige Zahl:", zahl)
````

In einer echten Eingabe würde man schreiben:

```python
while True:
    zahl = int(input("Bitte eine positive Zahl eingeben: "))

    if zahl > 0:
        break
```

Für ein Notebook ist die feste Beispielvariable angenehmer, weil die Zelle ohne manuelle Eingabe vollständig läuft.

### Ablauf einer fußgesteuerten Schleife

```text
Start
  ↓
Schleifenblock ausführen
  ↓
Abbruchbedingung prüfen
  ├── erfüllt → Ende
  └── nicht erfüllt → Schleifenblock erneut ausführen
```

<a id="kopfgesteuert-und-fußgesteuert-im-vergleich"></a>
## Kopfgesteuert und fußgesteuert im Vergleich

| Eigenschaft | Kopfgesteuert | Fußgesteuert |
|---|---|---|
| Prüfung | vor dem Schleifenblock | nach dem Schleifenblock |
| mindestens ein Durchlauf? | nein | ja |
| Python-Syntax | `while bedingung:` | Nachbildung mit `while True` |
| Java-Syntax | `while (...)` | `do ... while (...)` |

<a id="endlosschleifen-erkennen"></a>
## Endlosschleifen erkennen

Eine Endlosschleife entsteht, wenn ihre Abbruchbedingung nie falsch wird.

Dieses Beispiel wäre gefährlich:

```python
zahl = 1

while zahl <= 5:
    print(zahl)
```

`zahl` wird nie verändert.

Deshalb bleibt die Bedingung dauerhaft wahr.

Richtig:

````python
zahl = 1

while zahl <= 5:
    print(zahl)
    zahl += 1
````

<a id="break-continue-und-else-bei-schleifen"></a>
## `break`, `continue` und `else` bei Schleifen

`break` beendet eine Schleife sofort.

````python
for zahl in range(1, 10):
    if zahl == 5:
        break

    print(zahl)
````

`continue` überspringt nur den aktuellen Durchlauf.

````python
for zahl in range(1, 6):
    if zahl == 3:
        continue

    print(zahl)
````

Python erlaubt außerdem ein `else` nach einer Schleife.

Das `else` wird ausgeführt, wenn die Schleife normal endet und nicht mit `break` beendet wurde.

````python
for zahl in range(1, 4):
    print(zahl)
else:
    print("Schleife normal beendet")
````

Mit `break`:

````python
for zahl in range(1, 6):
    if zahl == 3:
        break

    print(zahl)
else:
    print("Diese Ausgabe erscheint nicht")
````

## Eine Liste mit Index durchlaufen

Eine Liste kann auf mehrere Arten mit einem Index durchlaufen werden.

Alle folgenden Varianten funktionieren. Sie unterscheiden sich aber in Lesbarkeit, Aufwand und Fehlerrisiko.

### Variante 1: Manueller Zähler

Der Index wird selbst verwaltet.

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]
i = 0

for obst in obstkorb:
    print(f"Nr. {i} = {obst}")
    i += 1
````

Diese Variante ist gut, um das Prinzip eines Zählers zu verstehen.

Nachteil:

- `i` muss vor der Schleife angelegt werden
- `i += 1` darf nicht vergessen werden
- der Zähler kann leicht falsch verändert werden

### Variante 2: `range(len(...))`

Hier werden die gültigen Indizes der Liste erzeugt.

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]

for i in range(len(obstkorb)):
    print(f"Nr. {i} = {obstkorb[i]}")
````

Erklärung:

```python
len(obstkorb)
```

liefert die Anzahl der Elemente.

```python
range(len(obstkorb))
```

erzeugt die Indizes:

```text
0, 1, 2
```

Diese Variante ist korrekt, aber in Python meist weniger elegant als `enumerate()`.

### Variante 3: `enumerate()` – die typische Python-Lösung

`enumerate()` liefert automatisch:

- den Index
- das aktuelle Element

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]

for index, obst in enumerate(obstkorb):
    print(f"Nr. {index} = {obst}")
````

Das ist meistens die beste Lösung, wenn Index und Wert gleichzeitig benötigt werden.

★ **Merksatz**

> `enumerate()` liefert Index und Element zusammen.

### Nummerierung ab 1

Für eine benutzerfreundliche Nummerierung:

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]

for nummer, obst in enumerate(obstkorb, start=1):
    print(f"Nr. {nummer} = {obst}")
````

Ausgabe:

```text
Nr. 1 = Apfel
Nr. 2 = Banane
Nr. 3 = Kiwi
```

### Variante 4: `zip()`

`zip()` verbindet mehrere durchlaufbare Objekte paarweise.

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]
nummern = range(len(obstkorb))

for index, obst in zip(nummern, obstkorb):
    print(f"Nr. {index} = {obst}")
````

Diese Variante funktioniert, ist hier aber unnötig aufwendig.

`zip()` ist sinnvoller, wenn zwei echte Sammlungen parallel verarbeitet werden:

```python
namen = ["Anna", "Ben", "Clara"]
punkte = [90, 75, 88]
```

````python
namen = ["Anna", "Ben", "Clara"]
punkte = [90, 75, 88]

for name, punktzahl in zip(namen, punkte):
    print(f"{name}: {punktzahl} Punkte")
````

### Variante 5: Mit `while`

Die Liste kann auch über einen Index mit `while` durchlaufen werden.

````python
obstkorb = ["Apfel", "Banane", "Kiwi"]
i = 0

while i < len(obstkorb):
    print(f"Nr. {i} = {obstkorb[i]}")
    i += 1
````

Diese Variante ist korrekt.

Für das einfache Durchlaufen einer Liste ist `for` jedoch meistens besser geeignet.

`while` passt eher, wenn eine Bedingung die Wiederholung steuert.

### Direkter Vergleich

| Variante | Verständlich? | Typisch für Python? | Besonderheit |
|---|---:|---:|---|
| manueller Zähler | Ja | teilweise | gut zum Lernen |
| `range(len(...))` | Ja | gelegentlich | direkter Indexzugriff |
| `enumerate()` | Sehr | Ja | Index und Wert zusammen |
| `zip()` | Ja | Ja | mehrere Sammlungen parallel |
| `while` | Ja | selten für Listen | bedingungsgesteuert |

### Welche Variante sollte man verwenden?

Ohne Index:

```python
for obst in obstkorb:
    print(obst)
```

Mit Index:

```python
for index, obst in enumerate(obstkorb):
    print(index, obst)
```

Zwei Listen parallel:

```python
for name, punktzahl in zip(namen, punkte):
    print(name, punktzahl)
```

Manuelle Indexsteuerung nur dann, wenn sie wirklich gebraucht wird.

## Dictionaries mit Schleifen durchlaufen

Ein Dictionary speichert Schlüssel-Wert-Paare.

Beispiel:

```python
person = {
    "name": "Susi",
    "alter": 30,
    "stadt": "Köln"
}
```

Beim Durchlaufen eines Dictionaries gibt es mehrere sinnvolle Varianten.

### Nur die Schlüssel durchlaufen

Wird ein Dictionary direkt in einer `for`-Schleife verwendet, liefert Python die Schlüssel.

````python
person = {
    "name": "Susi",
    "alter": 30,
    "stadt": "Köln"
}

for schluessel in person:
    print(schluessel)
````

Das ist gleichbedeutend mit:

````python
for schluessel in person.keys():
    print(schluessel)
````

★ **Merksatz**

> `for schluessel in dictionary:` durchläuft standardmäßig die Schlüssel.

### Nur die Werte durchlaufen

Mit `.values()` werden nur die Werte ausgegeben.

````python
for wert in person.values():
    print(wert)
````

Ausgabe:

```text
Susi
30
Köln
```

### Schlüssel und Werte zusammen durchlaufen

Mit `.items()` erhält man Schlüssel und Wert gemeinsam.

````python
for schluessel, wert in person.items():
    print(f"{schluessel}: {wert}")
````

`items()` liefert bei jedem Durchlauf ein Tupel:

```text
("name", "Susi")
("alter", 30)
("stadt", "Köln")
```

Durch Unpacking werden die beiden Werte direkt auf Variablen verteilt:

```python
for schluessel, wert in person.items():
```

### Direkter Vergleich

| Schreibweise | Ergebnis |
|---|---|
| `for x in person:` | Schlüssel |
| `for x in person.keys():` | Schlüssel |
| `for x in person.values():` | Werte |
| `for k, v in person.items():` | Schlüssel und Werte |

### Auf einen Wert über den Schlüssel zugreifen

Auch über die Schlüssel kann auf Werte zugegriffen werden:

````python
for schluessel in person:
    print(f"{schluessel}: {person[schluessel]}")
````

Diese Variante funktioniert.

Wenn Schlüssel und Wert gleichzeitig gebraucht werden, ist `.items()` meist lesbarer.

### Dictionary mit `enumerate()` nummerieren

Auch Dictionaries können nummeriert ausgegeben werden.

````python
for nummer, (schluessel, wert) in enumerate(person.items(), start=1):
    print(f"{nummer}. {schluessel}: {wert}")
````

Hier passiert zweifaches Unpacking:

```python
nummer, (schluessel, wert)
```

- `nummer` kommt von `enumerate()`
- `schluessel` und `wert` kommen von `items()`

### Verschachteltes Dictionary durchlaufen

Ein verschachteltes Dictionary enthält weitere Dictionaries.

````python
personen = {
    "person_1": {
        "name": "Susi",
        "alter": 30
    },
    "person_2": {
        "name": "Max",
        "alter": 25
    }
}

for person_id, daten in personen.items():
    print(person_id)

    for schluessel, wert in daten.items():
        print(f"  {schluessel}: {wert}")
````

Die innere Schleife verarbeitet die Werte des jeweiligen Unter-Dictionaries.

### Liste mit Dictionaries durchlaufen

Sehr häufig begegnet dir eine Liste, die mehrere Dictionaries enthält.

````python
produkte = [
    {
        "name": "Tastatur",
        "preis": 49.99
    },
    {
        "name": "Maus",
        "preis": 29.99
    }
]

for produkt in produkte:
    print(
        f'{produkt["name"]}: '
        f'{produkt["preis"]:.2f} Euro'
    )
````

Diese Struktur ist typisch für:

- JSON-Daten
- API-Antworten
- Datenbankergebnisse
- Produktlisten

### Nur bestimmte Einträge ausgeben

Bedingungen können direkt innerhalb der Schleife verwendet werden.

````python
preise = {
    "Tastatur": 49.99,
    "Maus": 29.99,
    "Monitor": 199.99
}

for produkt, preis in preise.items():
    if preis >= 50:
        print(f"{produkt}: {preis:.2f} Euro")
````

### Dictionary während des Durchlaufens verändern

Vorsicht:

Ein Dictionary sollte nicht gleichzeitig in seiner Größe verändert und durchlaufen werden.

Dieses Beispiel wäre problematisch:

```python
for schluessel in person:
    person["neu"] = "Wert"
```

Dabei kann ein Laufzeitfehler entstehen:

```text
RuntimeError: dictionary changed size during iteration
```

Sicherer ist eine Kopie der Schlüssel:

````python
daten = {
    "a": 1,
    "b": 2
}

for schluessel in list(daten.keys()):
    daten[f"{schluessel}_kopie"] = daten[schluessel]

print(daten)
````

Oder Änderungen werden zunächst gesammelt und anschließend ausgeführt.

### Sortiert durch ein Dictionary laufen

Dictionaries behalten ihre Einfügereihenfolge.

Für eine alphabetische Ausgabe kann `sorted()` verwendet werden:

````python
person = {
    "stadt": "Köln",
    "name": "Susi",
    "alter": 30
}

for schluessel in sorted(person):
    print(f"{schluessel}: {person[schluessel]}")
````

Nach Werten sortieren:

````python
preise = {
    "Tastatur": 49.99,
    "Maus": 29.99,
    "Monitor": 199.99
}

for produkt, preis in sorted(
    preise.items(),
    key=lambda eintrag: eintrag[1]
):
    print(f"{produkt}: {preis:.2f} Euro")
````

### Typische Fehler

1. Erwartet wird ein Wert, aber die Schleife liefert einen Schlüssel.
2. `.items()` wird vergessen, obwohl Schlüssel und Wert benötigt werden.
3. Schlüssel und Wert werden nicht korrekt entpackt.
4. Das Dictionary wird während des Durchlaufens vergrößert oder verkleinert.
5. Bei verschachtelten Dictionaries fehlt eine zweite Schleife.
