Regex ist am Anfang ein bisschen gewöhnungsbedürftig.

Man schaut auf so etwas:


Aber im Grunde ist Regex nur eine Beschreibung dafür, wonach in einem Text gesucht werden soll.

## 1. Das Modul `re`

Regex gehört zum eingebauten Modul `re`.


````python
import re
````

## 2. Alle Treffer finden

`re.findall()` liefert alle Treffer als Liste.


````python
import re

text = "Artikel A123 kostet 49 Euro und Artikel B456 kostet 99 €."

print(re.findall(r"[0-9]+", text))
````

Das Muster `[0-9]+` bedeutet:

- eine Ziffer von 0 bis 9
- `+` bedeutet: mindestens einmal


## 3. Artikelnummern finden


````python
print(re.findall(r"[A-Z][0-9]+", text))
````

Das Muster bedeutet:

- ein Großbuchstabe von A bis Z
- danach eine oder mehrere Zahlen


## 4. Preise finden


````python
treffer = re.findall(r"(\d+)\s+(Euro|euro|€|EUR)", text)

preise = [eintrag[0] for eintrag in treffer]

print(preise)
````

## 5. Wichtige Funktionen

- `re.findall()` – alle Treffer
- `re.search()` – ersten Treffer suchen
- `re.fullmatch()` – komplette Zeichenkette prüfen
- `re.sub()` – Treffer ersetzen


## 6. Wichtige Bausteine

- `[0-9]` oder `\d` – eine Zahl
- `[A-Z]` – Großbuchstabe
- `[a-z]` – Kleinbuchstabe
- `+` – mindestens einmal
- `*` – beliebig oft
- `?` – optional
- `{2}` – genau zweimal
- `{1,3}` – ein- bis dreimal
- `\.` – echter Punkt
- `\s` – Leerraum
- `^` – Anfang
- `$` – Ende


## Warum steht vor Regex oft ein `r`?

Regex enthält viele Backslashes.

Mit einem Raw String bleibt das Muster besser lesbar:

```python
r"\d+"
```

Ohne Raw String müsste man den Backslash oft doppelt schreiben:

```python
"\\d+"
```


## `re.search()` praktisch


````python
import re

text = "Artikel 123 ist verfügbar."
treffer = re.search(r"\d+", text)

if treffer:
    print("Gefunden:", treffer.group())
else:
    print("Nichts gefunden")
````

`re.search()` liefert einen Treffer oder `None`.

Mit `.group()` wird der gefundene Text ausgegeben.


## `re.sub()` praktisch

Mit `re.sub()` können Treffer ersetzt werden.


````python
import re

text = "Meine Nummer ist 12345."
ergebnis = re.sub(r"[0-9]", "*", text)

print(ergebnis)
````

## Gruppen und Auswahl

Runde Klammern bilden Gruppen:

```python
(Euro|EUR|€)
```

Das `|` bedeutet **oder**.

Eine Gruppe mit `?:` speichert den Treffer nicht extra:

```python
(?:Euro|EUR|€)
```

Das ist praktisch, wenn die Gruppe nur zum Aufbau des Musters gebraucht wird.

## Aufgabe 1: Firmen-E-Mail-Adressen

Erlaubt sind nur E-Mail-Adressen mit der Endung `@meinefirma.de`.

Vor dem `@` erlauben wir Buchstaben, Zahlen, Punkte, Unterstriche und Bindestriche.


````python
emails = [
    "test@meinefirma.de",
    "TEST@meinefirma.de",
    "test123@meinefirmade",
    "v.nachname@meinefirma.de",
    "v-nachname@meinefirma.de",
    "v_nachname@meinefirma.de",
    "nicht gut@meinefirma.de",
    "nicht @meinefirma.de",
    "ok@meinefirma.de ",
    " ok@meinefirma.de ",
    "thomas:eses@meinefirma.de ",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

muster = r"[A-Za-z0-9._-]+@meinefirma\.de"

for email in emails:
    gueltig = bool(re.fullmatch(muster, email, re.IGNORECASE))
    print(f"{email!r}: {gueltig}")
````

### Muster erklärt

- `[A-Za-z0-9._-]+` erlaubt die Zeichen vor dem `@`.
- `@meinefirma` ist fester Text.
- `\.` steht für einen echten Punkt.
- `de` ist die feste Endung.
- `fullmatch()` verlangt, dass die komplette Eingabe passt.

Die Adresse `test123@meinefirmade` ist deshalb **nicht** gültig. Im Kommentar der Dozentenaufgabe war sie versehentlich als „ok“ markiert.


## Aufgabe 2: Deutsches Kfz-Kennzeichen

Wir prüfen nur ein vereinfachtes Format:

- 1 bis 3 Buchstaben für den Ort
- Leerzeichen
- 1 bis 2 Buchstaben
- Leerzeichen
- 1 bis 4 Ziffern
- optional `E` oder `H`

Ob das Ortskürzel wirklich existiert, wird nicht geprüft.


````python
kennzeichen = [
    "WÜ AB 12",
    "B M 1",
    "B MW 1",
    "WES A 1000",
    "WES A 1000E",
    "DU ME 123",
    "DU ME 123H",
    " DU ME 123H ",
    " DU ME 12-3H ",
    "ABC A 123",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

muster = r"[A-ZÄÖÜ]{1,3} [A-Z]{1,2} [1-9][0-9]{0,3}[EH]?"

for eintrag in kennzeichen:
    gueltig = bool(re.fullmatch(muster, eintrag))
    print(f"{eintrag!r}: {gueltig}")
````

### Muster erklärt

- `[A-ZÄÖÜ]{1,3}`: ein bis drei Großbuchstaben
- ein echtes Leerzeichen
- `[A-Z]{1,2}`: ein oder zwei Großbuchstaben
- `[1-9][0-9]{0,3}`: eine Zahl ohne führende Null, insgesamt höchstens vier Ziffern
- `[EH]?`: `E` oder `H` darf einmal vorkommen, muss aber nicht

`WÜ AB 12` und `ABC A 123` passen formal zu diesem Muster. Ob `ABC` wirklich als Ortskürzel existiert, ist eine andere Frage.


## Aufgabe 3: Euro-Beträge

Erlaubt sind:

- ganze Beträge mit oder ohne Tausenderpunkte
- optional genau zwei Nachkommastellen
- Leerzeichen und Eurozeichen am Ende


````python
zahlen = [
    "10.000,00 €",
    "10.000 €",
    "10000 €",
    "10000,00 €",
    "10.000",
    "10000",
    "0,90 €",
    " 1000",
    "222 2",
    "0900",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

muster = r"(?:0|[1-9][0-9]*|[1-9][0-9]{0,2}(?:\.[0-9]{3})+)(?:,[0-9]{2})? €"

for zahl in zahlen:
    gueltig = bool(re.fullmatch(muster, zahl))
    print(f"{zahl!r}: {gueltig}")
````

### Muster erklärt

- `(?:0|...)` erlaubt die einzelne Null oder einen größeren Betrag.
- `[1-9][0-9]*` erlaubt Zahlen ohne Tausenderpunkte.
- `[1-9][0-9]{0,2}(?:\.[0-9]{3})+` erlaubt korrekt gruppierte Tausenderpunkte.
- `(?:,[0-9]{2})?` erlaubt optional genau zwei Nachkommastellen.
- ` €` verlangt Leerzeichen und Eurozeichen.

Die fehlende Trennung zwischen `"0,90 €"` und `" 1000"` in der ursprünglichen Python-Liste wurde hier korrigiert.


## Aufgabe 4: IPv4-Adresse

Jeder der vier Zahlenblöcke muss zwischen `0` und `255` liegen.


````python
ip_adressen = [
    "192.168.1.100",
    "127.0.0.1",
    "255.255.255.255",
    "256.168.1.1",
    "192.168.1",
    "192.168.01.1",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

block = r"(?:25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])"
muster = rf"{block}\.{block}\.{block}\.{block}"

for ip in ip_adressen:
    gueltig = bool(re.fullmatch(muster, ip))
    print(f"{ip!r}: {gueltig}")
````

### Muster erklärt

Ein Block wird in mehrere Zahlenbereiche zerlegt:

- `25[0-5]` erlaubt `250` bis `255`
- `2[0-4][0-9]` erlaubt `200` bis `249`
- `1[0-9]{2}` erlaubt `100` bis `199`
- `[1-9]?[0-9]` erlaubt `0` bis `99`

Zwischen den Blöcken steht `\.`, also ein echter Punkt.


## Aufgabe 5: Deutsche IBAN

Geprüft wird nur das Format:

- `DE`
- zwei Prüfziffern
- danach zwanzig weitere Ziffern
- wahlweise ohne Leerzeichen oder in Vierergruppen

Die echte IBAN-Prüfsumme wird hier nicht berechnet.


````python
iban_nummern = [
    "DE89370400440532013000",
    "DE89 3704 0044 0532 0130 00",
    "DE89 3704 0044 0532",
    "FR89 3704 0044 0532 0130 00",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

muster = r"(?:DE[0-9]{20}|DE[0-9]{2}(?: [0-9]{4}){4} [0-9]{2})"

for iban in iban_nummern:
    gueltig = bool(re.fullmatch(muster, iban))
    print(f"{iban!r}: {gueltig}")
````

### Muster erklärt

- `DE[0-9]{20}` prüft die Schreibweise ohne Leerzeichen.
- Die zweite Möglichkeit prüft die gruppierte Schreibweise.
- `|` bedeutet „oder“.
- `(?:...)` fasst Teile zusammen, ohne einen eigenen Treffer zu speichern.


## Aufgabe 6: MAC-Adresse

Erlaubt wird hier bewusst nur die Schreibweise mit Doppelpunkten:

```text
00:1A:2B:3C:4D:5E
```


````python
mac_adressen = [
    "00:1A:2B:3C:4D:5E",
    "00:1a:2b:3c:4d:5e",
    "00-1A-2B-3C-4D-5E",
    "00:1A:2B:3C:4D",
    "GG:1A:2B:3C:4D:5E",
]
````

### Dein Versuch


````python
import re

# Ergänze hier dein Muster und die Prüfung.
````

### Musterlösung


````python
import re

muster = r"(?:[0-9A-Fa-f]{2}:){5}[0-9A-Fa-f]{2}"

for mac in mac_adressen:
    gueltig = bool(re.fullmatch(muster, mac))
    print(f"{mac!r}: {gueltig}")
````

### Muster erklärt

- `[0-9A-Fa-f]{2}` erlaubt genau zwei hexadezimale Zeichen.
- Der Block mit Doppelpunkt wird fünfmal wiederholt.
- Danach folgt der sechste Block ohne abschließenden Doppelpunkt.


## Aufgabe 7: Passwortregeln mit Regex

Das Passwort braucht mindestens:

- 8 Zeichen
- einen Großbuchstaben
- einen Kleinbuchstaben
- eine Zahl
- ein Sonderzeichen

Leerzeichen zählen nicht als Sonderzeichen.


````python
passwort = "MeinGeheimes&§123"
````

### Dein Versuch


````python
import re

# Ergänze hier deine Prüfungen.
````

### Musterlösung


````python
import re

mindestens_acht_zeichen = bool(re.search(r".{8,}", passwort))
enthaelt_grossbuchstaben = bool(re.search(r"[A-Z]", passwort))
enthaelt_kleinbuchstaben = bool(re.search(r"[a-z]", passwort))
enthaelt_zahl = bool(re.search(r"[0-9]", passwort))
enthaelt_sonderzeichen = bool(re.search(r"[^A-Za-z0-9\s]", passwort))

print("Mindestens 8 Zeichen:", mindestens_acht_zeichen)
print("Großbuchstabe:", enthaelt_grossbuchstaben)
print("Kleinbuchstabe:", enthaelt_kleinbuchstaben)
print("Zahl:", enthaelt_zahl)
print("Sonderzeichen:", enthaelt_sonderzeichen)

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
    and enthaelt_sonderzeichen
):
    print("Das Passwort erfüllt alle Regeln.")
else:
    print("Das Passwort erfüllt nicht alle Regeln.")
````

### Muster erklärt

- `.{8,}` prüft mindestens acht beliebige Zeichen.
- `[A-Z]`, `[a-z]` und `[0-9]` suchen die verlangten Zeichengruppen.
- `[^A-Za-z0-9\s]` sucht ein Zeichen, das weder Buchstabe noch Zahl noch Leerraum ist.
- `bool(...)` macht aus einem Treffer `True` und aus `None` den Wert `False`.
