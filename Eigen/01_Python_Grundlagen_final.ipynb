## 1. Was ist eine Variable?

Eine Variable ist ein benannter Speicherplatz für einen Wert.

```python
alter = 49
```

Dabei gilt:

- `alter` ist der Variablenname.
- `49` ist der gespeicherte Wert.
- `=` ist der Zuweisungsoperator.

Das Gleichheitszeichen bedeutet hier nicht „ist gleich“, sondern:

> Weise den Wert auf der rechten Seite der Variablen auf der linken Seite zu.


````python
alter = 49
name = "Max"

print(alter)
print(name)
````

### Kurz-Check

Was steht links vom Zuweisungsoperator?

Was steht rechts davon?

Ändere im nächsten Codeblock die Werte und führe ihn erneut aus.


````python
stadt = "Mönchengladbach"
anzahl = 10

print(stadt)
print(anzahl)
````

## 2. Einfache Datentypen

Wichtige Datentypen:

- `int` – ganze Zahl
- `float` – Kommazahl
- `str` – Zeichenkette
- `bool` – Wahrheitswert `True` oder `False`
- `None` – kein Wert vorhanden

Mit `type()` kannst du den Datentyp anzeigen.


````python
test_number = 5
test_decimal = 3.5
test_string = "123"
test_bool = True
test_none = None

print(test_number, type(test_number))
print(test_decimal, type(test_decimal))
print(test_string, type(test_string))
print(test_bool, type(test_bool))
print(test_none, type(test_none))
````

### Wichtig: `"123"` ist keine Zahl

Obwohl der Inhalt wie eine Zahl aussieht, machen die Anführungszeichen daraus einen String.

```python
123      # int
"123"    # str
```


````python
zahl = 123
text = "123"

print(zahl, type(zahl))
print(text, type(text))
````

## 3. Python ist dynamisch und stark typisiert

Python ist nicht „untypisiert“.

Python ist:

- **dynamisch typisiert**: Der Datentyp wird automatisch aus dem aktuellen Wert erkannt.
- **stark typisiert**: Unpassende Datentypen können nicht einfach miteinander verrechnet werden.

Eine Variable kann später einen Wert mit einem anderen Datentyp erhalten.


````python
wert = 5
print(wert, type(wert))

wert = "Hallo"
print(wert, type(wert))
````

## 4. Typannotationen

Typannotationen zeigen, welcher Datentyp vorgesehen ist.

```python
alter: int = 49
```

Sie verbessern die Lesbarkeit und helfen der Entwicklungsumgebung.

Python erzwingt diesen Typ aber nicht automatisch.


````python
alter: int = 49
preis: float = 9.99
name: str = "Max"
ist_aktiv: bool = True

print(alter, preis, name, ist_aktiv)
````

## 5. Variablennamen

In Python wird für Variablen meistens `snake_case` verwendet.

Gut:

```python
menge_der_dateien = 20
ist_angemeldet = True
```

Wichtige Regeln:

- nicht mit einer Zahl beginnen
- keine Leerzeichen
- möglichst verständliche Namen
- Groß- und Kleinschreibung beachten
- reservierte Wörter nicht als Namen verwenden


````python
menge_der_dateien = 20
ist_angemeldet = True

print(menge_der_dateien)
print(ist_angemeldet)
````

### Reservierte Wörter

Wörter wie `if`, `for`, `while`, `class` oder `def` haben in Python bereits eine Bedeutung.

Sie dürfen nicht als Variablennamen verwendet werden.

Falsch:

```python
if = 5
```


### Groß- und Kleinschreibung

Python ist **case-sensitive**.

Diese Variablen sind verschieden:


````python
variable_a = "Hallo"
Variable_A = "Welt"

print(variable_a)
print(Variable_A)
````

## 6. Konstanten als Vereinbarung

Python besitzt keine echten Konstanten.

Werte, die sich nicht ändern sollen, werden aber üblicherweise komplett groß geschrieben.

```python
MEHRWERTSTEUER = 0.19
VERSANDKOSTEN = 4.90
```

Das ist eine Vereinbarung für Menschen, keine technische Sperre.


````python
MEHRWERTSTEUER = 0.19
VERSANDKOSTEN = 4.90

print(MEHRWERTSTEUER)
print(VERSANDKOSTEN)
````

## 7. Eingaben mit `input()`

Mit `input()` kann der Benutzer etwas eingeben.

Wichtig:

> `input()` liefert immer einen String.

Auch wenn du `100` eingibst, ist das Ergebnis zunächst Text.


````python
eingabe = input("Bitte etwas eingeben: ")

print(eingabe)
print(type(eingabe))
````

### Eingaben in Zahlen umwandeln

Für Berechnungen musst du den Text umwandeln:

```python
int(...)     # ganze Zahl
float(...)   # Kommazahl
```


````python
alter = int(input("Bitte das Alter eingeben: "))
preis = float(input("Bitte einen Preis eingeben: "))

print(alter, type(alter))
print(preis, type(preis))
````

### Typischer Fehler bei Eingaben

```python
zahl = input("Zahl: ")
print(zahl + 5)
```

Das funktioniert nicht, weil `zahl` ein String ist.

Richtig:


````python
zahl = int(input("Bitte eine ganze Zahl eingeben: "))

print(zahl + 5)
````

## 8. Typumwandlung

Wichtige Funktionen:

- `int()`
- `float()`
- `str()`
- `bool()`

Nicht jede Umwandlung ist möglich.


````python
print(int(5.5))
print(float("5.5"))
print(str(5.5))
print(int(True))
print(bool(0))
print(bool(1))
````

### Ungültige Umwandlung

Das hier erzeugt einen Fehler:

```python
int("Hallo")
```

Der Text `"Hallo"` kann nicht in eine Zahl umgewandelt werden.

Solche Fehler werden später mit `try` und `except` behandelt.


## 9. Unterschied zwischen `=` und `==`

`=` weist einen Wert zu:

```python
alter = 18
```

`==` vergleicht zwei Werte:

```python
alter == 18
```

Das Ergebnis eines Vergleichs ist `True` oder `False`.


````python
alter = 18

print(alter == 18)
print(alter == 20)
````

> **Merksatz:** Ein `=` speichert. Zwei `==` vergleichen.


## 10. Arithmetische Operatoren

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `//` Ganzzahldivision
- `%` Rest einer Division
- `**` Potenz


````python
zahl_1 = 10
zahl_2 = 5

print("Addition:", zahl_1 + zahl_2)
print("Subtraktion:", zahl_1 - zahl_2)
print("Multiplikation:", zahl_1 * zahl_2)
print("Division:", zahl_1 / zahl_2)
print("Ganzzahldivision:", 10 // 3)
print("Modulo:", 10 % 3)
print("Potenz:", zahl_1 ** zahl_2)
````

### Division, Ganzzahldivision und Modulo

```python
10 / 3
```

liefert eine Kommazahl.

```python
10 // 3
```

liefert nur den ganzzahligen Anteil.

```python
10 % 3
```

liefert den Rest.


````python
print(10 / 3)
print(10 // 3)
print(10 % 3)
````

## 11. Variablen verändern

Eine Variable kann mit ihrem bisherigen Wert weitergerechnet werden.


````python
zahl = 10

zahl = zahl + 5
print(zahl)

zahl += 5
print(zahl)

zahl -= 2
print(zahl)

zahl *= 3
print(zahl)
````

Die Kurzformen bedeuten:

- `zahl += 1` → `zahl = zahl + 1`
- `zahl -= 1` → `zahl = zahl - 1`
- `zahl *= 2` → `zahl = zahl * 2`
- `zahl /= 2` → `zahl = zahl / 2`


## 12. Wahrheitswerte beim Rechnen

Intern verhält sich:

- `True` wie `1`
- `False` wie `0`


````python
print(True + True)
print(True + False)
print(False + False)
````

Das ist technisch möglich, sollte aber nur verwendet werden, wenn der Zusammenhang verständlich bleibt.


## 13. Genauigkeit bei Kommazahlen

Kommazahlen werden im Computer binär gespeichert.

Einige Dezimalzahlen können dadurch nicht exakt dargestellt werden.


````python
wert = 0

for _ in range(10):
    wert += 0.1

print(wert)
print(round(wert, 2))
````

>Mit `round(wert, 2)` wird auf zwei Nachkommastellen gerundet.

>Für besonders genaue Geldberechnungen kann `Decimal` verwendet werden.


## **Hinweis**

Also, falls Du dich fragst was for "_" ist - Das _ ist einfach ein Variablenname.
Normalerweise könnte man schreiben:

>**for i in range(10):**
>>    **wert += 0.1**

Aber **i** wird innerhalb der Schleife gar nicht gebraucht. Deshalb verwendet man in Python oft auch nur _.

Das **_** bedeutet sinngemäß:

Die Schleifenvariable existiert zwar, aber ihr Wert interessiert mich nicht, denn die Anzahl des Schleifendurchlaufs wird nicht gebraucht.
