## 1. Wahrheitswerte

Eine Bedingung ergibt immer `True` oder `False`.


````python
print(5 > 3)
print(5 < 3)
print(5 == 5)
print(5 != 5)
````

## 2. Vergleichsoperatoren

- `==` gleich
- `!=` ungleich
- `<` kleiner
- `>` größer
- `<=` kleiner oder gleich
- `>=` größer oder gleich

Wichtig:

`=` weist zu.

`==` vergleicht.


````python
alter = 18

print(alter == 18)
print(alter != 18)
print(alter >= 18)
````

## 3. Die `if`-Bedingung

Mit `if` wird Code nur ausgeführt, wenn eine Bedingung wahr ist.


````python
alter = 20

if alter >= 18:
    print("Du bist volljährig.")
````

### Einrückung

Der eingerückte Code gehört zur Bedingung.

Python verwendet Einrückungen zur Strukturierung.


## 4. `if-else`

Mit `else` wird festgelegt, was passieren soll, wenn die Bedingung falsch ist.


````python
alter = 17

if alter >= 18:
    print("Volljährig")
else:
    print("Minderjährig")
````

## 5. `if-elif-else`

Mit `elif` können mehrere Fälle geprüft werden.


````python
alter = 16

if alter < 13:
    status = "Kind"
elif alter < 18:
    status = "Jugendlich"
else:
    status = "Erwachsen"

print(status)
````

Die Bedingungen werden von oben nach unten geprüft.

Sobald eine Bedingung wahr ist, wird dieser Block ausgeführt. Die restlichen Blöcke werden übersprungen.


## 6. Logische Operatoren

### `and`

Beide Bedingungen müssen wahr sein.


````python
alter = 20
hat_ticket = True

if alter >= 18 and hat_ticket:
    print("Einlass erlaubt")
````

### `or`

Mindestens eine Bedingung muss wahr sein.


````python
ist_admin = False
ist_moderator = True

if ist_admin or ist_moderator:
    print("Zugriff erlaubt")
````

### `not`

`not` kehrt einen Wahrheitswert um.


````python
ist_gesperrt = False

if not ist_gesperrt:
    print("Konto ist aktiv")
````

## 7. Verschachtelte Bedingungen

Eine Bedingung kann innerhalb einer anderen Bedingung stehen.


````python
alter = 20
hat_ticket = True

if alter >= 18:
    if hat_ticket:
        print("Einlass erlaubt")
    else:
        print("Ticket fehlt")
else:
    print("Zu jung")
````

Oft ist eine Verbindung mit `and` übersichtlicher.


## 8. Bedingter Ausdruck

Für kurze Entscheidungen gibt es eine kompakte Schreibweise.

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```


````python
alter = 17

status = "Erwachsen" if alter >= 18 else "Minderjährig"

print(status)
````

### Leserichtung

```text
Nimm den ersten Wert,
wenn die Bedingung wahr ist,
sonst nimm den zweiten Wert.
```


### Bedingter Ausdruck in einem f-String


````python
alter = 20

print(f"Status: {'Volljährig' if alter >= 18 else 'Minderjährig'}")
````

## 9. Verschachtelte bedingte Ausdrücke

Technisch ist das möglich:


````python
alter = 16

status = "Kind" if alter < 13 else "Jugendlich" if alter < 18 else "Erwachsen"

print(status)
````

Das wird schnell unübersichtlich.

Bei mehreren Fällen ist `if-elif-else` meistens besser.


## 10. Typische Fehler

### `=` statt `==`

Falsch:

```python
if alter = 18:
```

Richtig:

```python
if alter == 18:
```

### Doppelpunkt vergessen

Nach `if`, `elif` und `else` steht ein Doppelpunkt.

### Einrückung vergessen

Der zugehörige Code muss eingerückt werden.


## Aufgabe 1: Gerade oder ungerade

### Aufgabe

Prüfe, ob eine Zahl gerade oder ungerade ist.

### Hinweis

Eine gerade Zahl hat bei der Division durch 2 den Rest 0.

### Dein Versuch


### Musterlösung


````python
zahl = int(input("Bitte eine ganze Zahl eingeben: "))

if zahl % 2 == 0:
    print("Die Zahl ist gerade.")
else:
    print("Die Zahl ist ungerade.")
````

## Aufgabe 2: Notenbewertung

### Aufgabe

Gib abhängig von einer Note eine Bewertung aus:

- 1 → sehr gut
- 2 → gut
- 3 → befriedigend
- 4 → ausreichend
- alles andere → ungültige Note

### Dein Versuch


### Musterlösung


````python
note = int(input("Bitte eine Note eingeben: "))

if note == 1:
    print("Sehr gut")
elif note == 2:
    print("Gut")
elif note == 3:
    print("Befriedigend")
elif note == 4:
    print("Ausreichend")
else:
    print("Ungültige Note")
````

## Aufgabe 3: Passwort-Grundprüfung

### Aufgabe

Prüfe:

- mindestens 8 Zeichen
- mindestens ein Großbuchstabe
- mindestens ein Kleinbuchstabe
- mindestens eine Zahl

### Dein Versuch


### Musterlösung


````python
passwort = input("Bitte ein Passwort eingeben: ")

lang_genug = len(passwort) >= 8
hat_grossbuchstaben = any(zeichen.isupper() for zeichen in passwort)
hat_kleinbuchstaben = any(zeichen.islower() for zeichen in passwort)
hat_zahl = any(zeichen.isdigit() for zeichen in passwort)

if lang_genug and hat_grossbuchstaben and hat_kleinbuchstaben and hat_zahl:
    print("Passwort erfüllt alle Bedingungen.")
else:
    print("Passwort erfüllt nicht alle Bedingungen.")
````
