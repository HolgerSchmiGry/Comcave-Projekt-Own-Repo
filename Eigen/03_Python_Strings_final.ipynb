## 1. Was ist ein String?

Ein String ist eine Zeichenkette.

Strings stehen in einfachen oder doppelten Anführungszeichen.


````python
name = "Max"
stadt = 'Würzburg'
nummer_als_text = "123"

print(name, type(name))
print(stadt, type(stadt))
print(nummer_als_text, type(nummer_als_text))
````

## 2. Strings verbinden

Mit `+` können Strings verbunden werden.

Das nennt man Verkettung oder Konkatenation.


````python
text_1 = "Hallo"
text_2 = "Welt"

print(text_1 + " " + text_2)
````

Eine Zahl muss vor der Verkettung mit `str()` in Text umgewandelt werden.


````python
alter = 49

print("Alter: " + str(alter))
````

## 3. Anführungszeichen und Escaping

Enthält ein String selbst Anführungszeichen, gibt es mehrere Möglichkeiten.


````python
satz_1 = 'Hans sagt: "Hallo"'
satz_2 = "Hans sagt: \"Hallo\""
satz_3 = "Hans sagt: 'Hallo'"

print(satz_1)
print(satz_2)
print(satz_3)
````

## 4. Steuerzeichen

- `\n` – Zeilenumbruch
- `\r` – Wagenrücklauf
- `\t` – Tabulator
- `\\` – Backslash


````python
print("Hallo\nWelt")
print("Name:\tMax")
````

## 5. Raw Strings

Raw Strings sind bei Windows-Pfaden praktisch.

Vor den String wird ein `r` geschrieben.


````python
pfad = r"c:\windows\tmp"

print(pfad)
````

## 6. Mehrzeilige Strings

Dreifache Anführungszeichen erzeugen einen mehrzeiligen String.


````python
gedicht = '''Rosen sind rot,
Veilchen sind blau,
Python macht Spaß,
das weißt du genau!'''

print(gedicht)
````

Wichtig:

Ein mehrzeiliger String ist kein echter Kommentar.


## 7. Formatierung mit `format()`

Mit `format()` werden Werte in Platzhalter eingesetzt.


````python
vorname = "Max"
stadt = "Würzburg"
alter = 49

satzmaske = "{name} ist {alter} Jahre alt und wohnt in {city}"

ergebnis = satzmaske.format(
    name=vorname,
    alter=alter,
    city=stadt
)

print(ergebnis)
````

## 8. Formatierung mit f-Strings

f-Strings sind heute meistens die übersichtlichste Lösung.


````python
name = "Max"
stadt = "Würzburg"
alter = 49

print(f"{name} wohnt in {stadt} und ist {alter} Jahre alt.")
````

### Berechnungen in f-Strings


````python
preis = 100
steuer = 0.19

print(f"Bruttopreis: {preis + preis * steuer:.2f} Euro")
````

### Methoden in f-Strings


````python
text = "hello this is my world"

print(f"{text.title()}")
````

### Dictionaries in f-Strings


````python
person = {"name": "John", "age": 19}

print(f"{person['name']} ist {person['age']} Jahre alt.")
````

## 9. Zahlen formatieren

- `.2f` – zwei Nachkommastellen
- `>8` – rechtsbündig
- `<8` – linksbündig
- `^8` – zentriert


````python
preis = 9.99875

print(f"{preis:.2f}")
print(f"{preis:>8.2f}")
print(f"{preis:<8.2f}")
print(f"{preis:^8.2f}")
````

### Wichtig

Eine formatierte Zahl ist danach ein String.


````python
zahl = 100
formatierte_zahl = f"{zahl:>8.2f}"

print(zahl, type(zahl))
print(formatierte_zahl, type(formatierte_zahl))
````

## 10. String-Methoden

Wichtige Methoden:

- `upper()`
- `lower()`
- `capitalize()`
- `title()`
- `strip()`
- `lstrip()`
- `rstrip()`
- `replace()`
- `center()`
- `split()`
- `join()`
- `find()`


````python
mein_string = "    hello mein, World!-------"

print(mein_string.upper())
print(mein_string.lower())
print(mein_string.capitalize())
print(mein_string.title())
print(mein_string.strip())
print(mein_string.strip().strip("-"))
print(mein_string.replace("World", "Mars"))
print(mein_string.center(60, "#"))
````

### `split()` und `join()`


````python
zeichenkette = "Apfel,Birne,Banane"

obst = zeichenkette.split(",")
print(obst)

ergebnis = " | ".join(obst)
print(ergebnis)
````

### `find()`

`find()` liefert die Position des ersten Treffers.

Wird nichts gefunden, lautet das Ergebnis `-1`.


````python
text = "Hallo Welt"

print(text.find("Welt"))
print(text.find("Python"))
````

## 11. Prüffunktionen

- `isspace()`
- `startswith()`
- `endswith()`
- `isalnum()`
- `isdecimal()`
- `isdigit()`
- `isnumeric()`
- `isidentifier()`
- `isalpha()`
- `isupper()`
- `islower()`


````python
datei = "bild.jpg"
test = "id_123"

print(datei.startswith("bild"))
print(datei.endswith(".jpg"))
print(test.isalnum())
print(test.isidentifier())
print(test.isalpha())
````

## 12. Slicing

Strings können über Indizes angesprochen werden.

Der erste Index ist `0`.

Negative Indizes zählen von hinten.


````python
text = "Hallo, Welt!"

print(text[0])
print(text[7])
print(text[-2])
print(text[:6])
print(text[7:])
````

### Schrittweite und Umkehren

Aufbau:

```python
text[start:ende:schrittweite]
```

Mit `[::-1]` wird die Reihenfolge umgedreht.


````python
text = "ABCDEFGHIJ"

print(text[::2])
print(text[::-1])
print(text[8:1:-2])
````

## 13. HTML-Entitäten

`&amp;` ist die HTML-Schreibweise für `&`.

Python wandelt das nicht automatisch um.


````python
import html

text = "Vance &amp; Hince"

print(text)
print(html.unescape(text))
````

## Aufgabe 1: Passwort analysieren

### Aufgabe

Prüfe, ob ein Passwort:

- mindestens 8 Zeichen lang ist,
- einen Großbuchstaben enthält,
- einen Kleinbuchstaben enthält,
- eine Zahl enthält.

### Vorgehen

1. Länge mit `len()` prüfen
2. Zeichen mit `isupper()`, `islower()` und `isdigit()` untersuchen
3. Ergebnisse verbinden

### Dein Versuch


### Musterlösung


````python
passwort = "MeinGeheimes&§123"

mindestens_acht_zeichen = len(passwort) >= 8
enthaelt_grossbuchstaben = any(zeichen.isupper() for zeichen in passwort)
enthaelt_kleinbuchstaben = any(zeichen.islower() for zeichen in passwort)
enthaelt_zahl = any(zeichen.isdigit() for zeichen in passwort)

print("Mindestens 8 Zeichen:", mindestens_acht_zeichen)
print("Großbuchstabe:", enthaelt_grossbuchstaben)
print("Kleinbuchstabe:", enthaelt_kleinbuchstaben)
print("Zahl:", enthaelt_zahl)
````

## Aufgabe 2: Wörter umkehren

### Aufgabe

Aus:

```text
Hallo du schöne Sonne
```

soll werden:

```text
Sonne schöne du Hallo
```

### Vorgehen

1. mit `split()` zerlegen
2. mit `[::-1]` umdrehen
3. mit `join()` verbinden

### Dein Versuch


### Musterlösung


````python
text = "Hallo du schöne Sonne"

woerter = text.split()
umgekehrte_woerter = woerter[::-1]
ergebnis = " ".join(umgekehrte_woerter)

print(ergebnis)
````
