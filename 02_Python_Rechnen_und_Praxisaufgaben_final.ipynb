# Aufgabe 1: Sozialversicherung berechnen

Schreibe ein Programm zur vereinfachten Berechnung der Sozialversicherungsbeiträge.

## Gegeben

- Bruttolohn in Euro
- Arbeitslosenversicherung in Prozent
- Rentenversicherung in Prozent
- Krankenversicherung in Prozent
- Pflegeversicherung in Prozent

Eine spätere Erweiterung könnte zusätzlich die Anzahl der Kinder berücksichtigen.

## Ausgegeben werden sollen

- Bruttolohn
- Beitrag zur Arbeitslosenversicherung
- Beitrag zur Rentenversicherung
- Beitrag zur Krankenversicherung
- Beitrag zur Pflegeversicherung
- Summe aller Sozialversicherungsbeiträge
- verbleibender Betrag nach diesen Abzügen

## Vereinfachung

Beitragsbemessungsgrenzen, Steuern, Kirchensteuer und weitere Regeln einer echten Lohnabrechnung werden hier noch nicht berücksichtigt.

> **Wichtig:** In der ursprünglichen Aufgabenstellung wird vom „zu versteuernden Einkommen“ gesprochen. In dieser vereinfachten Rechnung berechnen wir aber nur den Betrag, der nach den genannten Sozialversicherungsbeiträgen übrig bleibt.


## Planung

1. Bruttolohn und Prozentsätze speichern.
2. Jeden Prozentwert berechnen.
3. Alle Beiträge addieren.
4. Die Summe vom Bruttolohn abziehen.
5. Ergebnisse übersichtlich ausgeben.


## Dein Versuch


## Musterlösung


````python
bruttolohn = 3500.00

arbeitslosenversicherung_prozent = 1.3
rentenversicherung_prozent = 9.3
krankenversicherung_prozent = 7.3
pflegeversicherung_prozent = 1.7

arbeitslosenversicherung = (
    bruttolohn * arbeitslosenversicherung_prozent / 100
)
rentenversicherung = (
    bruttolohn * rentenversicherung_prozent / 100
)
krankenversicherung = (
    bruttolohn * krankenversicherung_prozent / 100
)
pflegeversicherung = (
    bruttolohn * pflegeversicherung_prozent / 100
)

sozialversicherungsbeitraege = (
    arbeitslosenversicherung
    + rentenversicherung
    + krankenversicherung
    + pflegeversicherung
)

verbleibender_betrag = bruttolohn - sozialversicherungsbeitraege

print(f"Bruttolohn: {bruttolohn:.2f} €")
print(f"Arbeitslosenversicherung: {arbeitslosenversicherung:.2f} €")
print(f"Rentenversicherung: {rentenversicherung:.2f} €")
print(f"Krankenversicherung: {krankenversicherung:.2f} €")
print(f"Pflegeversicherung: {pflegeversicherung:.2f} €")
print(f"Beiträge gesamt: {sozialversicherungsbeitraege:.2f} €")
print(f"Verbleibender Betrag: {verbleibender_betrag:.2f} €")
````

## Erklärung

Ein Prozentwert wird so berechnet:

```text
Grundwert × Prozentsatz ÷ 100
```

Beispiel für die Krankenversicherung:

```python
krankenversicherung = bruttolohn * 7.3 / 100
```

Danach werden alle Beiträge zusammengerechnet und vom Bruttolohn abgezogen.


## Beispielausgabe

Bei einem Bruttolohn von `3500,00 €` ergibt sich:

```text
Bruttolohn: 3500.00 €
Arbeitslosenversicherung: 45.50 €
Rentenversicherung: 325.50 €
Krankenversicherung: 255.50 €
Pflegeversicherung: 59.50 €
Beiträge gesamt: 686.00 €
Verbleibender Betrag: 2814.00 €
```


# Aufgabe 2: Warenwert mit Rabatten und Zusatzkosten

Ein Programm soll einen Online-Einkauf berechnen.

## Gegeben

- Warenwert
- Mengenrabatt: `10 %`
- Sonderrabatt: `5 %`
- Versandkosten: `4,90 €`
- Verpackungskosten: `2 %`
- Mehrwertsteuer: `19 %`

## Ausgegeben werden sollen

- ursprünglicher Warenwert
- Mengenrabatt
- Sonderrabatt
- Versandkosten
- Verpackungskosten
- Nettobetrag vor Mehrwertsteuer
- Mehrwertsteuer
- Endpreis

## Rechenreihenfolge

Die Aufgabenstellung sagt, dass Rabatte und Zuschläge **vom ursprünglichen Warenwert** berechnet werden.

Deshalb rechnen wir so:

1. Mengenrabatt vom ursprünglichen Warenwert berechnen.
2. Sonderrabatt vom ursprünglichen Warenwert berechnen.
3. Verpackungskosten vom ursprünglichen Warenwert berechnen.
4. Rabatte abziehen.
5. Versandkosten und Verpackungskosten addieren.
6. Auf den neuen Nettobetrag die Mehrwertsteuer berechnen.
7. Mehrwertsteuer zum Nettobetrag addieren.


## Dein Versuch


## Musterlösung


````python
warenwert = 200.00

mengenrabatt_prozent = 10
sonderrabatt_prozent = 5
versandkosten = 4.90
verpackungskosten_prozent = 2
mehrwertsteuer_prozent = 19

mengenrabatt = warenwert * mengenrabatt_prozent / 100
sonderrabatt = warenwert * sonderrabatt_prozent / 100
verpackungskosten = warenwert * verpackungskosten_prozent / 100

nettobetrag = (
    warenwert
    - mengenrabatt
    - sonderrabatt
    + versandkosten
    + verpackungskosten
)

mehrwertsteuer = nettobetrag * mehrwertsteuer_prozent / 100
endpreis = nettobetrag + mehrwertsteuer

print(f"Warenwert: {warenwert:.2f} €")
print(f"Mengenrabatt: -{mengenrabatt:.2f} €")
print(f"Sonderrabatt: -{sonderrabatt:.2f} €")
print(f"Versandkosten: +{versandkosten:.2f} €")
print(f"Verpackungskosten: +{verpackungskosten:.2f} €")
print(f"Nettobetrag: {nettobetrag:.2f} €")
print(f"Mehrwertsteuer: +{mehrwertsteuer:.2f} €")
print(f"Endpreis: {endpreis:.2f} €")
````

## Erklärung

Alle prozentualen Werte werden vom ursprünglichen Warenwert berechnet:

```python
mengenrabatt = warenwert * 10 / 100
sonderrabatt = warenwert * 5 / 100
verpackungskosten = warenwert * 2 / 100
```

Rabatte werden abgezogen. Versand und Verpackungskosten werden addiert.

Erst danach wird die Mehrwertsteuer auf den neuen Nettobetrag berechnet.


## Beispielausgabe

Bei einem Warenwert von `200,00 €`:

```text
Warenwert: 200.00 €
Mengenrabatt: -20.00 €
Sonderrabatt: -10.00 €
Versandkosten: +4.90 €
Verpackungskosten: +4.00 €
Nettobetrag: 178.90 €
Mehrwertsteuer: +33.99 €
Endpreis: 212.89 €
```

Durch die Ausgabe sieht man gut, welcher Betrag abgezogen und welcher hinzugerechnet wurde.


# Aufgabe 3: Zinseszins berechnen

Schreibe ein Programm, das folgende Werte speichert:

- Startkapital
- Zinssatz
- Anzahl der Jahre

Danach soll das Endkapital mit Zinseszins berechnet werden.

## Formel

```text
Endkapital = Startkapital × (1 + Zinssatz) ** Jahre
```

Der Zinssatz muss vorher von Prozent in eine Dezimalzahl umgerechnet werden.

Beispiel:

```text
3 % = 0,03
```


## Planung

1. Startkapital speichern.
2. Zinssatz in Prozent speichern.
3. Jahre speichern.
4. Zinssatz durch `100` teilen.
5. Endkapital mit der Formel berechnen.
6. Gewinn berechnen.
7. Ergebnisse ausgeben.


## Dein Versuch


## Musterlösung


````python
startkapital = 1000.00
zinssatz_prozent = 3
jahre = 5

zinssatz = zinssatz_prozent / 100

endkapital = startkapital * (1 + zinssatz) ** jahre
gewinn = endkapital - startkapital

print(f"Startkapital: {startkapital:.2f} €")
print(f"Zinssatz: {zinssatz_prozent:.2f} %")
print(f"Laufzeit: {jahre} Jahre")
print(f"Endkapital: {endkapital:.2f} €")
print(f"Gewinn: {gewinn:.2f} €")
````

## Erklärung

Der Ausdruck:

```python
(1 + zinssatz) ** jahre
```

sorgt dafür, dass die Zinsen jedes Jahr erneut mitverzinst werden.

Das ist der Unterschied zwischen einfachen Zinsen und Zinseszins.


## Beispielausgabe

```text
Startkapital: 1000.00 €
Zinssatz: 3.00 %
Laufzeit: 5 Jahre
Endkapital: 1159.27 €
Gewinn: 159.27 €
```
