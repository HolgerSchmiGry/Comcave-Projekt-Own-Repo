# Deutsche IBAN mit Regex prüfen


## Aufgabe

Prüfe nur das Format einer deutschen IBAN.

Erlaubt sind:

- `DE89370400440532013000`
- `DE89 3704 0044 0532 0130 00`

Die Prüfsumme soll nicht berechnet werden.


## Erklärung

Eine deutsche IBAN beginnt mit `DE` und enthält danach 20 Ziffern.

Das Muster erlaubt entweder die Schreibweise ohne Leerzeichen oder die gruppierte Schreibweise.

Die inhaltliche Prüfsumme wird nicht getestet. Dadurch kann eine formal richtige IBAN trotzdem inhaltlich ungültig sein.


## Code


```python
import re

iban_nummern = [
    "DE89370400440532013000",
    "DE89 3704 0044 0532 0130 00",
    "DE89 3704 0044 0532",
    "FR89 3704 0044 0532 0130 00"
]

muster = r"(?:DE[0-9]{20}|DE[0-9]{2}(?: [0-9]{4}){4} [0-9]{2})"

for iban in iban_nummern:
    if re.fullmatch(muster, iban):
        print(f"{iban!r}: Format gültig")
    else:
        print(f"{iban!r}: Format ungültig")
```
