# Deutsche Kennzeichen mit Regex prüfen


## Aufgabe

Prüfe verschiedene Zeichenketten darauf, ob sie wie ein deutsches Kfz-Kennzeichen aufgebaut sind.

Erlaubt sind:

- 1 bis 3 Buchstaben für den Ort,
- ein Leerzeichen,
- 1 bis 2 Buchstaben,
- ein Leerzeichen,
- 1 bis 4 Ziffern,
- optional ein `E` oder `H` am Ende.

Führende oder nachgestellte Leerzeichen und Bindestriche sollen nicht erlaubt sein.


## Erklärung

Mit `re.fullmatch()` wird geprüft, ob die gesamte Zeichenkette dem Muster entspricht.

`[A-Z]{1,3}` steht für ein bis drei Großbuchstaben.

Danach folgen genau ein Leerzeichen, ein oder zwei weitere Buchstaben und erneut ein Leerzeichen.

`[1-9][0-9]{0,3}` erlaubt Zahlen von 1 bis maximal 9999. Eine führende Null ist dadurch ausgeschlossen.

`[EH]?` erlaubt optional ein `E` oder `H`.


## Code


```python
import re

kennzeichen = [
    "WÜ AB 12",
    "B M 1",
    "B MW 1",
    "WES A 1000",
    "WES A 1000E",
    "DU ME 123",
    "DU ME 123H",
    " DU ME 123H ",
    " DU ME 12-3H ",
    " ABC A 123"
]

muster = r"[A-ZÄÖÜ]{1,3} [A-Z]{1,2} [1-9][0-9]{0,3}[EH]?"

for eintrag in kennzeichen:
    if re.fullmatch(muster, eintrag):
        print(f"{eintrag!r}: richtig")
    else:
        print(f"{eintrag!r}: falsch")
```
