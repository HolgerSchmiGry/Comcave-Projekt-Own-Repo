# Passwort analysieren


## Aufgabe

Prüfe, ob das Passwort `MeinGeheimes&§123`:

- mindestens 8 Zeichen lang ist,
- mindestens einen Großbuchstaben enthält,
- mindestens einen Kleinbuchstaben enthält,
- mindestens eine Zahl enthält.


## Erklärung

Mit `len()` wird die Länge geprüft.

`any()` liefert `True`, sobald mindestens ein Zeichen die jeweilige Bedingung erfüllt.

Dafür werden `isupper()`, `islower()` und `isdigit()` verwendet.

Am Ende werden alle Ergebnisse mit `and` verbunden.


## Code


```python
passwort = "MeinGeheimes&§123"

mindestens_acht_zeichen = len(passwort) >= 8
enthaelt_grossbuchstaben = any(zeichen.isupper() for zeichen in passwort)
enthaelt_kleinbuchstaben = any(zeichen.islower() for zeichen in passwort)
enthaelt_zahl = any(zeichen.isdigit() for zeichen in passwort)

print("--- Passwortprüfung ---")
print(f"Passwort:                   {passwort}")
print(f"Mindestens 8 Zeichen:       {mindestens_acht_zeichen}")
print(f"Großbuchstabe enthalten:    {enthaelt_grossbuchstaben}")
print(f"Kleinbuchstabe enthalten:   {enthaelt_kleinbuchstaben}")
print(f"Zahl enthalten:             {enthaelt_zahl}")

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
):
    print("Das Passwort erfüllt alle Bedingungen.")
else:
    print("Das Passwort erfüllt nicht alle Bedingungen.")
```
