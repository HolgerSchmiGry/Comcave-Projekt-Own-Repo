# MAC-Adressen mit Regex prüfen


## Aufgabe

Prüfe, ob eine Zeichenkette wie eine MAC-Adresse aufgebaut ist.

Eine gültige MAC-Adresse besteht aus sechs Zweiergruppen mit hexadezimalen Zeichen.

Die Gruppen werden durch Doppelpunkte getrennt.

Beispiel:

`00:1A:2B:3C:4D:5E`


## Erklärung

`[0-9A-Fa-f]{2}` erlaubt genau zwei hexadezimale Zeichen.

Dieser Teil wird sechsmal benötigt.

Die ersten fünf Gruppen werden jeweils mit einem Doppelpunkt abgeschlossen. Danach folgt die letzte Gruppe ohne weiteren Doppelpunkt.


## Code


```python
import re

mac_adressen = [
    "00:1A:2B:3C:4D:5E",
    "00:1a:2b:3c:4d:5e",
    "00-1A-2B-3C-4D-5E",
    "00:1A:2B:3C:4D",
    "GG:1A:2B:3C:4D:5E"
]

muster = r"(?:[0-9A-Fa-f]{2}:){5}[0-9A-Fa-f]{2}"

for mac in mac_adressen:
    if re.fullmatch(muster, mac):
        print(f"{mac!r}: gültig")
    else:
        print(f"{mac!r}: ungültig")
```
