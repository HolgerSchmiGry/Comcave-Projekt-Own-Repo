# Wörter eines Satzes umkehren


## Aufgabe

Drehe die Reihenfolge der Wörter im Satz:

`Hallo du schöne Sonne`

Das Ergebnis soll sein:

`Sonne schöne du Hallo`


## Erklärung

Mit `split()` wird der Satz in einzelne Wörter zerlegt.

Mit `[::-1]` wird die Liste der Wörter umgedreht.

Mit `" ".join()` werden die Wörter wieder mit Leerzeichen zusammengesetzt.


## Code


```python
text = "Hallo du schöne Sonne"

woerter = text.split()
umgekehrte_woerter = woerter[::-1]
umgekehrter_text = " ".join(umgekehrte_woerter)

print("Original:")
print(text)

print("\nUmgekehrte Reihenfolge:")
print(umgekehrter_text)
```
