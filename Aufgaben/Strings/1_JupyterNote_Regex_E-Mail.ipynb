# E-Mail-Adressen mit Regex prüfen


## Aufgabe

Prüfe mehrere E-Mail-Adressen.

Es dürfen nur Adressen gültig sein, die:

- keine Leerzeichen enthalten,
- nur erlaubte Zeichen vor dem `@` besitzen,
- genau auf `@meinefirma.de` enden.

Die Groß- und Kleinschreibung soll keine Rolle spielen.


## Erklärung

Das Muster wird mit `re.fullmatch()` geprüft. Dadurch muss die komplette Zeichenkette zum regulären Ausdruck passen.

Der erste Teil `[A-Za-z0-9._-]+` erlaubt Buchstaben, Zahlen, Punkte, Unterstriche und Bindestriche vor dem `@`.

Der Punkt in `meinefirma.de` wird als `\.` geschrieben. Ohne den Backslash würde der Punkt in Regex für ein beliebiges Zeichen stehen.

Mit `re.IGNORECASE` werden auch Großbuchstaben akzeptiert.


## Code


```python
import re

emails = [
    "test@meinefirma.de",
    "TEST@meinefirma.de",
    "test123@meinefirmade",
    "v.nachname@meinefirma.de",
    "v-nachname@meinefirma.de",
    "v_nachname@meinefirma.de",
    "nicht gut@meinefirma.de",
    "nicht @meinefirma.de",
    "ok@meinefirma.de ",
    " ok@meinefirma.de ",
    "thomas:eses@meinefirma.de "
]

muster = r"[A-Za-z0-9._-]+@meinefirma\.de"

for email in emails:
    if re.fullmatch(muster, email, re.IGNORECASE):
        print(f"{email!r}: richtig")
    else:
        print(f"{email!r}: falsch")
```
