# Zwei Variablen tauschen


## Aufgabe

Tausche die Inhalte der Variablen `VariableA` und `VariableB`.

Vorher:

- `VariableA = "Hallo"`
- `VariableB = "Welt"`

Nachher sollen die Werte vertauscht sein.


## Erklärung

Python kann zwei Variablen direkt tauschen:

`VariableA, VariableB = VariableB, VariableA`

Die Werte auf der rechten Seite werden zuerst gelesen. Danach werden sie in umgekehrter Reihenfolge zugewiesen.

Eine zusätzliche Hilfsvariable ist nicht nötig.


## Code


```python
VariableA = "Hallo"
VariableB = "Welt"

print("--- Vor dem Tauschen ---")
print(VariableA, VariableB)

VariableA, VariableB = VariableB, VariableA

print("\n--- Nach dem Tauschen ---")
print(VariableA, VariableB)
```
