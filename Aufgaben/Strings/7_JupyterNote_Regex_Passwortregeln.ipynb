# Passwortregeln mit Regex prüfen


## Aufgabe

Prüfe ein Passwort auf folgende Regeln:

- mindestens 8 Zeichen,
- mindestens ein Großbuchstabe,
- mindestens ein Kleinbuchstabe,
- mindestens eine Zahl,
- mindestens ein Sonderzeichen.


## Erklärung

Für jede Regel wird ein eigener regulärer Ausdruck verwendet.

Mit `re.search()` wird geprüft, ob im Passwort mindestens ein passendes Zeichen vorkommt.

`[^A-Za-z0-9]` erkennt ein Zeichen, das weder Buchstabe noch Zahl ist. Damit wird das Sonderzeichen geprüft.

Nur wenn alle Bedingungen wahr sind, gilt das Passwort als gültig.


## Code


```python
import re

passwort = input("Bitte ein Passwort eingeben: ")

mindestens_acht_zeichen = bool(re.search(r".{8,}", passwort))
enthaelt_grossbuchstaben = bool(re.search(r"[A-Z]", passwort))
enthaelt_kleinbuchstaben = bool(re.search(r"[a-z]", passwort))
enthaelt_zahl = bool(re.search(r"[0-9]", passwort))
enthaelt_sonderzeichen = bool(re.search(r"[^A-Za-z0-9]", passwort))

print("\n--- Passwortprüfung ---")
print(f"Mindestens 8 Zeichen:       {mindestens_acht_zeichen}")
print(f"Großbuchstabe enthalten:    {enthaelt_grossbuchstaben}")
print(f"Kleinbuchstabe enthalten:   {enthaelt_kleinbuchstaben}")
print(f"Zahl enthalten:             {enthaelt_zahl}")
print(f"Sonderzeichen enthalten:    {enthaelt_sonderzeichen}")

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
    and enthaelt_sonderzeichen
):
    print("Das Passwort erfüllt alle Regeln.")
else:
    print("Das Passwort erfüllt nicht alle Regeln.")
```
