# Euro-Beträge mit Regex prüfen


## Aufgabe

Prüfe verschiedene Texte darauf, ob sie einen gültigen Euro-Betrag darstellen.

Erlaubt sind zum Beispiel:

- `10.000,00 €`
- `10.000 €`
- `10000 €`
- `10000,00 €`
- `0,90 €`

Das Eurozeichen ist Pflicht. Vor oder nach dem Betrag dürfen keine zusätzlichen Leerzeichen stehen.


## Erklärung

Der Ausdruck erlaubt entweder eine normale Zahl ohne Tausenderpunkte oder eine korrekt gruppierte Zahl mit Tausenderpunkten.

`[0-9]+` erlaubt normale Zahlen wie `10000`.

`[0-9]{1,3}(?:\.[0-9]{3})+` erlaubt gruppierte Zahlen wie `10.000` oder `1.000.000`.

`(?:,[0-9]{2})?` erlaubt optional genau zwei Nachkommastellen.

Am Ende müssen ein Leerzeichen und das Eurozeichen folgen.


## Code


```python
import re

zahlen = [
    "10.000,00 €",
    "10.000 €",
    "10000 €",
    "10000,00 €",
    "10.000",
    "10000",
    "0,90 €",
    " 1000",
    "222 2",
    "0900"
]

muster = r"(?:[0-9]+|[0-9]{1,3}(?:\.[0-9]{3})+)(?:,[0-9]{2})? €"

for zahl in zahlen:
    if re.fullmatch(muster, zahl):
        print(f"{zahl!r}: gültig")
    else:
        print(f"{zahl!r}: ungültig")
```
