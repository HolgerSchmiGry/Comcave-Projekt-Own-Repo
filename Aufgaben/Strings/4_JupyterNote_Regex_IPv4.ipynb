# IPv4-Adressen mit Regex prüfen


## Aufgabe

Prüfe, ob eine Zeichenkette wie eine IPv4-Adresse aufgebaut ist.

Eine IPv4-Adresse besteht aus vier Zahlenblöcken, die durch Punkte getrennt sind.

Jeder Zahlenblock darf nur einen Wert von 0 bis 255 enthalten.


## Erklärung

Für einen Zahlenblock von 0 bis 255 werden mehrere Möglichkeiten kombiniert:

- `25[0-5]` erlaubt 250 bis 255.
- `2[0-4][0-9]` erlaubt 200 bis 249.
- `1[0-9]{2}` erlaubt 100 bis 199.
- `[1-9]?[0-9]` erlaubt 0 bis 99.

Dieser Block wird viermal verwendet und jeweils mit einem Punkt getrennt.


## Code


```python
import re

ip_adressen = [
    "192.168.1.100",
    "127.0.0.1",
    "255.255.255.255",
    "256.168.1.1",
    "192.168.1",
    "192.168.01.1"
]

block = r"(?:25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])"
muster = rf"{block}\.{block}\.{block}\.{block}"

for ip in ip_adressen:
    if re.fullmatch(muster, ip):
        print(f"{ip!r}: gültig")
    else:
        print(f"{ip!r}: ungültig")
```
