# Aufgabe 1: Einlasskontrolle einer Diskothek

## Aufgabenstellung

Vor der Diskothek steht ein Roboter. Seine Sensoren erkennen das Alter einer Person, den VIP-Status, die Farbe des T-Shirts und die aktuelle Uhrzeit. Als Ausgabe besitzt der Roboter nur ein Laufband, auf dem eine kurze Meldung angezeigt wird.

Dabei gelten folgende Regeln:

- Menschen unter 18 Jahren dürfen sich nur bis 00:00 Uhr in der Diskothek aufhalten.
- VIP-Gäste bekommen einen Sekt.
- Menschen mit einem gelben T-Shirt bekommen ebenfalls einen Sekt.
- Alle anderen Personen kommen ganz normal hinein.
- Menschen unter 18 Jahren dürfen keinen Alkohol bekommen.
- Jede Person bekommt höchstens einen Sekt.

Die Eingabe soll über feste Variablen erfolgen.  
Die Ausgabe soll mit einem einfachen `print()` ausgegeben werden.


## Planung der Variablen

Für die Aufgabe werden folgende Eingangsparameter benötigt:

| Variable | Bedeutung | Beispiel |
|---|---|---:|
| `alter` | Alter der Person | `20` |
| `uhrzeit` | Uhrzeit als Zahl; für diese Übung darf `24` für 00:00 Uhr verwendet werden | `22` |
| `ist_vip` | Gibt an, ob die Person VIP ist | `False` |
| `gelbes_tshirt` | Gibt an, ob die Person ein gelbes T-Shirt trägt | `True` |

> **Hinweis:** Für diese vereinfachte Übung wird Mitternacht als `24` dargestellt. In einem echten Programm würde man dafür mit Datum und Uhrzeit arbeiten.


````python
def einlass_pruefen(alter, uhrzeit, ist_vip, gelbes_tshirt):
    """Prüft den Einlass und die Sekt-Ausgabe für eine Person."""

    if alter < 18:
        if uhrzeit >= 24:
            print("Kein Einlass: Personen unter 18 dürfen nur bis 00:00 Uhr bleiben.")
        else:
            print("Einlass erlaubt, aber kein Alkohol.")
    elif ist_vip or gelbes_tshirt:
        print("Einlass erlaubt. Die Person bekommt einen Sekt.")
    else:
        print("Einlass erlaubt. Die Person kommt ganz normal hinein.")


def main():
    # Beispielaufruf
    einlass_pruefen(20, 22, False, True)


main()
````

## Erklärung der Lösung

Die gesamte Prüfung befindet sich in der Funktion:

```python
def einlass_pruefen(alter, uhrzeit, ist_vip, gelbes_tshirt):
```

Die vier Werte werden beim Funktionsaufruf als Argumente übergeben.

Zuerst wird geprüft, ob die Person unter 18 Jahre alt ist:

```python
if alter < 18:
```

Für Minderjährige gelten besondere Regeln. Nach Mitternacht ist kein Aufenthalt mehr erlaubt. Vor Mitternacht dürfen sie hinein, bekommen aber keinen Alkohol.

Erst danach wird bei volljährigen Personen geprüft, ob mindestens eine der beiden Sekt-Bedingungen erfüllt ist:

```python
elif ist_vip or gelbes_tshirt:
```

Der Operator `or` bedeutet **oder**. Die Bedingung ist wahr, wenn die Person VIP ist, ein gelbes T-Shirt trägt oder beides zutrifft.

Da beide Bedingungen gemeinsam geprüft werden, wird trotzdem nur einmal ausgegeben, dass die Person einen Sekt bekommt.

Der Beispielaufruf wird in einer eigenen `main()`-Funktion ausgeführt:

```python
def main():
    einlass_pruefen(20, 22, False, True)

main()
```

Die Namen der Funktionen können frei gewählt werden. Wichtig ist nur, dass der verwendete Name bei Definition und Aufruf übereinstimmt.

# Aufgabe 2: Statusanzeige eines Parkhauses

## Aufgabenstellung

Es soll ein Programm für die Statusanzeige eines Parkhauses entwickelt werden.

Das Parkhaus besitzt insgesamt **500 Stellplätze**. Die Anzahl der belegten Stellplätze wird automatisch ermittelt.

Für die Statusanzeige gelten folgende Bedingungen:

- Die Auslastung kann über 95 Prozent liegen.
- Die Ein- oder Ausfahrt kann blockiert sein.
- Das Parkhaus kann für eine Veranstaltung angemietet sein.

Als Ausgabe soll genau einer der folgenden Texte erscheinen:

- `Parkhaus belegt`
- `Parkhaus frei`
- `Parkhaus gesperrt`


## Entscheidungstabelle

`J` bedeutet **Ja**, `N` bedeutet **Nein** und `X` kennzeichnet die auszuführende Aktion.

| Bedingungen und Aktionen | R1 | R2 | R3 | R4 | R5 | R6 | R7 | R8 |
|---|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
| Auslastung über 95 % | J | J | J | J | N | N | N | N |
| Ein- oder Ausfahrt blockiert | J | J | N | N | J | J | N | N |
| Für Veranstaltung angemietet | J | N | J | N | J | N | J | N |
| **Ausgabe „Parkhaus belegt“** |  |  |  | X |  |  |  |  |
| **Ausgabe „Parkhaus frei“** |  |  |  |  |  |  |  | X |
| **Ausgabe „Parkhaus gesperrt“** | X | X | X |  | X | X | X |  |

## Auswertung der Tabelle

Aus der Entscheidungstabelle lässt sich erkennen:

1. Sobald die Ein- oder Ausfahrt blockiert ist, wird das Parkhaus gesperrt.
2. Sobald das Parkhaus für eine Veranstaltung angemietet ist, wird es ebenfalls gesperrt.
3. Nur wenn keine Sperrung vorliegt, wird die Auslastung geprüft.
4. Bei mehr als 95 Prozent Auslastung lautet der Status `Parkhaus belegt`.
5. In allen übrigen Fällen lautet der Status `Parkhaus frei`.


## Planung der Variablen

| Variable | Bedeutung | Beispiel |
|---|---|---:|
| `anzahl_stellplaetze` | Gesamtzahl aller Stellplätze | `500` |
| `belegte_stellplaetze` | Anzahl der aktuell belegten Stellplätze | `480` |
| `zufahrt_blockiert` | Gibt an, ob Ein- oder Ausfahrt blockiert ist | `False` |
| `fuer_veranstaltung_gemietet` | Gibt an, ob das Parkhaus vermietet ist | `False` |

Die Auslastung wird mit folgender Formel berechnet:

```text
Auslastung = belegte Stellplätze / gesamte Stellplätze × 100
```


````python
def status_anzeigen(
    belegte_stellplaetze,
    zufahrt_blockiert,
    fuer_veranstaltung_gemietet,
    anzahl_stellplaetze=500
):
    """Ermittelt den aktuellen Status des Parkhauses."""

    auslastung = belegte_stellplaetze / anzahl_stellplaetze * 100

    if zufahrt_blockiert or fuer_veranstaltung_gemietet:
        print("Parkhaus gesperrt")
    elif auslastung > 95:
        print("Parkhaus belegt")
    else:
        print("Parkhaus frei")


def main():
    # Beispielaufruf
    status_anzeigen(330, True, True)


main()
````

## Erklärung der Lösung

Die Parkhauslogik befindet sich in der Funktion:

```python
def status_anzeigen(
    belegte_stellplaetze,
    zufahrt_blockiert,
    fuer_veranstaltung_gemietet,
    anzahl_stellplaetze=500
):
```

Die Gesamtzahl der Stellplätze besitzt mit `500` bereits einen Standardwert. Dadurch reicht beim Aufruf zum Beispiel:

```python
status_anzeigen(330, True, True)
```

Zuerst wird die Auslastung berechnet:

```python
auslastung = belegte_stellplaetze / anzahl_stellplaetze * 100
```

Danach wird zuerst geprüft, ob das Parkhaus gesperrt werden muss:

```python
if zufahrt_blockiert or fuer_veranstaltung_gemietet:
```

Diese Prüfung steht an erster Stelle, weil eine Sperrung Vorrang vor der Auslastung hat.

Nur wenn keine Sperrung vorliegt, wird geprüft, ob die Auslastung über 95 Prozent liegt:

```python
elif auslastung > 95:
```

Der Funktionsaufruf wird anschließend in `main()` geschrieben:

```python
def main():
    status_anzeigen(330, True, True)

main()
```

Auch hier sind die Funktionsnamen frei wählbar. `status_anzeigen()` und `main()` sind nur sinnvolle und gut lesbare Namen.

# Zusammenfassung

Bei beiden Aufgaben ist die Reihenfolge der Bedingungen besonders wichtig.

## Diskothek

1. Minderjährigkeit und Uhrzeit prüfen.
2. Minderjährigen keinen Alkohol geben.
3. Bei Volljährigen VIP-Status oder gelbes T-Shirt prüfen.
4. Durch eine gemeinsame `or`-Bedingung maximal einen Sekt ausgeben.

## Parkhaus

1. Zuerst mögliche Sperrgründe prüfen.
2. Danach die Auslastung prüfen.
3. Wenn keine Bedingung zutrifft, ist das Parkhaus frei.

## Merksatz

> `if` prüft die erste Bedingung, `elif` weitere Bedingungen und `else` behandelt alle übrigen Fälle.
