## 1. Was ist eine Variable?

Eine Variable speichert einen Wert unter einem Namen.

Beispiel:

```python
alter = 49
```

- `alter` ist der Variablenname.
- `49` ist der gespeicherte Wert.
- Mit `=` wird der Wert der Variablen zugewiesen.

Das Gleichheitszeichen bedeutet hier nicht „ist gleich“, sondern „weise den Wert auf der rechten Seite der Variablen auf der linken Seite zu“.


````python
alter = 49
name = "Max"

print(alter)
print(name)
````

## 2. Einfache Datentypen

Python kennt verschiedene Datentypen.

Wichtige einfache Datentypen sind:

- `int` – ganze Zahlen
- `float` – Kommazahlen
- `str` – Zeichenketten
- `bool` – Wahrheitswerte
- `None` – kein Wert vorhanden

Mit `type()` kann der Datentyp eines Wertes angezeigt werden.


````python
test_number = 5
test_decimal = 3.5
test_string = "123"
test_bool = True
test_none = None

print("test_number:", test_number, type(test_number))
print("test_decimal:", test_decimal, type(test_decimal))
print("test_string:", test_string, type(test_string))
print("test_bool:", test_bool, type(test_bool))
print("test_none:", test_none, type(test_none))
````

## 3. Python ist dynamisch typisiert

Die Aussage „Python ist nicht typisiert“ ist nicht ganz richtig.

Python ist **dynamisch typisiert**.

Das bedeutet:

- Jede Variable besitzt zu einem bestimmten Zeitpunkt einen Datentyp.
- Der Datentyp wird automatisch aus dem gespeicherten Wert erkannt.
- Eine Variable kann später einen Wert mit einem anderen Datentyp erhalten.

Python ist außerdem **stark typisiert**. Unterschiedliche Datentypen können nicht ohne Weiteres miteinander verrechnet werden.


````python
wert = 5
print(wert, type(wert))

wert = "Hallo"
print(wert, type(wert))
````

## 4. Typannotationen

Eine Typannotation zeigt, welcher Datentyp für eine Variable vorgesehen ist.

Beispiel:

```python
alter: int = 49
```

Die Typannotation verbessert die Lesbarkeit und hilft Entwicklungsumgebungen bei der Prüfung.

Python erzwingt den Datentyp dadurch aber nicht automatisch.


````python
test_number: int = 5
test_decimal: float = 3.5
test_string: str = "123"
test_bool: bool = True
test_none: None = None

print(test_number)
print(test_decimal)
print(test_string)
print(test_bool)
print(test_none)
````

## 5. Benennung von Variablen

Ein Variablenname sollte verständlich sein.

In Python wird meistens `snake_case` verwendet:

```python
menge_der_dateien = 20
```

Regeln:

- Der Name darf nicht mit einer Zahl beginnen.
- Leerzeichen sind nicht erlaubt.
- Buchstaben, Zahlen und Unterstriche sind erlaubt.
- Python unterscheidet Groß- und Kleinschreibung.
- Abkürzungen sollten sparsam verwendet werden.

Für Variablen ist `snake_case` in Python üblicher als CamelCase oder PascalCase.


````python
menge_der_dateien = 20
du_bist_ein_wert = "Hallo"
zahl_22 = 100

print(menge_der_dateien)
print(du_bist_ein_wert)
print(zahl_22)
````

## 6. Python ist case-sensitive

Python unterscheidet Groß- und Kleinschreibung.

Diese beiden Variablen sind deshalb verschieden:

```python
variable_a
Variable_A
```


````python
Variable_a = "Hallo"
variable_A = "Welt"

print(Variable_a)
print(variable_A)
````

## 7. Typumwandlung

Mit Typumwandlung wird ein Wert in einen anderen Datentyp umgewandelt.

Wichtige Funktionen:

- `int()`
- `float()`
- `str()`
- `bool()`

Nicht jede Umwandlung ist möglich.


````python
print(int(5.5))
print(float("5.5"))
print(str(5.5))
print(int(True))
print(bool(0))
print(bool(1))
````

### Zahl und String können nicht direkt addiert werden

Das hier erzeugt einen Fehler:

```python
10 + "Hallo"
```

Der Grund:

- `10` ist ein `int`.
- `"Hallo"` ist ein `str`.

Beide Werte müssen zuerst in einen passenden gemeinsamen Datentyp gebracht werden.


````python
zahl = 10
text = "Hallo"

print(str(zahl) + " " + text)
````

## 8. Arithmetische Operatoren

Python unterstützt die üblichen Rechenoperatoren:

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `//` Ganzzahldivision
- `%` Rest einer Division
- `**` Potenz


````python
zahl_1 = 10
zahl_2 = 5

print("Addition:", zahl_1 + zahl_2)
print("Subtraktion:", zahl_1 - zahl_2)
print("Multiplikation:", zahl_1 * zahl_2)
print("Potenz:", zahl_1 ** zahl_2)

division = zahl_1 / zahl_2
print("Division:", division)
print("Datentyp der Division:", type(division))

modulo = 10 % 3
print("Modulo:", modulo)

ganzzahldivision = 10 // 3
print("Ganzzahldivision:", ganzzahldivision)
print("Datentyp:", type(ganzzahldivision))
````

## 9. Division, Ganzzahldivision und Modulo

### Normale Division `/`

Das Ergebnis ist normalerweise ein `float`.

```python
10 / 3
```

Ergebnis:

```text
3.3333333333333335
```

### Ganzzahldivision `//`

Die Nachkommastellen werden abgeschnitten.

```python
10 // 3
```

Ergebnis:

```text
3
```

### Modulo `%`

Modulo liefert den Rest einer Division.

```python
10 % 3
```

Ergebnis:

```text
1
```


````python
print(10 / 3)
print(10 // 3)
print(10 % 3)
````

## 10. Wahrheitswerte beim Rechnen

In Python verhält sich:

- `True` wie die Zahl `1`
- `False` wie die Zahl `0`

Deshalb ist folgende Rechnung möglich:


````python
bool_1 = True
bool_2 = True

print(bool_1 + bool_2)
print(True + False)
print(False + False)
````

Das ist technisch erlaubt, sollte aber nur verwendet werden, wenn der Zusammenhang verständlich bleibt.


## 11. Genauigkeit bei Kommazahlen

Kommazahlen werden im Computer binär gespeichert.

Einige Dezimalzahlen können dabei nicht exakt dargestellt werden. Deshalb entstehen manchmal kleine Rundungsabweichungen.


````python
wert = 0

wert += 0.12
wert += 2.1
wert += 3.1
wert += 3.1
wert += 3.1

for _ in range(11):
    wert += 0.1

print("Ungefilterter Wert:", wert)
print("Gerundet:", round(wert, 2))
````

### Vorsicht bei Multiplikation und Division

Auch Rechnungen, die mathematisch exakt aussehen, können intern kleine Abweichungen erzeugen.


````python
print(49.0 * (1.0 / 49.0))
print((1 / 49) * 49)
````

Für normale Ausgaben reicht oft:

```python
round(wert, 2)
```

Bei Geldbeträgen und besonders genauen Berechnungen kann später `Decimal` sinnvoll sein.


## 12. Aufgabe: Zwei Variablen tauschen

Gegeben sind:

```python
variable_a = "Hallo"
variable_b = "Welt"
```

Tausche die Inhalte der beiden Variablen.

Vorher:

```text
Hallo Welt
```

Nachher:

```text
Welt Hallo
```


````python
variable_a = "Hallo"
variable_b = "Welt"

print("Vorher:", variable_a, variable_b)

variable_a, variable_b = variable_b, variable_a

print("Nachher:", variable_a, variable_b)
````
