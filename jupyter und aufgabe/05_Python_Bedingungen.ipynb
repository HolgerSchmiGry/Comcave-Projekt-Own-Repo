# Python – Bedingungen und bedingte Ausdrücke

Dieses Notebook behandelt die kurze Schreibweise für einfache `if-else`-Entscheidungen.


## Normaler `if-else`-Block


````python
alter = 17

if alter >= 18:
    status = "Erwachsen"
else:
    status = "Minderjährig"

print(status)
````

## Bedingter Ausdruck

Kurze Schreibweise:

```python
wert_wenn_wahr if bedingung else wert_wenn_falsch
```

Diese Schreibweise eignet sich für kurze und übersichtliche Entscheidungen.


````python
alter = 17

status = "Erwachsen" if alter >= 18 else "Minderjährig"

print(status)
````

## Bedingung in einem f-String

Auch innerhalb eines f-Strings kann ein bedingter Ausdruck verwendet werden.


````python
alter = 20

print(f"Status: {'Volljährig' if alter >= 18 else 'Minderjährig'}")
````
