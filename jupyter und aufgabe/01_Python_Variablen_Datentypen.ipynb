## Variablen und Datentypen

Eine Variable speichert einen Wert unter einem Namen.

Python ist **dynamisch typisiert**. Der Datentyp wird also anhand des gespeicherten Wertes erkannt.

Wichtige Datentypen:

- `int` – ganze Zahl
- `float` – Kommazahl
- `str` – Text
- `bool` – `True` oder `False`
- `None` – kein Wert vorhanden

Mit `type()` kann der Datentyp angezeigt werden.


````python
test_number = 5
test_decimal = 3.5
test_string = "123"
test_bool = True
test_none = None

print(test_number, type(test_number))
print(test_decimal, type(test_decimal))
print(test_string, type(test_string))
print(test_bool, type(test_bool))
print(test_none, type(test_none))
````

## Typannotationen

Mit einer Typannotation wird angegeben, welcher Datentyp vorgesehen ist.

Python erzwingt den Datentyp dadurch nicht automatisch, aber der Code wird verständlicher.


````python
zahl: int = 5
kommazahl: float = 3.5
text: str = "Hallo"
ist_aktiv: bool = True

print(zahl, kommazahl, text, ist_aktiv)
````

## Variablennamen

In Python wird meistens `snake_case` verwendet.

Beispiele:

```python
anzahl_dateien = 20
vorname = "Max"
ist_angemeldet = True
```

Wichtig:

- keine Zahl am Anfang
- keine Leerzeichen
- Groß- und Kleinschreibung wird unterschieden
- verständliche Namen benutzen


````python
variable_a = "Hallo"
VariableA = "Welt"

print(variable_a)
print(VariableA)
````

## Typumwandlung

Mit diesen Funktionen können Werte umgewandelt werden:

- `int()`
- `float()`
- `str()`
- `bool()`

Nicht jede Umwandlung ist möglich. `"Hallo"` kann zum Beispiel nicht in eine Zahl umgewandelt werden.


````python
print(int(5.5))
print(float("5.5"))
print(str(5.5))
print(int(True))
print(bool(0))
print(bool(1))
````

## Aufgabe: Zwei Variablen tauschen

Die Variablen enthalten zuerst:

```python
variable_a = "Hallo"
variable_b = "Welt"
```

Tausche die Inhalte und gib die Werte vorher und nachher aus.

Python kann zwei Variablen direkt in einer Zeile tauschen.


````python
variable_a = "Hallo"
variable_b = "Welt"

print("Vorher:", variable_a, variable_b)

variable_a, variable_b = variable_b, variable_a

print("Nachher:", variable_a, variable_b)
````
