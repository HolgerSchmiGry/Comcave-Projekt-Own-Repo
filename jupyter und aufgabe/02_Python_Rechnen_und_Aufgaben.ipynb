## Arithmetische Operatoren

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `//` Ganzzahldivision
- `%` Rest einer Division
- `**` Potenz


````python
zahl_1 = 10
zahl_2 = 5

print("Addition:", zahl_1 + zahl_2)
print("Subtraktion:", zahl_1 - zahl_2)
print("Multiplikation:", zahl_1 * zahl_2)
print("Division:", zahl_1 / zahl_2)
print("Ganzzahldivision:", 10 // 3)
print("Modulo:", 10 % 3)
print("Potenz:", zahl_1 ** zahl_2)
````

## Genauigkeit bei Kommazahlen

Kommazahlen werden intern binär gespeichert. Dadurch können kleine Rundungsabweichungen entstehen.

Mit `round(wert, 2)` wird auf zwei Nachkommastellen gerundet.


````python
wert = 0

wert += 0.12
wert += 2.1
wert += 3.1
wert += 3.1
wert += 3.1

for _ in range(11):
    wert += 0.1

print("Ungefiltert:", wert)
print("Gerundet:", round(wert, 2))

print(49.0 * (1.0 / 49.0))
print((1 / 49) * 49)
````

## Aufgabe: Sozialversicherung berechnen

Der Bruttolohn wird selbst eingegeben. Die Beitragssätze sind fest hinterlegt.

Berechnet werden:

- Arbeitslosenversicherung
- Rentenversicherung
- Krankenversicherung
- Pflegeversicherung
- Summe der Beiträge
- vereinfachtes Einkommen nach den Abzügen

Die echte Lohnabrechnung ist komplizierter. Hier wird nur die Aufgabenstellung umgesetzt.


````python
bruttolohn = float(input("Bitte den Bruttolohn in Euro eingeben: "))

arbeitslosenversicherung_prozent = 1.3
rentenversicherung_prozent = 9.3
krankenversicherung_prozent = 8.15
pflegeversicherung_prozent = 1.8

arbeitslosenversicherung = bruttolohn * arbeitslosenversicherung_prozent / 100
rentenversicherung = bruttolohn * rentenversicherung_prozent / 100
krankenversicherung = bruttolohn * krankenversicherung_prozent / 100
pflegeversicherung = bruttolohn * pflegeversicherung_prozent / 100

sozialversicherung_gesamt = (
    arbeitslosenversicherung
    + rentenversicherung
    + krankenversicherung
    + pflegeversicherung
)

einkommen_nach_abzuegen = bruttolohn - sozialversicherung_gesamt

print("\n--- Ergebnis ---")
print(f"Bruttolohn:                     {bruttolohn:10.2f} Euro")
print(f"Arbeitslosenversicherung:       {arbeitslosenversicherung:10.2f} Euro")
print(f"Rentenversicherung:             {rentenversicherung:10.2f} Euro")
print(f"Krankenversicherung:            {krankenversicherung:10.2f} Euro")
print(f"Pflegeversicherung:             {pflegeversicherung:10.2f} Euro")
print(f"Sozialversicherung insgesamt:   {sozialversicherung_gesamt:10.2f} Euro")
print(f"Einkommen nach Abzügen:         {einkommen_nach_abzuegen:10.2f} Euro")
````

## Aufgabe: Rabatte und Zusatzkosten

Berechnet werden:

- Mengenrabatt: 10 %
- Sonderrabatt: 5 %
- Versandkosten: 4,90 Euro
- Verpackungskosten: 2 %
- Mehrwertsteuer: 19 %
- Endpreis

Die Verpackungskosten werden hier auf den Betrag nach den Rabatten berechnet.


````python
warenwert = float(input("Bitte den Warenwert in Euro eingeben: "))

mengenrabatt = warenwert * 10 / 100
sonderrabatt = warenwert * 5 / 100

betrag_nach_rabatten = warenwert - mengenrabatt - sonderrabatt

versandkosten = 4.90
verpackungskosten = betrag_nach_rabatten * 2 / 100

nettobetrag = betrag_nach_rabatten + versandkosten + verpackungskosten
mehrwertsteuer = nettobetrag * 19 / 100
endpreis = nettobetrag + mehrwertsteuer

print("\n--- Online-Einkauf ---")
print(f"Warenwert:             {warenwert:10.2f} Euro")
print(f"Mengenrabatt:         -{mengenrabatt:10.2f} Euro")
print(f"Sonderrabatt:         -{sonderrabatt:10.2f} Euro")
print(f"Versandkosten:        +{versandkosten:10.2f} Euro")
print(f"Verpackungskosten:    +{verpackungskosten:10.2f} Euro")
print(f"Mehrwertsteuer:       +{mehrwertsteuer:10.2f} Euro")
print("--------------------------------")
print(f"Endpreis:              {endpreis:10.2f} Euro")
````

## Aufgabe: Zinseszinsen

Formel:

```text
Endkapital = Startkapital * (1 + Zinssatz / 100) ** Jahre
```

`zinssatz / 100` wandelt den Prozentwert in eine Dezimalzahl um.

Mit `**` wird potenziert. Dadurch werden auch die Zinsen der vorherigen Jahre wieder verzinst.


````python
startkapital = float(input("Bitte das Startkapital in Euro eingeben: "))
zinssatz = float(input("Bitte den Zinssatz in Prozent eingeben: "))
jahre = int(input("Bitte die Anzahl der Jahre eingeben: "))

endkapital = startkapital * (1 + zinssatz / 100) ** jahre

print("\n--- Zinsberechnung ---")
print(f"Startkapital: {startkapital:.2f} Euro")
print(f"Zinssatz:     {zinssatz:.2f} Prozent")
print(f"Laufzeit:     {jahre} Jahre")
print(f"Endkapital:   {endkapital:.2f} Euro")
````
