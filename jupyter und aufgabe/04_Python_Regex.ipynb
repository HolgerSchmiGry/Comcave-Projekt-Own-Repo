## Regex-Grundlagen

Regex wird verwendet, um Texte nach Mustern zu durchsuchen oder vollständig zu prüfen.

Wichtige Funktionen:

- `re.findall()` – alle Treffer finden
- `re.search()` – einen Treffer suchen
- `re.fullmatch()` – die gesamte Zeichenkette prüfen
- `re.sub()` – Treffer ersetzen

Wichtige Bestandteile:

- `[0-9]` oder `\d` – Ziffer
- `[A-Z]` – Großbuchstabe
- `[a-z]` – Kleinbuchstabe
- `+` – mindestens einmal
- `?` – optional
- `{2}` – genau zweimal
- `{1,3}` – ein- bis dreimal
- `\.` – echter Punkt
- `\s` – Leerraum


````python
import re

text = "Artikel A123 kostet 49 Euro und Artikel B456 kostet 99 €."

print(re.findall(r"[0-9]+", text))
print(re.findall(r"[A-Z][0-9]+", text))

treffer = re.findall(r"(\d+)\s+(Euro|euro|€|EUR)", text)
preise = [eintrag[0] for eintrag in treffer]

print(preise)
````

## Aufgabe 1: E-Mail-Adressen

Nur Adressen mit `@meinefirma.de` sind erlaubt.

Vor dem `@` sind Buchstaben, Zahlen, Punkte, Bindestriche und Unterstriche erlaubt.

Die komplette Zeichenkette wird mit `re.fullmatch()` geprüft.


````python
import re

emails = [
    "test@meinefirma.de",
    "TEST@meinefirma.de",
    "test123@meinefirmade",
    "v.nachname@meinefirma.de",
    "v-nachname@meinefirma.de",
    "v_nachname@meinefirma.de",
    "nicht gut@meinefirma.de",
    "nicht @meinefirma.de",
    "ok@meinefirma.de ",
    " ok@meinefirma.de ",
    "thomas:eses@meinefirma.de "
]

muster = r"[A-Za-z0-9._-]+@meinefirma\.de"

for email in emails:
    print(f"{email!r}: {bool(re.fullmatch(muster, email, re.IGNORECASE))}")
````

## Aufgabe 2: Kfz-Kennzeichen

Vereinfachtes Format:

- 1 bis 3 Buchstaben
- Leerzeichen
- 1 bis 2 Buchstaben
- Leerzeichen
- 1 bis 4 Ziffern
- optional `E` oder `H`

Es wird nur das Format geprüft, nicht ob das Ortskürzel existiert.


````python
import re

kennzeichen = [
    "WÜ AB 12",
    "B M 1",
    "B MW 1",
    "WES A 1000",
    "WES A 1000E",
    "DU ME 123",
    "DU ME 123H",
    " DU ME 123H ",
    " DU ME 12-3H ",
    " ABC A 123"
]

muster = r"[A-ZÄÖÜ]{1,3} [A-Z]{1,2} [1-9][0-9]{0,3}[EH]?"

for eintrag in kennzeichen:
    print(f"{eintrag!r}: {bool(re.fullmatch(muster, eintrag))}")
````

## Aufgabe 3: Euro-Beträge

Erlaubte Beispiele:

- `10.000,00 €`
- `10.000 €`
- `10000 €`
- `10000,00 €`
- `0,90 €`

Das Eurozeichen ist Pflicht.


````python
import re

zahlen = [
    "10.000,00 €",
    "10.000 €",
    "10000 €",
    "10000,00 €",
    "10.000",
    "10000",
    "0,90 €",
    " 1000",
    "222 2",
    "0900"
]

muster = r"(?:[0-9]+|[0-9]{1,3}(?:\.[0-9]{3})+)(?:,[0-9]{2})? €"

for zahl in zahlen:
    print(f"{zahl!r}: {bool(re.fullmatch(muster, zahl))}")
````

## Aufgabe 4: IPv4-Adresse

Eine IPv4-Adresse besteht aus vier Zahlenblöcken.

Jeder Block muss zwischen 0 und 255 liegen.


````python
import re

ip_adressen = [
    "192.168.1.100",
    "127.0.0.1",
    "255.255.255.255",
    "256.168.1.1",
    "192.168.1",
    "192.168.01.1"
]

block = r"(?:25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])"
muster = rf"{block}\.{block}\.{block}\.{block}"

for ip in ip_adressen:
    print(f"{ip!r}: {bool(re.fullmatch(muster, ip))}")
````

## Aufgabe 5: Deutsche IBAN

Geprüft wird nur das Format.

Erlaubt sind die Schreibweisen:

- ohne Leerzeichen
- mit Vierergruppen

Die Prüfsumme wird nicht geprüft.


````python
import re

iban_nummern = [
    "DE89370400440532013000",
    "DE89 3704 0044 0532 0130 00",
    "DE89 3704 0044 0532",
    "FR89 3704 0044 0532 0130 00"
]

muster = r"(?:DE[0-9]{20}|DE[0-9]{2}(?: [0-9]{4}){4} [0-9]{2})"

for iban in iban_nummern:
    print(f"{iban!r}: {bool(re.fullmatch(muster, iban))}")
````

## Aufgabe 6: MAC-Adresse

Eine MAC-Adresse besteht aus sechs Gruppen mit jeweils zwei hexadezimalen Zeichen.

Beispiel:

```text
00:1A:2B:3C:4D:5E
```


````python
import re

mac_adressen = [
    "00:1A:2B:3C:4D:5E",
    "00:1a:2b:3c:4d:5e",
    "00-1A-2B-3C-4D-5E",
    "00:1A:2B:3C:4D",
    "GG:1A:2B:3C:4D:5E"
]

muster = r"(?:[0-9A-Fa-f]{2}:){5}[0-9A-Fa-f]{2}"

for mac in mac_adressen:
    print(f"{mac!r}: {bool(re.fullmatch(muster, mac))}")
````

## Aufgabe 7: Passwortregeln

Das Passwort benötigt:

- mindestens 8 Zeichen
- einen Großbuchstaben
- einen Kleinbuchstaben
- eine Zahl
- ein Sonderzeichen

Jede Regel wird einzeln geprüft. Das ist leichter nachzuvollziehen als ein sehr langer Regex.


````python
import re

passwort = input("Bitte ein Passwort eingeben: ")

mindestens_acht_zeichen = bool(re.search(r".{8,}", passwort))
enthaelt_grossbuchstaben = bool(re.search(r"[A-Z]", passwort))
enthaelt_kleinbuchstaben = bool(re.search(r"[a-z]", passwort))
enthaelt_zahl = bool(re.search(r"[0-9]", passwort))
enthaelt_sonderzeichen = bool(re.search(r"[^A-Za-z0-9]", passwort))

print("Mindestens 8 Zeichen:", mindestens_acht_zeichen)
print("Großbuchstabe:", enthaelt_grossbuchstaben)
print("Kleinbuchstabe:", enthaelt_kleinbuchstaben)
print("Zahl:", enthaelt_zahl)
print("Sonderzeichen:", enthaelt_sonderzeichen)

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
    and enthaelt_sonderzeichen
):
    print("Das Passwort erfüllt alle Regeln.")
else:
    print("Das Passwort erfüllt nicht alle Regeln.")
````
