## String-Grundlagen

Strings sind Zeichenketten und werden in Anführungszeichen geschrieben.

Mit `+` können Strings miteinander verbunden werden.


````python
text_1 = "Hallo"
text_2 = "Welt"

print(text_1 + " " + text_2)
````

## Escaping und Anführungszeichen

Enthält ein String selbst Anführungszeichen, können unterschiedliche Anführungszeichen oder ein Backslash verwendet werden.


````python
satz_1 = 'Hans sagt: "Hallo"'
satz_2 = "Hans sagt: 'Hallo'"
satz_3 = "Hans sagt: \"Hallo\""

print(satz_1)
print(satz_2)
print(satz_3)
````

## Steuerzeichen und Raw Strings

- `\n` – Zeilenumbruch
- `\t` – Tabulator
- `\\` – Backslash

Bei Windows-Pfaden ist ein Raw String praktisch.


````python
print("Hallo\nWelt")
print("Name:\tMax")

pfad = r"c:\windows\tmp"
print(pfad)
````

## Mehrzeilige Strings

Dreifache Anführungszeichen erzeugen einen mehrzeiligen String.

Sie sind kein echter Kommentar, sondern weiterhin ein String.


````python
gedicht = '''Rosen sind rot,
Veilchen sind blau,
Python macht Spaß,
das weißt du genau!'''

print(gedicht)
````

## f-Strings

Mit f-Strings können Variablen direkt in einen Text eingesetzt werden.

Mit `.2f` werden zwei Nachkommastellen angezeigt.

Mit `>10` wird rechtsbündig formatiert.


````python
name = "Max"
stadt = "Würzburg"
alter = 49
preis = 9.99875

print(f"{name} wohnt in {stadt} und ist {alter} Jahre alt.")
print(f"Preis: {preis:.2f} Euro")
print(f"{preis:>10.2f}")
````

## `format()`

`format()` ist eine ältere Möglichkeit zur String-Formatierung.


````python
vorname = "Max"
stadt = "Würzburg"
alter = 49

satzmaske = "{name} ist {alter} Jahre alt und wohnt in {city}."

ergebnis = satzmaske.format(
    name=vorname,
    alter=alter,
    city=stadt
)

print(ergebnis)
````

## String-Methoden

Wichtige Methoden:

- `upper()`
- `lower()`
- `capitalize()`
- `title()`
- `strip()`
- `replace()`
- `split()`
- `find()`
- `startswith()`
- `endswith()`


````python
mein_string = "    hello mein, World!-------"

print("Länge:", len(mein_string))
print("Groß:", mein_string.upper())
print("Klein:", mein_string.lower())
print("Titel:", mein_string.title())
print("Bereinigt:", mein_string.strip().strip("-"))
print("Ersetzt:", mein_string.replace("World", "Mars"))

zeichenkette = "Apfel,Birne,Banane"
obst = zeichenkette.split(",")

print(obst)
print(" | ".join(obst))
````

## Prüffunktionen für Strings

- `isupper()`
- `islower()`
- `isdigit()`
- `isalpha()`
- `isalnum()`
- `isidentifier()`


````python
test = "id_123"

print("isalnum:", test.isalnum())
print("isdigit:", test.isdigit())
print("isidentifier:", test.isidentifier())
print("isalpha:", test.isalpha())
````

## Slicing

Mit Slicing können Teile eines Strings ausgewählt werden.

Aufbau:

```python
text[start:ende:schrittweite]
```

Mit `[::-1]` wird ein String umgedreht.


````python
text = "Hallo, Welt!"

print(text[0])
print(text[-2])
print(text[:6])
print(text[7:])
print(text[::-1])

buchstaben = "ABCDEFGHIJ"
print(buchstaben[8:1:-2])
````

## Aufgabe: Passwort analysieren

Prüfe, ob das Passwort:

- mindestens 8 Zeichen lang ist
- einen Großbuchstaben enthält
- einen Kleinbuchstaben enthält
- eine Zahl enthält

`any()` liefert `True`, sobald mindestens ein passendes Zeichen gefunden wurde.


````python
passwort = "MeinGeheimes&§123"

mindestens_acht_zeichen = len(passwort) >= 8
enthaelt_grossbuchstaben = any(zeichen.isupper() for zeichen in passwort)
enthaelt_kleinbuchstaben = any(zeichen.islower() for zeichen in passwort)
enthaelt_zahl = any(zeichen.isdigit() for zeichen in passwort)

print("Mindestens 8 Zeichen:", mindestens_acht_zeichen)
print("Großbuchstabe:", enthaelt_grossbuchstaben)
print("Kleinbuchstabe:", enthaelt_kleinbuchstaben)
print("Zahl:", enthaelt_zahl)

if (
    mindestens_acht_zeichen
    and enthaelt_grossbuchstaben
    and enthaelt_kleinbuchstaben
    and enthaelt_zahl
):
    print("Das Passwort erfüllt alle Bedingungen.")
else:
    print("Das Passwort erfüllt nicht alle Bedingungen.")
````

## Aufgabe: Wörter eines Satzes umkehren

Die Reihenfolge der Wörter soll umgedreht werden.

Vorgehen:

1. Satz mit `split()` zerlegen
2. Liste mit `[::-1]` umdrehen
3. Wörter mit `" ".join()` zusammensetzen


````python
text = "Hallo du schöne Sonne"

woerter = text.split()
umgekehrte_woerter = woerter[::-1]
ergebnis = " ".join(umgekehrte_woerter)

print(ergebnis)
````
