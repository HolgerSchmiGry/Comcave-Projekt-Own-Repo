## Doppelpunkt, Einrückung und Codeblöcke

Alles, was eingerückt unter dem `if` steht, gehört zu dieser Bedingung.

Python nimmt Einrückungen ernst. Das ist nicht nur für die Optik.


Ein **Codeblock** ist eine Gruppe zusammengehörender Anweisungen.

In Python wird ein solcher Block durch gleichmäßige Einrückung sichtbar gemacht. Ein Doppelpunkt kündigt bei Anweisungen wie `if`, `else`, `for`, `while` oder `def` meist an, dass danach ein eingerückter Codeblock folgt.

> **Kontrollstrukturen** steuern, welche Anweisungen ausgeführt oder wiederholt werden. Dazu gehören zum Beispiel `if`, `for` und `while`.


````python
alter = 20

if alter >= 18:
    print("Diese Zeile gehört zur Bedingung.")
    print("Diese hier auch.")

print("Diese Zeile kommt danach.")
````

````python
alter = 20

if alter >= 18:
    print("volljährig")
````

Hier kündigt der Doppelpunkt an:

> Jetzt beginnt der Codeblock, der zur Bedingung gehört.

Die eingerückte Zeile wird nur ausgeführt, wenn die Bedingung wahr ist.

```python
if alter >= 18:
print("volljährig")
```

Dadurch entsteht ein `IndentationError`, weil der eingerückte Block fehlt.

Richtig:

````python
if alter >= 18:
    print("volljährig")
````

````python
alter = 20

if alter >= 18:
    print("volljährig")
    print("Zutritt erlaubt")

print("Programm beendet")
````

Die letzte Zeile ist nicht eingerückt und gehört deshalb nicht mehr zum `if`-Block.
Sie wird immer ausgeführt.

### Nicht jeder Doppelpunkt beginnt einen Codeblock

Der Doppelpunkt hat auch andere Aufgaben.

### Doppelpunkt im Dictionary

Im Dictionary trennt der Doppelpunkt einen Schlüssel von seinem Wert:

````python
person = {
    "name": "Susi",
    "alter": 30
}

print(person)
````

Hier bedeutet:

```python
"name": "Susi"
```

- `"name"` ist der Schlüssel
- `"Susi"` ist der Wert

Der Doppelpunkt beginnt hier keinen neuen Codeblock.

### Doppelpunkt beim Slicing

Beim Slicing trennt der Doppelpunkt Start, Ende und gegebenenfalls Schrittweite:

````python
zahlen = [1, 2, 3, 4, 5]

print(zahlen[1:4])
print(zahlen[::-1])
````

Bedeutung:

```python
zahlen[1:4]
```

> Von Index `1` bis vor Index `4`

```python
zahlen[::-1]
```

> Die komplette Liste rückwärts

### Sauberer Merksatz

> Ein Doppelpunkt leitet bei Kontrollstrukturen, Funktionen und Klassen meist einen eingerückten Codeblock ein.  
> In Dictionaries und beim Slicing hat er eine andere Bedeutung.

<a id="grenzwerte"></a>

### Bedingungen lesbar formulieren und Grenzwerte richtig prüfen

Eine Bedingung sollte möglichst so formuliert sein, dass der **gewünschte oder aktive Fall** direkt im `if`-Block steht.

Das macht den Code leichter zu lesen und später einfacher zu verstehen.

Beispiel:

> Ab 10 Dateien sollen die Daten kopiert werden.

````python

if dateien >= 10:
    print("Kopiere Daten.")
else:
    print("Die Menge von 10 Dateien ist noch nicht erreicht.")
````

Hier steht der gewünschte Fall direkt im `if`:

```python
dateien >= 10
```

Das bedeutet:

> Die Anzahl der Dateien ist größer als oder gleich 10.

Bei `dateien = 10` ist die Bedingung also wahr.

### Warum `>= 10` und nicht `> 10`?

Die Operatoren unterscheiden sich an der Grenze:

| Ausdruck | Bedeutung | Ist bei `dateien = 10` wahr? |
|---|---|---:|
| `dateien > 10` | mehr als 10 | Nein |
| `dateien >= 10` | mindestens 10 | Ja |
| `dateien < 10` | weniger als 10 | Nein |
| `dateien <= 10` | höchstens 10 | Ja |

Wenn bereits bei genau 10 Dateien kopiert werden soll, ist `>= 10` richtig.

### Das gleiche Beispiel andersherum formuliert

Man könnte auch zuerst den negativen Fall prüfen:

````python
dateien = 10

if dateien < 10:
    print("Die Menge von 10 Dateien ist noch nicht erreicht.")
else:
    print("Kopiere Daten.")
````

Diese Variante funktioniert ebenfalls.

Sie ist aber oft weniger direkt, weil der eigentliche gewünschte Fall erst im `else` steht.

Für einen späteren Leser ist meistens diese Form schneller verständlich:

```python
if dateien >= 10:
    print("Kopiere Daten.")
```

★ **Merksatz**

> Prüfe möglichst direkt den Fall, der eintreten soll.

### Wann ist die umgekehrte Form trotzdem sinnvoll?

Manchmal ist der Fehlerfall wichtiger und soll zuerst beendet werden.

Beispiel:

````python
dateien = 5

if dateien < 10:
    print("Noch nicht genügend Dateien vorhanden.")
else:
    print("Kopiervorgang wird gestartet.")
````

Hier ist der Hinweis auf die fehlenden Dateien möglicherweise die wichtigste Information.

Es gibt also nicht nur eine erlaubte Schreibweise. Entscheidend sind:

- fachlich richtige Grenzwerte
- verständliche Formulierung
- klarer Zweck der Bedingung

### Typischer Denkfehler an der Grenze

Angenommen, der Kopiervorgang soll **ab 10 Dateien** starten.

Dann wäre diese Bedingung falsch:

````python
dateien = 10

if dateien > 10:
    print("Kopiere Daten.")
else:
    print("Noch nicht genügend Dateien vorhanden.")
````

Bei genau 10 Dateien ist `dateien > 10` falsch.

Die Ausgabe wäre deshalb:

```text
Noch nicht genügend Dateien vorhanden.
```

Das widerspricht der Aufgabenstellung „ab 10 Dateien“.

Richtig ist:

```python
dateien >= 10
```

### Logische Operatoren: `and`, `or` und `not`

Mit logischen Operatoren werden mehrere Bedingungen miteinander verbunden oder umgekehrt.

Python kennt dafür vor allem:

- `and`
- `or`
- `not`

## Logische Operatoren

`and` liefert nur dann `True`, wenn beide Bedingungen wahr sind.

````python
alter = 20
hat_ticket = True

if alter >= 18 and hat_ticket:
    print("Einlass erlaubt")
else:
    print("Einlass nicht erlaubt")
````

Wahrheitstabelle für `and`:

| Bedingung A | Bedingung B | Ergebnis |
|---:|---:|---:|
| `True` | `True` | `True` |
| `True` | `False` | `False` |
| `False` | `True` | `False` |
| `False` | `False` | `False` |

★ **Merksatz**

> Bei `and` müssen alle verbundenen Bedingungen wahr sein.

### `or` – mindestens eine Bedingung muss wahr sein

`or` liefert `True`, wenn mindestens eine der Bedingungen wahr ist.

````python
wochentag = "Samstag"

if wochentag == "Samstag" or wochentag == "Sonntag":
    print("Wochenende")
else:
    print("Werktag")
````

Wahrheitstabelle für `or`:

| Bedingung A | Bedingung B | Ergebnis |
|---:|---:|---:|
| `True` | `True` | `True` |
| `True` | `False` | `True` |
| `False` | `True` | `True` |
| `False` | `False` | `False` |

★ **Merksatz**

> Bei `or` reicht eine wahre Bedingung aus.

### `not` – Wahrheitswert umkehren

`not` kehrt einen Wahrheitswert um:

- aus `True` wird `False`
- aus `False` wird `True`

````python
eingeloggt = False

if not eingeloggt:
    print("Bitte anmelden")
else:
    print("Willkommen")
````

Wahrheitstabelle für `not`:

| Ausgangswert | Ergebnis |
|---:|---:|
| `True` | `False` |
| `False` | `True` |

### Mehrere Operatoren kombinieren

Bedingungen können auch kombiniert werden:

````python
alter = 20
hat_ticket = False
ist_mitarbeiter = True

if (alter >= 18 and hat_ticket) or ist_mitarbeiter:
    print("Einlass erlaubt")
else:
    print("Einlass nicht erlaubt")
````

Die Klammern machen sichtbar, was zusammengehört:

1. `alter >= 18 and hat_ticket`
2. Ergebnis davon `or ist_mitarbeiter`

Ohne Klammern wäre der Ausdruck zwar ebenfalls auswertbar, aber für Menschen schwerer zu lesen.

### Reihenfolge der logischen Operatoren

Python wertet logische Operatoren in dieser Reihenfolge aus:

1. `not`
2. `and`
3. `or`

Beispiel:

````python
ergebnis = not False and True or False

print(ergebnis)
````

Für Anfänger ist es trotzdem besser, zusammengehörende Teile mit Klammern zu kennzeichnen:

```python
ergebnis = (not False and True) or False
```

### Kurzschlussauswertung

Python prüft nur so viel wie nötig.

Bei `and`:

```python
False and weitere_bedingung
```

Sobald der erste Teil `False` ist, kann das Gesamtergebnis nicht mehr `True` werden.

Bei `or`:

```python
True or weitere_bedingung
```

Sobald der erste Teil `True` ist, bleibt das Gesamtergebnis `True`.

Dieses Verhalten nennt man **Kurzschlussauswertung**.

````python
zahl = 0

if zahl != 0 and 10 / zahl > 2:
    print("Bedingung erfüllt")
else:
    print("Bedingung nicht erfüllt")
````

Obwohl `10 / zahl` bei `zahl = 0` eigentlich einen Fehler erzeugen würde, tritt hier kein Fehler auf.

Der erste Teil ist bereits `False`:

```python
zahl != 0
```

Deshalb prüft Python den zweiten Teil nicht mehr.

### Häufiger Anfängerfehler bei `or`

Falsch:

````python
tag = "Samstag"

if tag == "Samstag" or "Sonntag":
    print("Wochenende")
````

Diese Bedingung ist praktisch immer wahr, weil der Text `"Sonntag"` selbst als wahr gilt.

Richtig:

````python
tag = "Samstag"

if tag == "Samstag" or tag == "Sonntag":
    print("Wochenende")
````

Noch lesbarer mit `in`:

````python
tag = "Samstag"

if tag in ("Samstag", "Sonntag"):
    print("Wochenende")
````

## elif-Kaskaden

Mit `if`, `elif` und `else` können mehrere Möglichkeiten geprüft werden.

Python arbeitet die Bedingungen **von oben nach unten** ab:

1. `if` wird zuerst geprüft.
2. Ist `if` falsch, wird das erste `elif` geprüft.
3. Weitere `elif`-Bedingungen folgen nacheinander.
4. `else` wird ausgeführt, wenn vorher keine Bedingung wahr war.

Sobald eine Bedingung wahr ist, wird der zugehörige Block ausgeführt.  
Die restliche Kaskade wird danach übersprungen.

★ **Merksatz**

> Eine `elif`-Kaskade endet beim ersten Treffer.

### Beispiel: Zwei Zahlen vergleichen

Hier gibt es genau drei mögliche Fälle:

- `a` ist kleiner als `b`
- `a` ist gleich `b`
- `a` ist größer als `b`

````python
a = 5
b = 10

if a < b:
    print("a ist kleiner als b")
elif a == b:
    print("a ist gleich b")
else:
    print("a ist größer als b")
````

Bei `a = 5` und `b = 10` ist bereits die erste Bedingung wahr:

```python
a < b
```

Deshalb wird ausgegeben:

```text
a ist kleiner als b
```

`elif` und `else` werden nicht mehr geprüft.

### Warum heißt es Kaskade?

Eine Kaskade besteht aus mehreren aufeinanderfolgenden Prüfungen:

```text
if
  ↓ falsch
elif
  ↓ falsch
elif
  ↓ falsch
else
```

Man kann sich das wie eine Treppe vorstellen: Python geht nur so weit nach unten, bis eine passende Bedingung gefunden wurde.

### Beispiel: Altersfreigabe

Bei Grenzwerten muss die größte Grenze zuerst geprüft werden.

````python
alter = 17

if alter >= 18:
    print("Film ab 18 erlaubt")
elif alter >= 16:
    print("Film ab 16 erlaubt")
elif alter >= 12:
    print("Film ab 12 erlaubt")
elif alter >= 6:
    print("Film ab 6 erlaubt")
else:
    print("TV aus")
````

Bei `alter = 17` passiert Folgendes:

```text
alter >= 18  → falsch
alter >= 16  → wahr
```

Danach stoppt die Kaskade. Die Bedingungen `>= 12` und `>= 6` werden nicht mehr geprüft.

### Warum muss die höchste Grenze zuerst stehen?

Diese Reihenfolge wäre falsch:

````python
alter = 17

if alter >= 6:
    print("Film ab 6 erlaubt")
elif alter >= 12:
    print("Film ab 12 erlaubt")
elif alter >= 16:
    print("Film ab 16 erlaubt")
````

Eine 17-jährige Person erfüllt bereits:

```python
alter >= 6
```

Python würde deshalb sofort `"Film ab 6 erlaubt"` ausgeben und die höheren Altersstufen niemals erreichen.

★ **Merksatz**

> Bei überlappenden Zahlenbereichen beginnt eine `elif`-Kaskade meistens mit der höchsten oder speziellsten Bedingung.

### Mehrere einzelne `if` oder eine `elif`-Kaskade?

Mehrere einzelne `if`-Anweisungen werden unabhängig voneinander geprüft:

````python
zahl = 12

if zahl > 0:
    print("Die Zahl ist positiv")

if zahl % 2 == 0:
    print("Die Zahl ist gerade")
````

Hier sind beide Aussagen wahr. Deshalb werden beide Ausgaben angezeigt.

Eine `elif`-Kaskade prüft dagegen zusammengehörende Alternativen:

````python
zahl = 12

if zahl < 0:
    print("negativ")
elif zahl == 0:
    print("null")
else:
    print("positiv")
````

★ **Unterschied**

- mehrere `if` → mehrere Blöcke können ausgeführt werden
- `if-elif-else` → nur der erste passende Block wird ausgeführt

### Verschachtelte Bedingungen

Eine Bedingung kann innerhalb einer anderen Bedingung stehen.

Das nennt man eine **verschachtelte Bedingung**.

````python
alter = 20
hat_ticket = True

if alter >= 18:
    if hat_ticket:
        print("Einlass erlaubt")
    else:
        print("Kein Einlass: Ticket fehlt")
else:
    print("Kein Einlass: zu jung")
````

Der innere `if`-Block wird nur erreicht, wenn die äußere Bedingung `alter >= 18` wahr ist.

Verschachtelungen sind manchmal sinnvoll. Zu viele Ebenen machen Code jedoch schwer lesbar. Häufig lässt sich eine einfache Prüfung auch mit `and` ausdrücken:

````python
alter = 20
hat_ticket = True

if alter >= 18 and hat_ticket:
    print("Einlass erlaubt")
else:
    print("Einlass nicht erlaubt")
````

<a id="identity"></a>

### Identity Operator: `is` und `is not`

Der Identity Operator prüft, ob zwei Variablen auf **dasselbe Objekt im Speicher** zeigen.

Dafür gibt es:

- `is`
- `is not`

★ **Merksatz**

> `==` vergleicht Werte.  
> `is` vergleicht Identität.

### Beispiel 1: Zwei Variablen zeigen auf dieselbe Liste

````python
meine_liste = ["Hallo", "Welt"]
meine_liste_2 = meine_liste

print(meine_liste == meine_liste_2)
print(meine_liste is meine_liste_2)
````

Beide Ausgaben sind `True`.

Warum?

- Die Inhalte sind gleich.
- Beide Variablen zeigen auf dasselbe Listenobjekt.

````python
meine_liste_2.append("Test")

print("meine_liste:", meine_liste)
print("meine_liste_2:", meine_liste_2)
````

Beide Listen zeigen danach `"Test"`, weil keine Kopie erzeugt wurde.

Bildlich:

```text
meine_liste   ─┐
               ├──> ["Hallo", "Welt", "Test"]
meine_liste_2 ─┘
```

### Beispiel 2: Gleicher Inhalt, aber zwei verschiedene Listen

````python
testliste_1 = ["Hallo", "Welt"]
testliste_2 = ["Hallo", "Welt"]

print(testliste_1 == testliste_2)
print(testliste_1 is testliste_2)
````

Ergebnis:

```text
True
False
```

Die Inhalte sind gleich, aber es sind zwei getrennte Listenobjekte.

### Präzise Formulierung

Statt:

> „Beide Listen sind gleich.“

sollte man unterscheiden:

- `==` → Die Inhalte sind gleich.
- `is` → Beide Variablen zeigen auf dasselbe Objekt.

### Eine echte Kopie erstellen

Mit `.copy()` entsteht eine unabhängige Liste:

````python
original = ["Hallo", "Welt"]
kopie = original.copy()

kopie.append("Test")

print("Original:", original)
print("Kopie:", kopie)
print("Gleicher Inhalt:", original == kopie)
print("Dasselbe Objekt:", original is kopie)
````

Das Original bleibt unverändert.

### Vorsicht bei Zahlen und Strings

Diese Schreibweise ist fachlich falsch:

````python
a = 5
b = 5

print(a is b)
````

Das Ergebnis kann bei kleinen Zahlen zufällig `True` sein.

Python verwendet intern teilweise bereits vorhandene Objekte erneut. Dieses Verhalten nennt man **Interning** beziehungsweise **Caching**.

Deshalb darf `is` nicht verwendet werden, um Zahlen oder Strings inhaltlich zu vergleichen.

Richtig:

````python
a = 5
b = 5

print(a == b)
````

Auch bei Strings gilt:

```python
text_1 == text_2
```

prüft den Inhalt.

```python
text_1 is text_2
```

prüft nur, ob beide Variablen auf dasselbe Objekt zeigen.

### Ist `is` nur für komplexe Datentypen?

Nein.

`is` kann grundsätzlich bei beliebigen Python-Objekten verwendet werden.

Die entscheidende Frage ist aber:

> Möchte ich den Wert vergleichen oder die Identität?

In der Praxis wird `is` besonders häufig verwendet für:

```python
wert is None
wert is not None
```

Für normale Wertevergleiche verwendet man fast immer `==` oder `!=`.

### Übersicht

| Prüfung | Bedeutung |
|---|---|
| `a == b` | Haben beide denselben Wert? |
| `a != b` | Haben beide unterschiedliche Werte? |
| `a is b` | Zeigen beide auf dasselbe Objekt? |
| `a is not b` | Zeigen beide auf verschiedene Objekte? |
