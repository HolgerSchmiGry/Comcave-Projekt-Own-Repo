## 1. Was ist eine Funktion?

Eine Funktion ist ein benannter Codeblock.

Sie wird erst ausgeführt, wenn sie aufgerufen wird.


````python
def begruessung():
    print("Hallo!")

begruessung()
````

`def` bedeutet sinngemäß:

> Hier wird eine Funktion definiert.

`begruessung()` führt die Funktion aus.


## 2. Parameter und Argumente

Eine Funktion kann Werte entgegennehmen.


````python
def begruesse_person(name):
    print(f"Hallo {name}!")

begruesse_person("Max")
````

Dabei gilt:

- `name` ist der Parameter
- `"Max"` ist das Argument


## 3. Mehrere Parameter


````python
def beschreibe_person(name, alter, stadt):
    print(f"{name} ist {alter} Jahre alt und lebt in {stadt}.")

beschreibe_person("Max", 25, "Berlin")
````

Die Reihenfolge muss bei normalen Positionsargumenten stimmen.


## 4. Rückgabewerte mit `return`

Eine Funktion kann ein Ergebnis zurückgeben.


````python
def addieren(zahl_1, zahl_2):
    ergebnis = zahl_1 + zahl_2
    return ergebnis

summe = addieren(10, 20)

print(summe)
````

`return` gibt den Wert an die Stelle zurück, an der die Funktion aufgerufen wurde.


## 5. Standardwerte

Ein Parameter kann einen Standardwert bekommen.


````python
def begruesse_person(name="Gast"):
    print(f"Hallo {name}!")

begruesse_person()
begruesse_person("Max")
````

Wird kein Argument übergeben, nutzt Python den Standardwert.


## 6. Schlüsselwortargumente

Schlüsselwortargumente werden über den Parameternamen übergeben.


````python
def beschreibe_person(name, alter, stadt):
    print(f"{name} ist {alter} Jahre alt und lebt in {stadt}.")

beschreibe_person(
    alter=25,
    stadt="Berlin",
    name="Max"
)
````

Die Reihenfolge ist hier egal, weil jeder Wert eindeutig einem Namen zugeordnet wird.


# Teil 2: Packing und Unpacking

Packing bedeutet:

> Mehrere Werte werden zusammengepackt.

Unpacking bedeutet:

> Eine Sammlung wird wieder in einzelne Werte zerlegt.


## 7. Einfaches Unpacking eines Tupels


````python
person = ("Max", 25, "Berlin")

name, alter, stadt = person

print(name)
print(alter)
print(stadt)
````

Die drei Werte werden auf drei Variablen verteilt.


## 8. Anzahl muss passen

Das funktioniert nur, wenn Anzahl der Variablen und Werte zusammenpassen.


````python
# Dieser Code würde einen Fehler verursachen:
#
# person = ("Max", 25, "Berlin")
# name, alter = person
````

Drei Werte können nicht ohne Weiteres auf nur zwei Variablen verteilt werden.


## 9. Unpacking mit `*`

Mit `*` können mehrere Werte zusammengefasst werden.


````python
zahlen = [10, 20, 30, 40, 50]

erste, *mitte, letzte = zahlen

print("Erste:", erste)
print("Mitte:", mitte)
print("Letzte:", letzte)
````

`mitte` wird hier zu einer Liste.


## 10. Liste beim Funktionsaufruf entpacken


````python
def addieren(zahl_1, zahl_2, zahl_3):
    return zahl_1 + zahl_2 + zahl_3

zahlen = [10, 20, 30]

ergebnis = addieren(*zahlen)

print(ergebnis)
````

`*zahlen` entpackt die Liste.

Aus:

```python
addieren(*zahlen)
```

wird sinngemäß:

```python
addieren(10, 20, 30)
```


## 11. Dictionary mit `**` entpacken


````python
def beschreibe_person(name, alter, stadt):
    print(f"{name} ist {alter} Jahre alt und lebt in {stadt}.")

daten = {
    "name": "Max",
    "alter": 25,
    "stadt": "Berlin"
}

beschreibe_person(**daten)
````

`**daten` entpackt Schlüssel und Werte.

Die Schlüssel im Dictionary müssen zu den Parameternamen passen.


# Teil 3: `*args`

Mit `*args` kann eine Funktion beliebig viele Positionsargumente aufnehmen.


## 12. Einfaches Beispiel


````python
def werte_anzeigen(*args):
    print(args)

werte_anzeigen(10, 20, 30)
````

Innerhalb der Funktion ist `args` ein Tupel.


````python
def werte_anzeigen(*args):
    print(type(args))

werte_anzeigen(10, 20, 30)
````

## 13. Mit `*args` rechnen


````python
def addieren(*args):
    summe = 0

    for zahl in args:
        summe += zahl

    return summe

print(addieren(10, 20))
print(addieren(10, 20, 30, 40))
````

Die Funktion kann unterschiedlich viele Zahlen bekommen.


## 14. Der Name `args` ist nur eine Gewohnheit

Technisch geht auch:


````python
def ausgeben(*werte):
    print(werte)

ausgeben(1, 2, 3)
````

Entscheidend ist der Stern `*`, nicht der Name `args`.


# Teil 4: `**kwargs`

Mit `**kwargs` kann eine Funktion beliebig viele Schlüsselwortargumente aufnehmen.


## 15. Einfaches Beispiel


````python
def daten_anzeigen(**kwargs):
    print(kwargs)

daten_anzeigen(
    name="Max",
    alter=25,
    stadt="Berlin"
)
````

Innerhalb der Funktion ist `kwargs` ein Dictionary.


````python
def daten_anzeigen(**kwargs):
    print(type(kwargs))

daten_anzeigen(name="Max", alter=25)
````

## 16. Auf Werte zugreifen


````python
def person_anzeigen(**kwargs):
    print("Name:", kwargs["name"])
    print("Alter:", kwargs["alter"])

person_anzeigen(name="Max", alter=25)
````

## 17. Sicherer Zugriff mit `get()`


````python
def person_anzeigen(**kwargs):
    name = kwargs.get("name", "Unbekannt")
    alter = kwargs.get("alter", "nicht angegeben")

    print("Name:", name)
    print("Alter:", alter)

person_anzeigen(name="Max")
````

`get()` kann einen Standardwert liefern, wenn ein Schlüssel fehlt.


## 18. `*args` und `**kwargs` zusammen


````python
def alles_anzeigen(*args, **kwargs):
    print("Positionsargumente:", args)
    print("Schlüsselwortargumente:", kwargs)

alles_anzeigen(
    10,
    20,
    name="Max",
    alter=25
)
````

> **Merksatz:**  
> `*args` sammelt Positionsargumente in einem Tupel.  
> `**kwargs` sammelt Schlüsselwortargumente in einem Dictionary.


## 19. Reihenfolge in einer Funktionsdefinition

Eine typische Reihenfolge sieht so aus:

```python
def funktion(normal, standard="Wert", *args, **kwargs):
    ...
```

Nicht jede Funktion braucht alle vier Arten.

Schreibe eine Funktion, die zuerst alle Positionsargumente und danach alle Schlüsselwortargumente ausgibt.

### Musterlösung
